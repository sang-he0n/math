> ## **크래머 공식(Cramer's Rule)**

#### **(1) 정의** : 행렬식을 이용하여 선형 시스템 $ A\textbf{x} = \textbf{b} $의 특정 변수 $ \, x_{i} $의 해를 구하는 방법
#### $ \Rrightarrow{} x_{i} = \frac{\det{}(A_{i}(\textbf{b}))}{\det{}(A)} \;\;\ \text{ where } \,A_{i}(\textbf{b}) = \begin{bmatrix} \textbf{a}_{1} & \cdots{} & \textbf{a}_{i-1} & \textbf{b} & \textbf{a}_{i+1} & \cdots{} & \textbf{a}_{n} \end{bmatrix} $

##### **(`WHY?`)**
##### $ \hspace{0.15cm} A I_{i}(\textbf{x}) = A \begin{bmatrix} \textbf{e}_{1}&\cdots{}&\textbf{e}_{i-1}&\textbf{x}&\textbf{e}_{i-1}&\cdots{}&\textbf{e}_{n} \end{bmatrix} $
##### $ \hspace{1.25cm} = \begin{bmatrix} A\textbf{e}_{1}&\cdots{}&A\textbf{e}_{i-1}&A\textbf{x}&A\textbf{e}_{i-1}&\cdots{}&A\textbf{e}_{n} \end{bmatrix} $
##### $ \hspace{1.25cm}= \begin{bmatrix} \textbf{a}_{1}&\cdots{}&\textbf{a}_{i-1}&A\textbf{x}&\textbf{a}_{i+1}&\cdots{}&\textbf{a}_{n} \end{bmatrix} $
##### $ \hspace{1.25cm} = A_{i}(\textbf{b}) $
##### $ \hspace{0.15cm} \det{}(A I_{i}(\textbf{x})\big){} = \det{}(A) \times{} \det{}\big({}I_{i}(\textbf{x})\big){} = \det{}(A) \times{} x_{i} = \det{}(A_{i}(\textbf{b})\big){} \;\; \because{} $ Thoerem 3-2
##### $ \hspace{0.15cm} \therefore{} x_{i} = \frac{\det{}(A_{i}(\textbf{b}))}{\det{}(A)} \;\; $ (Theorem 3-7)

#### **(2) 예시** : 
##### $ \hspace{0.15cm} $ ① 아래 선형 시스템의 해를 크래머 공식을 이용하여 구하고자 할 때
##### $ \hspace{0.45cm} \begin{cases} 3x_{1} - 2x_{2} &= 6 \\ -5x_{1} + 4x_{2} &= 8 \end{cases} \; \rightarrow{} \; A\textbf{x} = \textbf{b} \; \rightarrow{} \; \begin{bmatrix} 3 & -2 \\ -5 & 4 \end{bmatrix} \begin{bmatrix} x_{1} \\ x_{2} \end{bmatrix} = \begin{bmatrix} 6 \\ 8 \end{bmatrix} $
##### $ \hspace{0.45cm} A = \begin{bmatrix} 3 & -2 \\ -5 & 4 \end{bmatrix}, \;\; \det{}(A) = 2 $
##### $ \hspace{0.45cm} A_{1}(\textbf{b}) = \begin{bmatrix} 6&-2 \\ 8&4 \end{bmatrix}, \;\; \det{}\big({}A_{1}(\textbf{b})\big){} = 40 $ 
##### $ \hspace{0.45cm} A_{2}(\textbf{b}) = \begin{bmatrix} 3&6 \\ -5&8 \end{bmatrix}, \;\; \det{}\big({}A_{2}(\textbf{b})\big){} = 54 $ 
##### $ \hspace{0.45cm} \therefore{} x_{1} = \frac{\det{}(A_{1}(\textbf{b}))}{\det{}(A)} = \frac{40}{2} = 20, \;\; x_{2} = \frac{\det{}(A_{2}(\textbf{b}))}{\det{}(A)} = \frac{54}{2} = 27 $ 
##### $ \hspace{0.15cm} $ ② **[LATEX]**

##### **(`PLUS`)** 크래머 공식은 행렬의 크기(shape)가 커질수록 계산이 비효율적임

> ## (Theorem 3-8) **an Inverse Formular**

#### 어떠한 정방행렬의 역행렬을 아래와 같이 일반화할 수 있다.
#### $ \Rrightarrow{} A^{-1} = \frac{1}{\det{}(A)} \begin{bmatrix} C_{1,1}&C_{1,2}&\cdots{}&C_{1,n} \\ C_{2,1}&C_{2,2}&\cdots{}&C_{2,n} \\ \vdots{}&\vdots{}&\ddots{}&\vdots{} \\ C_{n,1}&C_{n,2}&\cdots{}&C_{n,n} \end{bmatrix}^{T} = \frac{1}{\det{}(A)} \begin{bmatrix} C_{1,1}&C_{2,1}&\cdots{}&C_{n,1} \\ C_{1,2}&C_{2,2}&\cdots{}&C_{n,1} \\ \vdots{}&\vdots{}&\ddots{}&\vdots{} \\ C_{1,n}&C_{2,n}&\cdots{}&C_{n,n} \end{bmatrix} $

##### **(`WHY?`)**
##### $ \hspace{0.15cm} \text{if } \,A\textbf{x} = \textbf{e}_{k}, \;\; \textbf{x} = A^{-1}\textbf{e}_{k} = \textbf{a}^{-1}_{k} $
##### $ \hspace{0.3cm} (i$-$k) \, \text{ entry of } \, A^{-1} = a^{-1}_{i,k} = \frac{\det{}(A_{i}(\textbf{e}_{k})\big){}}{\det{}(A)} = \frac{\bf{C_{k,i}}}{\det{}(A)} \;\; $ ($ \text{not} \, C_{i,k} $)
##### $ \hspace{0.3cm} \therefore{} A^{-1} = \begin{bmatrix} a^{-1}_{1,1} & a^{-1}_{1,2} & \cdots{} & a^{-1}_{1,n} \\ a^{-1}_{2,1} & a^{-1}_{2,2} & \cdots{} & a^{-1}_{2,n} \\ \vdots{} & \vdots{} & \ddots{} & \vdots{} \\ a^{-1}_{n,1} & a^{-1}_{n,2} & \cdots{} & a^{-1}_{n,n} \end{bmatrix} = \frac{1}{\det{}(A)}\begin{bmatrix} C_{1,1} & C_{2,1} & \cdots{} & C_{n,1} \\ C_{1,2} & C_{2,2} & \cdots{} & C_{n,1} \\ \vdots{} & \vdots{} & \ddots{} & \vdots{} \\ C_{1,n} & C_{2,n} & \cdots{} & C_{nn} \end{bmatrix} $

##### **(`PLUS`)** 여인수로 이루어진 행렬의 전치 행렬을 수반행렬(adjoint matrix of $ A ; \text{adj}(A) $)라고 정의함
##### $ \hspace{0.15cm} \Rrightarrow{} \text{adj}(A) = \begin{bmatrix} C_{1,1}&C_{1,2}&\cdots{}&C_{1,n} \\ C_{2,1}&C_{2,2}&\cdots{}&C_{2,n} \\ \vdots{}&\vdots{}&\ddots{}&\vdots{} \\ C_{n,1}&C_{n,2}&\cdots{}&C_{n,n} \end{bmatrix}^{T} = \begin{bmatrix} C_{1,1}&C_{2,1}&\cdots{}&C_{n,1} \\ C_{1,2}&C_{2,2}&\cdots{}&C_{n,2} \\ \vdots{}&\vdots{}&\ddots{}&\vdots{} \\ C_{1,n}&C_{2,n}&\cdots{}&C_{n,n} \end{bmatrix} $

> ## **행렬식을 이용한 넓이와 부피(Determinants as Area or Volume)**

#### **(1) 넓이(area)** : $ \mathbb{R}^{2} $ 공간에서 벡터 $ \textbf{v}_{1}, \textbf{v}_{2} $가 형성하는 평행사변형의 넓이는 행렬식을 이용해 계산할 수 있음
#### $ \Rrightarrow{} S = | \det{}(\begin{bmatrix} \textbf{v}_{1} & \textbf{v}_{2} \end{bmatrix}) | $

##### **(`WHY?`)** (Theorem 3-9)
##### $ \hspace{0.15cm} \cdot{} \, A = \begin{bmatrix} \textbf{v}_{1} & \textbf{v}_{2} \end{bmatrix} $ ~ $ \begin{bmatrix} \textbf{v}_{1}^{*} & \textbf{v}_{2}^{*} \end{bmatrix} $
##### $ \hspace{0.15cm} \cdot{} \, S = | \det{}(\begin{bmatrix} \textbf{v}_{1} & \textbf{v}_{2} \end{bmatrix}) | = | \det{}(\begin{bmatrix} \textbf{v}_{1}^{*} & \textbf{v}_{2}^{*} \end{bmatrix}) | $ 
#### **[GRPAH]**

#### **(2) 부피(volume)** : $ \mathbb{R}^{3} $ 공간에서 벡터 $ \textbf{v}_{1},\, \textbf{v}_{2},\, \textbf{v}_{3} $ 가 형성하는 평행육면체의 부피는 행렬식을 이용해 계산
#### $ \Rrightarrow{} V = | \det{}(\begin{bmatrix} \textbf{v}_{1} & \textbf{v}_{2} & \textbf{v}_{3} \end{bmatrix}) | $
