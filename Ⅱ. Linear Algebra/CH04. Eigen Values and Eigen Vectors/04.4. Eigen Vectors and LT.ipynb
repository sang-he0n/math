# CH04.4. **Eigen Vectors and Linear Transformation**

> ## **선형 변환의 행렬 표현(Matrix Representation of Linear Transformation)**

#### **(1) 정의** : 어떠한 좌표 벡터 $ \begin{bmatrix} \textbf{x} \end{bmatrix}_{\mathcal{B}} $와 선형 변환 $ T(\textbf{x}) $의 좌표 벡터 $ \begin{bmatrix} T(\textbf{x}) \end{bmatrix}_{\mathcal{C}} $의 관계를 파악하는 행렬로 표현하는 방법
#### $ \Rrightarrow{} \begin{bmatrix} T(\textbf{x}) \end{bmatrix}_{\mathcal{C}} := \begin{bmatrix} \begin{bmatrix} T(\textbf{b}_{1}) \end{bmatrix}_{\mathcal{C}} & \begin{bmatrix} T(\textbf{b}_{2}) \end{bmatrix}_{\mathcal{C}} & \cdots{} & \begin{bmatrix} T(\textbf{b}_{n}) \end{bmatrix}_{\mathcal{C}} \end{bmatrix} \begin{bmatrix} \textbf{x} \end{bmatrix}_{\mathcal{B}} $
#### $ \hspace{1.9cm} = M \begin{bmatrix} \textbf{x} \end{bmatrix}_{\mathcal{B}} $ 
#### $ \hspace{0.45cm} \text{ where } \, V, W \, \text{ is vector space} , \; \; T(\textbf{x}) : V \rightarrow{} W $
#### $ \hspace{0.45cm} \text{ and } \; \mathcal{B} = \{\textbf{b}_{1}, \cdots{}, \textbf{b}_{n} \} \, \text{ is basis of } \, V, \;\; \mathcal{C} \, \text{ is basis of } \, W. $ 

##### **(`WHY?`)**
##### $ \hspace{0.15cm} \text{if } \, \mathcal{B} = \{ \textbf{b}_{1}, \textbf{b}_{2}, \cdots{}, \textbf{b}_{n} \} \, \text{ is basis of } \, V, \;\; \textbf{x} = r_{1} \textbf{b}_{1} + r_{2} \textbf{b}_{2} + \cdots{} + r_{n} \textbf{b}_{n} $
##### $ \hspace{0.3cm} T(\textbf{x}) = T(r_{1} \textbf{b}_{1} + r_{2} \textbf{b}_{2} + \cdots{} + r_{n} \textbf{b}_{n}) = r_{1} T(\textbf{b}_{1}) + r_{2} T(\textbf{b}_{2}) + \cdots{} + r_{n} T(\textbf{b}_{n}) $
##### $ \hspace{0.3cm} \begin{bmatrix} T(\textbf{x}) \end{bmatrix}_{\mathcal{C}} = \begin{bmatrix} r_{1} T(\textbf{b})_{1} + r_{2} T(\textbf{b})_{2} + \cdots{} + r_{n} T(\textbf{b})_{n} \end{bmatrix}_{\mathcal{C}} $ 
##### $ \hspace{1.6cm} = r_{1} \begin{bmatrix} T(\textbf{b}_{1}) \end{bmatrix}_{\mathcal{C}} + r_{2} \begin{bmatrix} T(\textbf{b}_{2}) \end{bmatrix}_{\mathcal{C}} + \cdots{} + r_{n} \begin{bmatrix} T(\textbf{b}_{n}) \end{bmatrix}_{\mathcal{C}} $
##### $ \hspace{1.6cm} = \begin{bmatrix} \begin{bmatrix} T(\textbf{b}_{1}) \end{bmatrix}_{\mathcal{C}} & \begin{bmatrix} T(\textbf{b}_{2}) \end{bmatrix}_{\mathcal{C}} & \cdots{} & \begin{bmatrix} T(\textbf{b}_{n}) \end{bmatrix}_{\mathcal{C}} \end{bmatrix} \begin{bmatrix} r_{1} \\ r_{2} \\ \cdots{} \\ r_{n} \end{bmatrix} $
##### $ \hspace{1.6cm} = \begin{bmatrix} \begin{bmatrix} T(\textbf{b}_{1}) \end{bmatrix}_{\mathcal{C}} & \begin{bmatrix} T(\textbf{b}_{2}) \end{bmatrix}_{\mathcal{C}} & \cdots{} & \begin{bmatrix} T(\textbf{b}_{n}) \end{bmatrix}_{\mathcal{C}} \end{bmatrix} \begin{bmatrix} \textbf{x} \end{bmatrix}_\mathcal{B} $
##### $ \hspace{1.6cm} = M \begin{bmatrix} \textbf{x} \end{bmatrix}_\mathcal{B} $
##### $ \hspace{0.15cm} \therefore{} M = \begin{bmatrix} \begin{bmatrix} T(\textbf{b}_{1}) \end{bmatrix}_{\mathcal{C}} & \begin{bmatrix} T(\textbf{b}_{2}) \end{bmatrix}_{\mathcal{C}} & \cdots{} & \begin{bmatrix} T(\textbf{b}_{n}) \end{bmatrix}_{\mathcal{C}} \end{bmatrix} $

#### **(2) 기하학적 해석** :
#### **[GRAPH]**

#### **(3) 선형 변환과 좌표 벡터 간 관계** :
#### $ \begin{CD} \textbf{x} @>A>> T(\textbf{x}) \\ @VVV @AAA \\ \begin{bmatrix} \textbf{x} \end{bmatrix}_{\mathcal{B}} @>M>> \begin{bmatrix} T(\textbf{x}) \end{bmatrix}_{\mathcal{C}} \end{CD} $ 

#### **(4) 예시** :
##### $ \hspace{0.15cm} $ ① 아래 선형 변환 $ T $에 대하여 표현 행렬 $ M $을 구하면
##### $ \hspace{0.45cm} \text{if } \,V = \text{span}(\{\textbf{b}_{1}, \textbf{b}_{2} \}) = \text{span}(\mathcal{B}), \;\; W = \text{span}(\mathcal{C}) = \text{span}(\{ \textbf{c}_{1}, \textbf{c}_{2}, \textbf{c}_{3} \}) $
##### $ \hspace{0.45cm} \text{and } \; T(\textbf{x}) : V \rightarrow{} W , \; \; T(\textbf{b}_{1}) = 3\textbf{c}_{1}-2\textbf{c}_{2}+5\textbf{c}_{3}, \;\; T(\textbf{b}_{2}) = 4\textbf{c}_{1}+7\textbf{c}_{2}-1\textbf{c}_{3}. $
##### $ \hspace{0.6cm} \begin{bmatrix} T(\textbf{b}_{1}) \end{bmatrix}_{\mathcal{C}} = \begin{bmatrix} 3 \\ -2 \\ 5 \end{bmatrix}, \;\; \begin{bmatrix} T(\textbf{b}_{2}) \end{bmatrix}_{\mathcal{C}} = \begin{bmatrix} 4 \\ 7 \\ -1 \end{bmatrix}$
##### $ \hspace{0.45cm} \therefore{} M = \begin{bmatrix} \begin{bmatrix} T(\textbf{b}_{1})_{\mathcal{C}} \end{bmatrix}&\begin{bmatrix} T(\textbf{b}_{2})_{\mathcal{C}} \end{bmatrix} \end{bmatrix} = \begin{bmatrix} 3&4 \\ -2&7 \\ 5&-1 \end{bmatrix} $
##### $ \hspace{0.15cm} $ ② **[LATEX]**

##### **(`PLUS`) 같은 벡터 공간으로 선형 변환할 경우의 벡터 표현** : 
##### $ \hspace{0.15cm} M = \begin{bmatrix} \begin{bmatrix} T(\textbf{b}_{1}) \end{bmatrix}_{\mathcal{C}} & \begin{bmatrix} T(\textbf{b}_{2}) \end{bmatrix}_{\mathcal{C}} & \cdots{} & \begin{bmatrix} T(\textbf{b}_{n}) \end{bmatrix}_{\mathcal{C}} \end{bmatrix} $
##### $ \hspace{0.625cm} = \begin{bmatrix} \begin{bmatrix} T(\textbf{b}_{1}) \end{bmatrix}_{\mathcal{B}} & \begin{bmatrix} T(\textbf{b}_{2}) \end{bmatrix}_{\mathcal{B}} & \cdots{} & \begin{bmatrix} T(\textbf{b}_{n}) \end{bmatrix}_{\mathcal{B}} \end{bmatrix} \;\; \because{} \mathcal{B}=\mathcal{C}$
##### $ \hspace{0.625cm} = \begin{bmatrix} T \end{bmatrix}_{\mathcal{B}} $
##### $ \hspace{0.15cm} \Rrightarrow{} \begin{bmatrix} T(\textbf{x}) \end{bmatrix}_{\mathcal{B}} = M \begin{bmatrix} \textbf{x} \end{bmatrix}_{\mathcal{B}} = \begin{bmatrix} T \end{bmatrix}_{\mathcal{B}} \begin{bmatrix} \textbf{x} \end{bmatrix}_{\mathcal{B}} $

> ## **(Theorem 4-7) Diagonal Matrix Representaition**

#### 어떠한 정방 행렬 $ A $가 대각화가능할 때 $ A $를 이용하여 선형 변환할 경우 행렬 표현 $ M $은 대각행렬이 된다.
#### $ \Rrightarrow{} M = \begin{bmatrix} T \end{bmatrix}_{\mathcal{B}} = D $
#### $ \hspace{0.45cm} \text{where } \, T(\textbf{x}) = A\textbf{x} : \mathbb{R}^{n}\ \rightarrow{} \mathbb{R}^{n} $
#### $ \hspace{0.45cm} \text{and } \; A = PDP^{-1} \;\; (\text{diagonalizable}) $
#### $ \hspace{0.45cm} \text{and } \; P = \begin{bmatrix} \textbf{b}_{1}, \textbf{b}_{2}, \cdots{}, \textbf{b}_{n} \end{bmatrix}, \;\; \mathcal{B} = \{ \textbf{b}_{1}, \textbf{b}_{2}, \cdots{}, \textbf{b}_{n} \} \, \text{ is basis for } \, \mathbb{R}^{n} $

##### **(`WHY?`)**
##### $ \hspace{0.15cm} \text{if } \,\textbf{x} = r_{1} \textbf{b}_{1} + r_{2} \textbf{b}_{2} + \cdots{} + r_{n} \textbf{b}_{n} = P \begin{bmatrix} r_{1} \\ r_{2} \\ \vdots{} \\ r_{n} \end{bmatrix} = P\begin{bmatrix}\textbf{x}\end{bmatrix}_{\mathcal{B}} $
##### $ \hspace{0.3cm} \begin{bmatrix}\textbf{x}\end{bmatrix}_{\mathcal{B}} = P^{-1}\textbf{x} $
##### $ \hspace{0.3cm} \Rightarrow{} \begin{bmatrix} T \end{bmatrix}_{\mathcal{B}} = \begin{bmatrix} \begin{bmatrix} T(\textbf{b}_{1}) \end{bmatrix}_{\mathcal{B}} & \begin{bmatrix} T(\textbf{b}_{2}) \end{bmatrix}_{\mathcal{B}} & \cdots{} & \begin{bmatrix} T(\textbf{b}_{n}) \end{bmatrix}_{\mathcal{B}} \end{bmatrix} $ 
##### $ \hspace{1.6cm} = \begin{bmatrix} \begin{bmatrix} A\textbf{b}_{1} \end{bmatrix}_{\mathcal{B}} & \begin{bmatrix} A\textbf{b}_{2} \end{bmatrix}_{\mathcal{B}} & \cdots{} & \begin{bmatrix} A\textbf{b}_{n} \end{bmatrix}_{\mathcal{B}} \end{bmatrix} $
##### $ \hspace{1.6cm} = \begin{bmatrix} P^{-1}A\textbf{b}_{1} & P^{-1}A\textbf{b}_{2} & \cdots{} & P^{-1}A\textbf{b}_{n} \end{bmatrix} \;\; \because{} \begin{bmatrix}\textbf{x}\end{bmatrix}_{\mathcal{B}} = P^{-1}\textbf{x} $
##### $ \hspace{1.6cm} = P^{-1}A \begin{bmatrix} \textbf{b}_{1} & \cdots{} & \textbf{b}_{n} \end{bmatrix} $
##### $ \hspace{1.6cm} = P^{-1}AP = D $ 
##### $ \hspace{0.15cm} \therefore{} \begin{bmatrix} T(\textbf{x}) \end{bmatrix}_{\mathcal{B}} = \begin{bmatrix} T \end{bmatrix}_{\mathcal{B}} \begin{bmatrix}\textbf{x}\end{bmatrix}_{\mathcal{B}} = D \begin{bmatrix}\textbf{x}\end{bmatrix}_{\mathcal{B}} \;\; \because{} T(\textbf{x}) : \mathbb{R}^{n} \rightarrow{} \mathbb{R}^{n} $

##### **(`PLUS`)** 행렬 $ D $는 $ T(\textbf{x}) $의 표준 행렬 $ A $가 기저 $ \mathcal{B} $ 에서 어떻게 작용하는지를 나타내는 행렬을 의미함

##### **(`PLUS`) 선형 변환과 좌표 벡터 간 관계** :
#### $ \begin{CD} \textbf{x} @>A>> T(\textbf{x}) \\ @VP^{-1}VV @AAPA \\ \textbf{u} @>D>> D\textbf{u} \end{CD} $
