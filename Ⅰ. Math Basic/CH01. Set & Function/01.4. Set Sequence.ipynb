# CH01.4. **Set Sequence**

> ## **_(collection)**

#### **(1) 정의** : 집합을 원소로 갖는 집합
#### $ \Rrightarrow{} \mathcal{C} = \{ C_{1}, C_{2} \} \;\; \text{ where } \, C_{i} \, \text{ is set} $

##### **(`PLUS`) 멱집합(Power set)** : 어떠한 집합의 모든 가능한 부분집합을 원소로 가지는 집합
##### $ \hspace{0.15cm} \Rrightarrow{} 2^{B} = \{ A \, : \, A \subseteq{} B \} $

#### **(2) 예시** :
##### $ \hspace{0.15cm} $ ① $ B = \{ 1, 2, 3 \}, \;\; \mathcal{C} = \{ \{ 1 \}, \{ 1, 2 \}, \{ 2, 3 \} \} $
##### $ \hspace{0.15cm} $ ② $ B = \{ 1, 2 \}, \;\; \mathcal{C} = 2^{B} = \{ \emptyset{}, \{ 1 \}, \{ 2 \}, \{ 1, 2 \} \} $
##### $ \hspace{0.15cm} $ ③ $ B = \mathbb{R}, \;\; \mathcal{C} = \{ (a, b) \mid{} a < b, \; a,b \in{} \mathbb{R} \} $

> ## **집합열(Set Sequence)**

#### **(1) 정의** : 어떤 집합 $ U $가 있을 때 인덱스(라벨링) 집합 $ \, I $에서 $ U $의 멱집합(power set)으로 가는 함수 $ A $ 
#### $ \Rrightarrow{} A : \mathbb{N} \rightarrow{} 2^{U} \; \Leftrightarrow{} \; A(n) = A_{n} \in{} 2^{U} $

##### **(`PLUS`) 관례적 표기** : 어떤 집합 $ U $ 위의 부분집합들을 인덱스로 나열한 집합족
##### $ \hspace{0.15cm} \Rrightarrow{} (A_{k})_{k=1}^{n} \equiv{} \{ A_{k} \}_{k=1}^{n} := (A_{1}, A_{2}, \cdots{}, A_{n}) \;\; \text{ where } \, A_{k} \subseteq{} U $

#### **(2) 단조 집합열(Monotone Set Sequence)** : 
##### $ \hspace{0.15cm} $ ① 단조 증가 집합열 : 집합의 인덱스가 커질수록 점점 커지는(포함관계로 증가하는) 집합열
##### $ \hspace{0.3cm} \Rrightarrow{} \{A_{k}\}_{k=1}^{n} \uparrow{} := A_{1}\subseteq{}A_{2}\subseteq{}\cdots{}\subseteq{}A_{n} $
##### $ \hspace{0.15cm} $ ② 단조 감소 집합열 : 집합이 인덱스가 커질수록 점점 작아지는(포함관계로 감소하는) 집합열
##### $ \hspace{0.3cm} \Rrightarrow{} \{A_{k}\}_{k=1}^{n} \downarrow{} := A_{1}\supseteq{}A_{2}\supseteq{}\cdots{}\supseteq{}A_{n} $

<b></b>

> ## **집합열의 __(_)**

#### **(1) 집합열의 상극한(Limit Superior)** : 원소가 $ A_{n} $들에 무한히 자주 포함되는 원소들의 집합 (infinitely often; i.o.)
#### $ \hspace{0.15cm} \Rrightarrow{} \displaystyle{} \limsup_{n\to{}\infty{}} A_{n} := \bigcap_{n=1}^{\infty{}} \bigcup_{k\geq{}n} A_{k} $

#### **(2) 집합열의 하극한(Limit Inferior)** : 어떤 시점 이후로 원소가 $A_{n}$들에 항상(계속) 포함되는 원소들의 집합 (eventually always)
#### $ \hspace{0.15cm} \Rrightarrow{} \displaystyle{} \liminf_{n\to{}\infty{}} A_{n} := \bigcup_{n=1}^{\infty{}} \bigcap_{k\geq{}n} A_{k} $

<b></b>

> ## (Theorem 1-n) **_**

#### **[~]**
#### $ \Rrightarrow{} \displaystyle{} \liminf_{n\to{}\infty{}} A_{n} \subseteq{} \limsup_{n\to{}\infty{}} A_{n} $

<b></b>

> ## (Theorem 1-4) **Countable-Operations Relationship**

#### **(1)** 어떠한 집합 $ A_{1}, A_{2}, \cdots{} $ 가 가산 집합이면, 이들의 합집합도 가산 집합이다.
#### $ \Rrightarrow{} |A| \leq{} \aleph{}_{0} \;\; \text{ s.t. } \, |A_{i}| \leq{} \aleph{}_{0}, \;\; A = \cup{}^{\infty{}}_{i=1} A_{i} $

#### **(2)** 어떠한 두 집합이 가산 집합이면, 두 집합의 곱집합도 가산 집합이다.
#### $ \Rrightarrow{} |A\times{}B| \leq{} \aleph{}_{0} \;\; \text{ s.t. } \, |A| \leq{} \aleph{}_{0}, \;\; |B| \leq{} \aleph{}_{0} $
