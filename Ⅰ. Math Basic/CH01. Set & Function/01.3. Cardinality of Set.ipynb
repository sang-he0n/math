> ## **유한 및 무한 집합(Finite and Infinite Set)**

#### **(1) 유한 집합(Finite Set)** : 집합의 크기가 자연수 혹은 $ 0 $인 집합
#### $ \Rrightarrow{} |A| = n \;\; \text{ where } \, n \in{} \mathbb{N}_{0} $

#### **(2) 무한 집합(Infinite Set)** : 유한 집합이 아닌 집합
#### $ \Rrightarrow{} |A| \not\in{} \mathbb{N} $

<b></b>

> ## **가산 및 불가산 집합(Countable and Uncountable set)**

#### **(1) 가산 집합(Countable set)** : 유한 집합이거나, 그 집합에서 자연수 집합 $ \mathbb{N} $으로 단사(one-to-one;injective) 함수가 존재하는 집합
#### $ = $ 집합의 원소들을 중복 없이 자연수와 짝을 지을 수 있는(인덱싱할 수 있는) 집합
#### $ \Rrightarrow{} |A| \leq{} \aleph{}_{0} \; \Leftrightarrow{} \; A \, \text{ is countable set}. $
#### $ \hspace{0.45cm} \text{where } \, \aleph{}_{0} \, \text{ is aleph-zero}. $

#### **(2) 불가산 집합(Uncountable set)** : 가산 집합이 아닌 집합
#### $ = $ 집합의 원소들을 중복 없이 자연수와 짝을 지을 수 없는 집합
#### $ \Rrightarrow{} |A| > \aleph{}_{0} \; \Leftrightarrow{} \; A \, \text{ is uncountable set}. $

> ## (Theorem 1-3) **Countable Set-Subset Relationship**

#### **(1)** 가산 집합의 부분집합은 가산 집합이다.
#### $ \Rrightarrow{} \text{if } \, |B| \leq{} \aleph{}_{0} \, \text{ and } \, A \subseteq{} B, \;\; \text{then } \, |A| \leq{} \aleph{}_{0} \;\; \ $

#### **(2)** 실수로 구성된 구간(interval)은 모두 불가산 집합이다.
#### $ \Rrightarrow{} \text{if } \, A \subseteq{} \mathbb{R} \, \text{ and } \, A \, \text{ is non-degenerate interval}, \;\; \text{then } \, |A| > \aleph{}_{0} \;\;  $

#### **(3)** 불가산 집합의 초집합(superset)은 불가산 집합이다.
#### $ \Rrightarrow{} \text{if } \, A \subset{} B \, \text{ and } \, |A| > \aleph{}_{0}, \;\; \text{then } \, |B| > \aleph{}_{0} \;\;  $

<b></b>

> ## (Theorem 1-5) **Inclusion-Exclusion Principle**

#### 어떠한 전체 집합의 유한한 부분집합에서 여러 개의 합집합의 크기를 구할 때 아래와 같다.
#### $ \Rrightarrow{} \displaystyle{} | \cup{}_{i=1}^{n} A_{i} | = \sum_{i=1}^{n}|A_{i}| - \sum_{i<j}^{n}|A_{i}\cap{}A_{j}| + \sum_{i<j<k}^{n}|A_{i}\cap{}A_{j}\cap{}A_{k}| - \cdots{} + (-1)^{n+1}|\cap{}_{i=1}^{n}A_{i}| $
#### $ \hspace{0.45cm} \text{s.t. } \, |A_{i}| = n_{i} \;\; \text{ where } \, n_{i} \in{} \mathbb{N} $
