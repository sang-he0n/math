> ## **인덱싱 함수(Indexing Function)**

#### **(1) 정의** : 어떠한 대상 객체(집합, 함수 등)를 구분하기 위해 인덱스 집합 $ I $의 원소로 구분(라벨링)하는 함수
#### $ \Rrightarrow{} x : I \rightarrow{} X, \;\; i \mapsto{} x_{i} $

##### $ \hspace{0.15cm} $ ② **[~]**

> ## **인덱스된 집합족(Indexed Family of Sets)**

#### **(1) 정의** : 집합 $ U $의 부분집합들을 인덱스 집합 $ I $로 대응시킨 모음(함수)
#### $ \Rrightarrow{} (A_{i})_{i\in{}I} \;\; \text{ where } \, A : I \rightarrow{} 2^{U}, \;\; i \mapsto{} A_{i} $ 

##### **(`PLUS`)** 인덱스된 집합족은 함수기에 서로 다른 인덱스 $ i \neq{} j $에서 $ A_{i} = A_{j} $가 가능함 (부분집합의 중복 허용)

<b></b>

> ## **집합열(Set Sequence)**

#### **(1) 정의** : 인덱스 집합이 자연수 집합 $ \mathbb{N} $인 집합족의 특수형
#### $ \Rrightarrow{} (A_{n})_{n\in{}\mathbb{N}} \;\; \text{ where } \, A : \mathbb{N} \rightarrow{} 2^{U}, \;\; n \mapsto{} A_{n} $

<b></b>

> ## **단조 집합열(Monotone Set Sequence)**

#### **(1) 정의** : 
##### $ \hspace{0.15cm} $ ① 단조 증가 집합열 : 모든 $ n $에 대하여 $ A_{n} \subseteq{} A_{n+1} $인 집합열
##### $ \hspace{0.3cm} \Rrightarrow{} (A_{n})_{n\in{}{}\mathbb{N}} \uparrow{} \; \Leftrightarrow{} \; A_{n} \subseteq{} A_{n+1} \; (\forall n \in{} \mathbb{N}) $

##### $ \hspace{0.15cm} $ ② 단조 감소 집합열 : 모든 $ n $에 대하여 $ A_{n} \supseteq{} A_{n+1} $인 집합열
##### $ \hspace{0.3cm} \Rrightarrow{} (A_{n})_{n\in{}{}\mathbb{N}} \downarrow{} \; \Leftrightarrow{} \; A_{n} \supseteq{} A_{n+1} \; (\forall n \in{} \mathbb{N}) $

<b></b>

> ## **집합열의 극한(Limit of Set Sequence)**

#### **(1) 상극한(Limit Superior)** : 원소가 $ A_{n} $들에 무한히 자주 포함되는 원소들의 집합
#### $ \Rrightarrow{} \displaystyle{} \limsup_{n\to{}\infty{}} A_{n} := \bigcap_{n=1}^{\infty{}} \bigcup_{k\geq{}n} A_{k} $

#### **(2) 하극한(Limit Inferior)** : 어떤 시점 이후로 원소가 $A_{n}$들에 항상(계속) 포함되는 원소들의 집합
#### $ \Rrightarrow{} \displaystyle{} \liminf_{n\to{}\infty{}} A_{n} := \bigcup_{n=1}^{\infty{}} \bigcap_{k\geq{}n} A_{k} $

<b></b>

> ## (Theorem 1-6) **집합열의 극한 성질(Property of Limit Of Sets)**

#### **(1)**
#### $ \Rrightarrow{} \displaystyle{} \liminf_{n\to{}\infty{}} A_{n} \subseteq{} \limsup_{n\to{}\infty{}} A_{n} $

#### **(2)** 
#### $ \Rrightarrow{} \displaystyle{} ( \limsup_{n\to{}\infty{}} A_{n} )^{c} = \liminf_{n\to{}\infty{}} A_{n}^{c} \hspace{0.45cm} \displaystyle{} ( \liminf_{n\to{}\infty{}} A_{n} )^{c} = \limsup_{n\to{}\infty{}} A_{n}^{c} $

#### **(3)** 집합열의 상극한과 하극한이 동일할 경우, 극한이 존재한다.
#### $ \Rrightarrow{} \displaystyle{} \lim_{n\rightarrow{}\infty{}} A_{n} \, \text{ exist} \; \Leftrightarrow{} \; \limsup_{n\to{}\infty{}} A_{n} = \liminf_{n\to{}\infty{}} A_{n} $

#### **(4)** 
#### $ \Rrightarrow{} \displaystyle{} (A_{n})_{n\in{}\mathbb{N}} \uparrow{} \; \Rightarrow{} \; \lim_{n\rightarrow{}\infty{}}A_{n} = \bigcup_{n=1}^{\infty{}} A_{n}, \;\; (A_{n})_{n\in{}\mathbb{N}} \downarrow{} \; \Rightarrow{} \; \lim_{n\rightarrow{}\infty{}}A_{n} = \bigcap_{n=1}^{\infty{}} A_{n} $

<b></b>

> ## **집합의 분할(Partition of a set)**

#### **(1) 정의** : 어떠한 집합족 $ (A_{i})_{i\in{}I} $에서 각 부분집합이 공집합이 아니며 서로 다른 두 부분집합이 서로소 관계인 집합족의 특수형
#### $ \Rrightarrow{} \displaystyle{} (A_{i})_{i\in{}I} \, \text{ is partition of } \, A \; \Leftrightarrow{} \; A = \bigcup_{i\in{}I}A_{i}, \;\; A_{i} \subseteq{} A, \;\; A_{i} \neq{} \emptyset{} , \; \; A_{i} \cap{} A_{j} = \emptyset{} \;\; (i \neq{} j) $

#### **(2) 예시** :
##### $ \hspace{0.15cm} $ ① 어떠한 분할된 집합 $ A $를 합집합으로 표현하면
##### $ \hspace{0.45cm} A = \{ 1, 2, 3, 4, 5 \}, \;\; A_{1} = \{ 1 \}, \;\; A_{2} = \{ 2, 4 \}, \;\; A_{3} = \{ 3, 5 \} $ 
##### $ \hspace{0.45cm} \therefore{} A_{1} \cup{} A_{2} \cup{} A_{3} = \displaystyle{} \bigcup^{3}_{i=1}A_{i} = A = \{ 1, 2, 3, 4, 5 \} $
##### $ \hspace{0.15cm} $ ② 어떠한 집합 $ B $를 분할된 부분집합으로 표현하면
##### $ \hspace{0.45cm} B = \{ a, b, c, d, e \}, \;\; \mathcal{B} = \{ B_{1}, B_{2}, B_{3} \} \, \text{ is partition of } \, B $
##### $ \hspace{0.45cm} B_{1} = \{ a \}, \;\; B_{2} = \{ b, d \}, \;\; B_{3} = \{ c, e \} \;\; \because{} \, B_{i} \neq{} \emptyset{}, \;\; B_{i} \cap{} B_{j} = \emptyset{} \; (i \neq{} j), \;\; \displaystyle{} \bigcup_{i=1}^{3}B_{i} = B $
