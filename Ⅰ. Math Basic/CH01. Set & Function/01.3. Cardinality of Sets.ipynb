# CH01.3. **Cardinality of Sets**

> ## **동수(Equinumerous;동치;Equivalence)**

#### **(1) 정의** : 두 집합 사이에 전단사(bijection) 함수가 존재할 때 갖는 두 집합의 관계
#### $ \Rrightarrow{} A \sim{} B \; \Leftrightarrow{} \; \exists{} \, f:A\rightarrow{}B \;\; \text{ s.t. } \, f \, \text{ is bijective} $

<b></b>

> ## **유한 및 무한 집합(Finite and Infinite Set)**

#### **(1) 유한 집합(Finite Set)** : 어떤 자연수 초기구간 $ [n] $과 동수인 집합
#### $ \Rrightarrow{} A \, \text{ is finite } \; \Leftrightarrow{} \; \exists{} \, n \in{} \mathbb{N}_{0} := \{0, 1, 2, \cdots{}\} \; \text{ s.t. } \, A \sim{} [n] $

##### **(`PLUS`) 자연수 초기 구간** : 처음 $ n $개의 자연수로 이루어진 집합
##### $ \hspace{0.15cm} \Rrightarrow{} [n] := \{1, 2, \cdots{}, n\}, \;\; [0] := \emptyset{} $

#### **(2) 무한 집합(Infinite Set)** : 어떠한 자연수 초기 구간 $ [n] $과 동수가 아닌 집합
#### $ \Rrightarrow{} A \, \text{ is infinite } \; \Leftrightarrow{} \; \forall{} \, n \in{} \mathbb{N}_{0}, \;\; A \not\sim{} [n] $

<b></b>

> ## **집합의 크기(Cardinality)**

#### **(1) 정의** : 동수인 두 집합이 갖는 공통된 크기
#### $ \Rrightarrow{} A \sim{} B \; \Leftrightarrow{} \; |A| = |B| $

##### **(`PLUS`)** 유한 집합의 크기는 그 집합의 원소의 개수로 정의됨
##### $ \hspace{0.15cm} \Rrightarrow{} A \sim{} [n] \; \Rightarrow{} \; |A| := n $

#### **(2) 성질** : 
##### $ \hspace{0.15cm} $ ① 
##### $ \hspace{0.15cm} $ ② $ A \subseteq{} B \; \Rightarrow{} \; |A| \leq{} |B| \; \text{ where } A, B \, \text{ is finite} $

<b></b>

> ## (Theorem 1-3) **Inclusion-Exclusion Principle**

#### $ n $개의 유한 집합 $ A_{1}, A_{2}, \cdots{}, A_{n} $에 대해 그 합집합의 크기는 아래와 같다.
#### $ \Rrightarrow{} \displaystyle{} | \bigcup_{i=1}^{n} A_{i} | = \sum_{i=1}^{n}|A_{i}| - \sum_{i<j}^{n}|A_{i}\cap{}A_{j}| + \sum_{i<j<k}^{n}|A_{i}\cap{}A_{j}\cap{}A_{k}| - \cdots{} + (-1)^{n+1}|\bigcap_{i=1}^{n}A_{i}| $
#### $ \hspace{0.45cm} \text{s.t. } \, |A_{i}| = n_{i} \in{} \mathbb{N}_{0} $

<b></b>

> ## **가산 및 불가산 집합(Countable and Uncountable Set)**

#### **(1) 가산 집합(Countable set)** : 유한 집합이거나, 그 집합에서 자연수 집합 $ \mathbb{N} $으로 단사 함수가 존재하는 집합
#### $ = $ 집합의 원소들을 중복 없이 자연수와 중복 없이 대응시킬 수 있는 집합
#### $ \Rrightarrow{} A \, \text{ is countable}. \; \Leftrightarrow{} \; A \, \text{ is finite} \, \text{ or } \, \exists{} \, f : A \rightarrow{} \mathbb{N} \;\; (\text{injective}) $

#### **(2) 가산 무한 집합(Countably Infinite Set)** : 그 집합에서 자연수 집합 $ \mathbb{N} $으로 전단사 함수가 존재하는 집합
#### $ \Rrightarrow{} A \, \text{ is countably infinite}. \; \Leftrightarrow{} \; \exists{} \, f : A \rightarrow{} \mathbb{N} \;\; (\text{bijective}) \; \Leftrightarrow{} \; A \sim{} \mathbb{N} $

#### **(3) 불가산 집합(Uncountable set)** : 가산 집합이 아닌 집합
#### $ = $ 집합의 원소들을 자연수와 중복 없이 대응시킬 수 없는 집합
#### $ \Rrightarrow{} A \, \text{ is uncountable}. \; \Leftrightarrow{} \; \nexists{} \, f : A \rightarrow{} \mathbb{N} \;\; (\text{injective}) $

> ## (Theorem 1-4) **Countable Set-Subset Relationship**

#### **(1)** 가산 집합의 부분집합은 가산 집합이다.
#### $ \Rrightarrow{} B \, \text{ is countable}, \;\; A \subseteq{} B \; \Rightarrow{} \; A \, \text{ is countable} $

#### **(2)** 실수로 구성된 비퇴화 구간(interval)은 모두 불가산 집합이다.
#### $ \Rrightarrow{} A \subseteq{} \mathbb{R}, \;\; A \, \text{ is (non-degenerate) interval} \; \Rightarrow{} \;  A \, \text{ is uncountable} $

#### **(3)** 불가산 집합의 초집합(superset)은 불가산 집합이다.
#### $ \Rrightarrow{} A \subseteq{} B, \;\;  A \, \text{ is uncountable}, \; \Rightarrow{} \;  B \, \text{ is uncountable} $

> ## (Theorem 1-5) **Countable-Operations Relationship**

#### **(1)** 어떠한 두 집합이 가산 집합이면, 두 집합의 곱집합도 가산 집합이다.
#### $ \Rrightarrow{} A, B \, \text{ are countable} \; \Rightarrow{} \; A\times{}B \, \text{ is countable} $

#### **(2)** 어떠한 집합 $ A_{1}, A_{2}, \cdots{} $들이 가산 집합이면, 이들의 합집합도 가산 집합이다.
#### $ \Rrightarrow{} \displaystyle{} \forall{} \, i \in{} \mathbb{N}, \;\; A_{i} \, \text{ is countable} \; \Rightarrow{} \; \bigcup^{\infty{}}_{i=1} A_{i} \, \text{ is countable}  $
