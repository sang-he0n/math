# CH01.2. **Function**

> ## **함수(Function)**

#### **(1) 정의** : 특정 집합에서 다른 집합으로의 대응 관계를 나타내는 규칙
#### $ \Rrightarrow{} f : A \rightarrow{} B, \;\; a \mapsto{} f(a) \;\; \text{ where } \, A, B \, \text{ are sets}, \;\; a \in{} A $

#### **(2) 형식** : 
##### $ \hspace{0.15cm} $ <img src="../img/01.2. Function (1).png" width="40%" height="40%"></img>
##### $ \hspace{0.15cm} $ ① 정의역(Domain) : 함수에 입력할 수 있는 모든 원소(input)들의 집합 $ A $
##### $ \hspace{0.15cm} $ ② 공역(Codomain) : 함수값이 속할 수 있는 가능한 출력값들의 집합 $ B $ 
##### $ \hspace{0.15cm} $ ③ 상(Image) : 정의역 내 입력값이 함수를 통해 대응되는 출력값들의 집합 $ f(A) := \{ f(a) \mid{} a \in{} A \} \subseteq{} B $

#### **(3) 함수의 분류** :
##### $ \hspace{0.15cm} $ ① 전사(Onto;Surjective) : 상(image)와 공역(codomain)이 동일한 관계
##### $ \hspace{0.45cm} \Rrightarrow{} f(A) = B $
##### $ \hspace{0.15cm} $ ② 단사(One-to-one;Injective) : 다른 입력(원소)은 항상 다른 출력을 갖는 관계
##### $ \hspace{0.45cm} \Rrightarrow{} f(a_{1}) \neq{} f(a_{2}) \;\; \text{ where } \, \forall{} \, a_{1} \neq{} a_{2}, \;\; a_{1}, a_{2} \in{} A $
##### $ \hspace{0.15cm} $ ③ 전단사(Bijective) : 전사이면서 단사인 관계
##### $ \hspace{0.45cm} \Rrightarrow{} f(A) = B, \; \; f(a_{1}) \neq{} f(a_{2}) $

> ## **역상(Preimage;Inverse image)**

#### **(1) 정의** : 함수 $ f $와 공역(codomain)의 부분집합 $ C \subseteq{} B $ 가 주어졌을 때 $ \, C $로 가는 입력(원소)들의 집합
#### $ \Rrightarrow{} f^{-1}(C) := \{ a \in A \mid f(a) \in C \} \;\; \text{ where } \, f : A \rightarrow{} B $
##### $ \hspace{0.15cm} $ <img src="../img/01.2. Function (5).png" width="40%" height="40%"></img>

#### **(2) 성질** : 
##### $ \hspace{0.15cm} $ ① $ f^{-1}(\empty{}) = \empty{}, \;\; f^{-1}(B)=A \;\; \text{ where } \, f : A \rightarrow{} B $
##### $ \hspace{0.15cm} $ ② $ f^{-1}(C^{c}) = \big({}f^{-1}(C)\big){}^{c} = A - f^{-1}(C) $
##### $ \hspace{0.15cm} $ ③ $ C \subseteq{} D \; \Rightarrow{} \; f^{-1}(C) \subseteq{} f^{-1}(D) $
##### $ \hspace{0.15cm} $ ④ $ \displaystyle{} f^{-1}\big({}\bigcup_{i}C_{i}\big){} = \bigcup_{i} f^{-1}(C_{i}), \;\; f^{-1}\big({}\bigcap_{i}C_{i}\big){} = \bigcap_{i} f^{-1}(C_{i}) $ 
##### $ \hspace{0.15cm} $ ⑤ $ \displaystyle{} f^{-1}(C-D) = f^{-1}(C)-f^{-1}(D) $ 

#### **(3) 예시** :
##### $ \hspace{0.15cm} $ ① 함수 $f : \{1,2,3\} \rightarrow{} \{a,b,c,d\} $가 $ f(1)=a, \;\; f(2)=c, \;\; f(3)=c $일 때, $ \, f^{-1}(\{c\}) $와 $ f^{-1}(\{a,c\}) $를 구하면
##### $ \hspace{0.45cm} \therefore{} \begin{cases} f^{-1}(\{c\})=\{x\in{} \{1,2,3\} \mid{} f(x)\in{}\{c\}\}=\{x\in{} \{1,2,3\} \mid{} f(x)=c\}=\{2,3\} \\ f^{-1}(\{a,c\})=\{x\in{} \{1,2,3\} \mid{} f(x)\in{}\{a,c\}\}=\{x\in{} \{1,2,3\} \mid{} f(x)=a \, \text{ or } \, f(x) = c\}=\{1,2,3\} \end{cases} $
##### $ \hspace{0.15cm} $ ② $ f : \mathbb{Z} \rightarrow{} \{ 0,1,2 \}, \;\; f(n) = n \bmod{} 3 $일 때, $ \, f^{-1}(\{ 1 \}) $를 구하면
##### $ \hspace{0.45cm} f^{-1}( \{1\}) = \{ n \in{} \mathbb{Z}\mid{} f(n)\in{}\{1\}\} = \{ n \in{} \mathbb{Z} \mid{} n\bmod{} 3 = 1\} $
##### $ \hspace{0.45cm} \therefore{} f^{-1}(\{1\})=\{\, 3k+1 \mid{} k\in{}\mathbb{Z}\,\} = \{\cdots{},-5,-2,1,4,7,10,\cdots{} \} $
##### $ \hspace{0.15cm} $ ③ $ f : \mathbb{R} \rightarrow{} \mathbb{R}, \;\; f(x) = x^{2} $일 때, $ \, f^{-1}([1,4]) $를 구하면
##### $ \hspace{0.45cm} f^{-1}([1,4]) = \{ x \in{} \mathbb{R} \mid{} f(x)\in{}[1,4]\} = \{ x \in{} \mathbb{R} \mid{} x^{2}\in{}[1,4]\} $
##### $ \hspace{0.45cm} 1 \leq{} x^{2} \leq{} 4 \;\; \Leftrightarrow{} \;\; 1 \leq{} |x| \leq{} 2 \;\; \Leftrightarrow{} \;\; x \in{} [-2,-1]\cup{}[1,2] $
##### $ \hspace{0.45cm} \therefore{} f^{-1}([1,4]) = [-2,-1]\cup{}[1,2] $

> ## **합성 함수(Composite Function)**

#### **(1) 정의** : 두 함수 $ f : A \rightarrow{} B, \;\; g : B \rightarrow{} C $를 순서대로 적용한 함수
#### $ \Rrightarrow{} (g\circ{} f):A\rightarrow{} C, \;\; (g\circ{} f)(a) = g \big({} f(a) \big){} \;\; \text{ where } \, a \in{} A $ 

#### **(2) 성질** : 
##### $ \hspace{0.15cm} $ ① $ (h \circ{} g) \circ{} f = h \circ{} (g \circ{} f) $ 
##### $ \hspace{0.15cm} $ ② $ (g \circ{} f)^{-1}(D) = f^{-1}\big({} g^{-1}(D) \big){} \;\; \text{ where } \, D \subseteq{} C \;\; (\text{preimage}) $

> ## **역함수(Inverse Function)**

#### **(1) 정의** : 어떠한 함수 $ f : A \rightarrow{} B $가 전단사(bijective) 함수일 때, $ B $의 각 원소에 유일하게 대응되는 $ A $의 원소로 역매핑하는 함수
#### $ \Rrightarrow{} f^{-1} : B \rightarrow{} A \;\; \text{ s.t. } \, f^{-1}\big({} f(a)\big){} = a, \;\; f\big({}f^{-1}(b)\big){} = b \;\;\;\; \forall{} \, a \in{} A,  \;\; \forall{} b \in{} B $ 

##### **(`PLUS`)** 역상은 집합 입력을 받는 항상 정의되는 연산이고 역함수는 원소 입력을 받고 전단사 함수일 때만 정의됨
##### $ \hspace{0.15cm} \Rrightarrow{} \begin{cases} f^{-1}(C) & \cdots{} \, (\text{preimage}) \hspace{1.225cm} \text{where } \, C \, \text{ is set} \\ f^{-1}(b) & \cdots{} \, (\text{inverse function}) \;\; \text{where } \, f \, \text{ is bijective function}, \;\; b \in{} B \end{cases} $

#### **(2) 성질** : 
##### $ \hspace{0.15cm} $ ① $ (f^{-1})^{-1} = f $
##### $ \hspace{0.15cm} $ ② $ (g\circ{} f)^{-1} = f^{-1}\circ{} g^{-1} \;\; \text{ s.t. } \, f, g \, \text{ are bijective functions}. $

#### **(3) 예시** :
##### $ \hspace{0.15cm} $ ① $ f : \mathbb{R} \rightarrow{} \mathbb{R}, \;\; f(x) = 3x+2 $일 때, 역함수 $ f^{-1} $를 구하면
##### $ \hspace{0.45cm} \text{let } \, y := f(x) = 3x+2 \in{} \mathbb{R}, \;\; 3x = y-2 \;\; \Leftrightarrow{} \;\; x = \frac{y-2}{3} $
##### $ \hspace{0.45cm} \therefore{} f^{-1}(y) = \frac{y-2}{3} $
##### $ \hspace{0.15cm} $ ② $ f : (0,\infty) \rightarrow{} \mathbb{R}, \;\; f(x) = \ln x $일 때, 역함수 $ f^{-1} $와 $ f^{-1}(2) $를 구하면
##### $ \hspace{0.45cm} \text{let } \, y := \ln x \in{} \mathbb{R}, \;\; y=\ln x \;\; \Leftrightarrow{} \;\; x=e^{y} $
##### $ \hspace{0.45cm} \therefore{} f^{-1}(y) = e^{y}, \;\; f^{-1}(2)=e^{2} $
