# CH01.1. **Set basic**

> ## **집합(Set)**

#### **(1) 정의** : 중복이 없고, 순서를 고려하지 않는 원소(아이템, 객체)들의 모음 
#### $ \Rrightarrow{} \{ \text{form} \mid{} \text{condition} \} \; \Leftrightarrow{} \; \{ \text{form} \; : \; \text{condition} \} $

#### **(2) 원소 포함 관계** : 집합 $ A = \{ 😀, 😡 \} $가 존재할 때 아래가 성립함
##### $ \hspace{0.15cm} $ ① 원소 😀가 집합 $ A $에 속함 $ \; \Rightarrow{} 😀 \in{} A $
##### $ \hspace{0.15cm} $ ② 원소 😭가 집합 $ A $에 속하지 않음 $ \; \Rightarrow{} 😭 \notin{} A $

#### **(3) 수의 체계 표현** :
##### $ \hspace{0.15cm} $ ① 자연수(Natural number) : $ \mathbb{N} := \{ 1, 2, 3, \cdots{} \} $

##### **(`PLUS`)** 해당 노트에서는 자연수의 정의에서는 $ 0 $을 포함하지 않음

##### $ \hspace{0.15cm} $ ② 정수(Integer) : $ \mathbb{Z} := \{ \cdots{}, -1, 0, 1, 2, 3, \cdots{} \} $
##### $ \hspace{0.15cm} $ ③ 유리수(Rational number) : $ \mathbb{Q} := \{ \frac{p}{q} \mid{} p,q \in{} \mathbb{Z}, \;\; q \neq{} 0 \} \approx{} \{ \cdots{} , -{\frac{1}{3}}, \cdots{}, 0, \cdots{}, {\frac{2}{5}}, \cdots{} \}$
##### $ \hspace{0.15cm} $ ④ 실수(Real number) : $ \mathbb{R} \approx{} \{ \cdots{}, -0.001, \cdots{}, 0, \cdots{}, \pi{}, \cdots{} \} $

#### **(4) 수의 범위 표현** : 
##### $ \hspace{0.15cm} $ ① 열린 구간(Open interval) : $ (a, b) := \{ x \mid{} a < x < b, \;\; a,b \in{} \mathbb{R} \} $
##### $ \hspace{0.15cm} $ ② 닫힌 구간(Closed interval) : $ [a, b] := \{ x \mid{} a \leq{} x \leq{} b, \;\; a,b \in{} \mathbb{R} \} $
##### $ \hspace{0.15cm} $ ③ 반닫힌 구간(Half-closed interval) : $ [a, b) := \{ x \mid{} a \leq{} x < b, \;\; a,b \in{} \mathbb{R} \} $
##### $ \hspace{0.15cm} $ ④ 퇴화 구간(Degenerate interval) : $ [a, a] = \{ a \} $

#### **(5) 예시** :
##### $ \hspace{0.15cm} $ ① $ \{ 1, 2, 3, 4 \} $
##### $ \hspace{0.15cm} $ ② $ \{ 😀, 😡 \} = \{ 😡, 😀 \} $
##### $ \hspace{0.15cm} $ ③ $ \{ $가$ , $나$ , $다$ \} $

<b></b>

> ## **우주 집합과 공집합(Universal Set and Empty Set)**

#### **(1) 우주 집합(Universal Set)** : 현재 문맥을 대표하는 전체 집합
#### $ \Rrightarrow{} U $

#### **(2) 공집합(Empty Set)** : 원소를 하나도 갖지 않는 집합
#### $ \Rrightarrow{} \emptyset{} := \{ \} $

<b></b>

> ## **부분집합(Subset)**

#### **(1) 정의** : 어떠한 집합 $ A $의 모든 원소가 다른 집합 $ B $에도 존재할 때의 집합 $ A $
#### $ \Rrightarrow{} A \subseteq{} B $
#### $ \Leftrightarrow{} \; (\forall{} \, x) \;\; x \in{} A \; \Rightarrow{} \; x \in{} B $

##### **(`PLUS`)** 집합 $ B $를 집합 $ A $의 초집합(superset)이라고 정의함

#### **(2) 예시** :
##### $ \hspace{0.15cm} $ ① $ \{ a, b \} \subseteq{} \{ a, b, c \} $
##### $ \hspace{0.15cm} $ ② $ \{ a, b, c \} \subseteq{} \{ a, b, c \} $
##### $ \hspace{0.15cm} $ ③ $ \{ 1, 4 \} \not\subseteq{} \{ 1, 2, 3 \} $

<b></b>

> ## **진부분집합(Proper Subset)**

#### **(1) 정의** : 어떠한 집합 $ A $의 모든 원소가 다른 집합 $ B $에도 존재하나 집합 $ B $와는 다른 집합 $ A $
#### $ \Rrightarrow{} A \subsetneqq{} B $

<b></b>

> ## **집합 연산(Set Operation)**

#### **(1) 정의** : 둘 이상의 집합에 대해 특정한 규칙을 적용하여 새로운 집합을 만들어내는 연산

#### **(2) 종류** : 
##### $ \hspace{0.15cm} $ ① 합집합(Union) : 두 집합의 모든 원소를 포함하는 집합 
##### $ \hspace{0.45cm} \Rrightarrow{} A \cup{} B := \{ x \mid{} x \in{} A \, \text{ or } \, x \in{} B \} $
##### $ \hspace{0.45cm} $ <img src="../img/01.1. Set basic (3).png" width="30%" height="30%"></img>
##### $ \hspace{0.15cm} $ ② 교집합(Intersection) : 두 집합의 공통 원소를 포함하는 집합 
##### $ \hspace{0.45cm} \Rrightarrow{} A \cap{} B := \{ x \mid{} x \in{} A \, \text{ and } \, x \in{} B \} $
##### $ \hspace{0.45cm} $ <img src="../img/01.1. Set basic (4).png" width="30%" height="30%"></img>

##### **(`PLUS`) 서로소(Pairwise Disjoint Set)** : 두 집합이 공통된 원소를 갖지 않는 관계
##### $ \hspace{0.15cm} A \cap{} B = \emptyset{} = \{ \} $

##### $ \hspace{0.15cm} $ ③ 여집합(Complement) : 전체 집합 중 특정 집합의 원소를 제외한 집합 
##### $ \hspace{0.45cm} \Rrightarrow{} A^{c} := \{ x \mid{} x \in{} U \, \text{ and } \, x \not\in{} A \} $

##### $ \hspace{0.15cm} $ ④ 차집합(Subtraction;Difference) : 한 집합에서 다른 집합의 원소를 제외한 집합 
##### $ \hspace{0.45cm} \Rrightarrow{} A \setminus{} B := A \cap{} B^{c} = \{ x \mid{} x \in{} A \, \text{ and } \, x \not\in{} B \} \;\; \text{ where } \, B^{c} = U \setminus{} B $

> ## **곱집합(Product set;Cartesian product)**

#### **(1) 정의** : 두 개 이상의 집합에서 원소를 하나씩 뽑아 만든 모든 순서 $n$-튜플의 집합
#### $ \Rrightarrow{} A_{1} \times{} A_{2} \times{} \cdots{} \times{} A_{n} = \{ (a_{1}, a_{2}, \cdots{}, a_{n}) \mid{} a_{i} \in{} A_{i} \; (i=1,2,\cdots{},n) \} $

#### **(2) 성질** : 
##### $ \hspace{0.15cm} $ ① 곱집합은 원소의 순서를 고려함
##### $ \hspace{0.45cm} \Rrightarrow{} (1,a) \neq{} (a,1), \;\; A \times{} B \neq{} B \times{} A \;\; (\text{generally}) $

#### **(2) 예시 :**
##### $ \hspace{0.15cm} $ ① 집합 $ A = \{ 1, 2 \}, \;\; B = \{ a, b, c \} $가 존재할 때 $ A \times{} B $와 $ B \times{} A $를 구하면
##### $ \hspace{0.45cm} A \times{} B = \{ (1,a), (1,b), (1,c), (2,a), (2,b), (2,c) \} $
##### $ \hspace{0.45cm} B \times{} A = \{ (a,1), (a,2), (b,1), (b,2), (c,1), (c,2) \} $
##### $ \hspace{0.15cm} $ ② $ \mathbb{R} \times{} \mathbb{R} \times{} \mathbb{R} = \{ (x, y, z) \mid{} x, y, z \in{} \mathbb{R} \} = \mathbb{R}^{3} $ ($ \text{space} $)
