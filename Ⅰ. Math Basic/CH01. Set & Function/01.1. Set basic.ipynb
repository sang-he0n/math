# CH01.1. **Set basic**

> ## **집합(Set)**

#### **(1) 정의** : 어떤 아이템, 객체 혹은 원소들의 유일하고(중복이 없고), 순서를 고려하지 않는 모음
#### $ \Rrightarrow{} \{ \text{form} \mid{} \text{condition} \} \; \Leftrightarrow{} \; \{ \text{form} \; : \; \text{condition} \} $

#### **(2) 포함 관계** : 집합 $ A = \{ 😀, 😡 \} $가 존재할 때 아래가 성립함
##### $ \hspace{0.15cm} $ ① 원소 😀가 집합 $ A $에 속함 $ \; \Rightarrow{} 😀 \in{} A $
##### $ \hspace{0.15cm} $ ② 원소 😭가 집합 $ A $에 속하지 않음 $ \; \Rightarrow{} 😭 \notin{} A $

#### **(3) 수의 체계 표현** :
##### $ \hspace{0.15cm} $ ① 자연수(Natural number) : $ \mathbb{N} := \{ 1, 2, 3, \cdots{} \} $

##### **(`PLUS`)** 해당 노트에서는 자연수의 정의에서는 $ 0 $을 포함하지 않음

##### $ \hspace{0.15cm} $ ② 정수(Integer) : $ \mathbb{Z} := \{ \cdots{}, -1, 0, 1, 2, 3, \cdots{} \} $
##### $ \hspace{0.15cm} $ ③ 유리수(Rational number) : $ \mathbb{Q} := \{ \frac{p}{q} \mid{} p,q \in{} \mathbb{Z}, \;\; q \neq{} 0 \} \approx{} \{ \cdots{} , -{\frac{2}{6}}, \cdots{}, 0, \cdots{}, {\frac{2}{5}}, \cdots{} \}$
##### $ \hspace{0.15cm} $ ④ 실수(Real number) : $ \mathbb{R} \approx{} \{ \cdots{}, -0.001, \cdots{}, 0, \cdots{}, \pi{}, \cdots{} \} $

#### **(4) 수의 범위 표현** : 
##### $ \hspace{0.15cm} $ ① 열린 구간(Open interval) : $ (a, b) = \{ x \mid{} a < x < b, \;\; a,b \in{} \mathbb{R} \} $
##### $ \hspace{0.15cm} $ ② 닫힌 구간(Closed interval) : $ [a, b] = \{ x \mid{} a \leq{} x \leq{} b, \;\; a,b \in{} \mathbb{R} \} $
##### $ \hspace{0.15cm} $ ③ 반닫힌 구간(Half-closed interval) : $ [a, b) \; \text{ or } \; (a, b] $
##### $ \hspace{0.15cm} $ ④ 퇴화 구간(Degenerate interval) : $ [a, a] = \{ a \} $

#### **(5) 예시** :
##### $ \hspace{0.15cm} $ ① $ \{ 1, 2, 3, 4 \} $
##### $ \hspace{0.15cm} $ ② $ \{ 😀, 😡 \} = \{ 😡, 😀 \} $
##### $ \hspace{0.15cm} $ ③ $ \{ $가$ , $나$ , $다$ \} $

> ## **우주 집합과 공집합(Universe Set and Empty Set)**

#### **(1) 우주 집합(Universe Set)** : 전체를 대표하는 기준 집합
#### $ \Rrightarrow{} U $

#### **(2) 공집합(Empty Set)** : 원소를 하나도 갖지 않는 집합
#### $ \Rrightarrow{} \empty{} := \{ \} $

<b></b>

> ## **부분집합(Subset)**

#### **(1) 정의** : 어떠한 집합 $ A $의 모든 원소가 다른 집합 $ B $에도 존재할 때의 집합 $ A $
#### $ \Rrightarrow{} A \subseteq{} B $
#### $ \Leftrightarrow{} \forall{} \, x, \;\; x \in{} A \; \Rightarrow{} \; x \in{} B $

##### **(`PLUS`)** 집합 $ B $를 집합 $ A $의 초집합(superset)이라고 정의함

#### **(2) 포함 관계** : 집합 $ A = \{ 😀, 😡 \} $와 집합 $ B = \{ 😀, 😡, 🙃 \} $가 존재할 때 아래가 성립함
##### $ \hspace{0.15cm} $ ① $ A $는 $ B $의 부분집합임 $ \; \Rightarrow{} A \subseteq{} B $ 
##### $ \hspace{0.15cm} $ ② $ B $는 $ A $의 부분집합이 아님 $ \; \Rightarrow{} B \not\subseteq{} A $ 

#### **(3) 예시** :
##### $ \hspace{0.15cm} $ ① $ \{ a, b \} \subseteq{} \{ a, b, c \} $
##### $ \hspace{0.15cm} $ ② $ \{ a, b, c \} \subseteq{} \{ a, b, c \} $
##### $ \hspace{0.15cm} $ ③ $ \{ 1, 4 \} \not\subseteq{} \{ 1, 2, 3 \} $

<b></b>

> ## **진부분집합(Proper Subset)**

#### **(1) 정의** : 어떠한 집합 $ A $의 모든 원소가 다른 집합 $ B $에도 존재하나 집합 $ B $와는 다른 집합 $ A $
#### $ \Rrightarrow{} A \subsetneqq{} B $

<b></b>

> ## **집합족(Family of Sets;collection)**

#### **(1) 정의** : 어떠한 집합 $ B $의 부분집합들의 모임 $ \, \mathcal{C} $
#### $ \Rrightarrow{} \mathcal{C} \subseteq{} 2^{B} \;\; \text{ where } \, B \, \text{ is set} $

##### **(`PLUS`) 멱집합(Power set)** : 어떠한 집합의 모든 가능한 부분집합을 원소로 가지는 집합
##### $ \hspace{0.15cm} \Rrightarrow{} 2^{B} = \{ A \, : \, A \subseteq{} B \} $

#### **(2) 예시** :
##### $ \hspace{0.15cm} $ ① $ B = \{ 1, 2, 3 \}, \;\; \mathcal{C} = \{ \{ 1 \}, \{ 1, 2 \}, \{ 2, 3 \} \} $
##### $ \hspace{0.15cm} $ ② $ B = \{ 1, 2 \}, \;\; \mathcal{C} = 2^{B} = \{ \emptyset{}, \{ 1 \}, \{ 2 \}, \{ 1, 2 \} \} $
##### $ \hspace{0.15cm} $ ③ $ B = \mathbb{R}, \;\; \mathcal{C} = \{ (a, b) \mid{} a < b, \; a,b \in{} \mathbb{R} \} $

> ## **집합 연산(Set Operation)**

#### **(1) 정의** : 둘 이상의 집합에 대해 특정한 규칙을 적용하여 새로운 집합을 만들어내는 연산

#### **(2) 종류** : 
#### $ \hspace{0.15cm} $ ① 합집합(Union) : 두 집합의 모든 원소를 포함하는 집합 
#### $ \hspace{0.45cm} \Rrightarrow{} A \cup{} B = \{ 1, 2, 3 \} \cup{} \{ 1, 4, 5 \} = \{ 1, 2, 3, 4, 5 \} $ 
#### $ \hspace{0.45cm} $ <img src="../img/01.1. Set (3).png" width="30%" height="30%"></img>

##### **(`PLUS`)** $ \displaystyle{} A_{1} \cup{} A_{2} \cup{} A_{3} \cup{} A_{4} \cdots{} \cup A_{n} = \bigcup_{i=1}^{n} A_{i} $

#### $ \hspace{0.15cm} $ ② 교집합(Intersection) : 두 집합의 공통 원소를 포함하는 집합 
#### $ \hspace{0.45cm} \Rrightarrow{} A \cap{} B = \{ 1, 2, 3 \} \cap{} \{ 1, 4, 5 \} = \{ 1 \} $
#### $ \hspace{0.45cm} $ <img src="../img/01.1. Set (4).png" width="30%" height="30%"></img>

##### **(`PLUS`)** $ \displaystyle{} A_{1} \cap{} A_{2} \cap{} A_{3} \cap{} A_{4} \cdots{} \cap A_{n} = \bigcap_{i=1}^{n} A_{i} $

#### $ \hspace{0.15cm} $ ③ 여집합(Complement) : 전체 집합 중 특정 집합의 원소를 제외한 집합

> ## (Theorem 1-2) **Distributive Law of Set**

#### 집합 연산은 아래와 같이 분배 법칙이 성립한다.

#### **(1)** $ A \cap{} ( B \cup{} C ) = ( A \cap{} B ) \cup{} ( A \cap{} C ) $

#### **(2)** $ A \cup{} ( B \cap{} C ) = ( A \cup{} B ) \cap{} ( A \cup{} C ) $

> ## **서로소 집합(Disjoint Set)**

#### **(1) 정의** : 두 집합이 공통된 원소를 갖지 않을 때의 집합 
#### $ \Rrightarrow{} A \cap{} B = \empty{} = \{ \} $

<b></b>

> ## **분할 집합(Partition Set)**

#### **(1) 정의** : 공집합이 아닌 서로소 집합을 합쳐서 하나의 집합을 완성하는 부분집합들의 모음
#### $ \Rrightarrow{} \displaystyle{} A = \bigcup^{n}_{i=1}A_{i} \;\; \text{ s.t. } \, A_{i} \subseteq{} A, \;\; A_{i} \neq{} \empty{} , \; \; A_{i} \cap{} A_{j} = \empty{} \;\; (i \neq{} j) $
<img src="../img/01.1. Set (8).png" width="30%" height="30%"></img>

#### **(2) 예시** :
##### $ \hspace{0.15cm} $ ① 어떠한 분할된 집합 $ A $를 합집합으로 표현하면
##### $ \hspace{0.45cm} A = \{ 1, 2, 3, 4, 5 \}, \;\; A_{1} = \{ 1 \}, \;\; A_{2} = \{ 2, 4 \}, \;\; A_{3} = \{ 3, 5 \} $ 
##### $ \hspace{0.45cm} \therefore{} A_{1} \cup{} A_{2} \cup{} A_{3} = \displaystyle{} \bigcup^{3}_{i=1}A_{i} = A = \{ 1, 2, 3, 4, 5 \} $
##### $ \hspace{0.15cm} $ ② 어떠한 집합 $ B $를 분할된 부분집합으로 표현하면
##### $ \hspace{0.45cm} B = \{ a, b, c, d, e \}, \;\; \text{Partition}(B) = \{ B_{1}, B_{2}, B_{3} \} $
##### $ \hspace{0.45cm} \Rrightarrow{} B_{1} = \{ a \}, \;\; B_{2} = \{ b, d \}, \;\; B_{3} = \{ c, e \} \;\; \because{} \, B_{i} \neq{} \emptyset{}, \;\; B_{i} \cap{} B_{j} = \emptyset{} \;\; \text{for } \, i \neq{} j, \;\; \displaystyle{} \bigcup_{i=1}^{3}B_{i} = B $

<b></b>

> ## **곱집합(Product set;Cartesian product)**

#### **(1) 정의** : 두 개 이상의 집합에서 원소를 하나씩 뽑아 만든 모든 순서쌍의 집합
#### $ \Rrightarrow{} A \times{} B = \{ (a, b) \mid{} a \in{} A, \;\; b \in{} B \} $

##### **(`PLUS`)** 곱집합은 원소의 순서를 고려하기 때문에 $ (1,a) \neq{} (a,1), \;\; A \times{} B \neq{} B \times{} A $

#### **(2) 예시 :**
##### $ \hspace{0.15cm} $ ① 집합 $ A = \{ 1, 2 \}, \;\; B = \{ a, b, c \} $가 존재할 때 $ A \times{} B $와 $ B \times{} A $를 구하면
##### $ \hspace{0.45cm} A \times{} B = \{ (1,a), (1,b), (1,c), (2,a), (2,b), (2,c) \} $
##### $ \hspace{0.45cm} B \times{} A = \{ (a,1), (a,2), (b,1), (b,2), (c,1), (c,2) \} $
##### $ \hspace{0.15cm} $ ② $ \mathbb{R} \times{} \mathbb{R} = \{ (x, y) \mid{} x \in{} \mathbb{R}, \;\; y \in{} \mathbb{R} \} = \mathbb{R}^{2} $ ($ \text{space} $)
