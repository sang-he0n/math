> ## **가우시안 과정(Gaussian Process)**

#### **(1) 정의** : 가우시안 성질(모든 유한차원 분포의 정규성)을 만족하는 확률 과정들의 클래스
#### $ = $ 임의의 유한 개 시점을 택해도 항상 다변량 정규분포를 따르는 랜덤 벡터를 생성하는 가우시안 확률 과정 클래스
#### $ \Rrightarrow{} (X_{t_{1}},\dots{},X_{t_{n}}) \sim{} \mathcal{N}{}\big( m(t_{1:n}), C(t_{1:n},t_{1:n}) \big){}, \;\; \forall{} n \in{} \mathbb{N}{},\;\forall{} t_{1},\dots{},t_{n} \in{} T $

<b></b>

> ## **마팅게일 과정(Martingale Process)**

#### **(1) 정의** : 마팅게일 성질을 만족하는 확률 과정 클래스($ \approx{} $성질의 조합)
#### $ = $ 주어진 필터레이션에 대해 기대 이득이 $ 0 $인 공정한 게임 구조를 갖는 확률 과정 클래스
#### $ \Rrightarrow{} \{ X_{t} \}_{t \in{} T},\;\;  \{ \mathcal{F}_{t} \}_{t \in{} T} \text{ s.t. } \mathbb{E}{}\big[{} |X_{t}| \big]{} < \infty{},\;\; \forall{} t \in{} T,\;\;  \mathbb{E}{}\big[{} X_{t} \mid{} \mathcal{F}_{s} \big]{} = X_{s},\;\; \forall{} s < t $

<b></b>

> ## **마르코프 과정(Markov Process)**

#### **(1) 정의** : 마르코프 성질을 만족하는 확률 과정들의 클래스
#### $ = $ 미래 분포가 과거 전체가 아니라 현재 상태에만 의존하는 메모리리스 구조의 확률 과정 클래스
#### $ \Rrightarrow{} \mathbb{P}{}\big( X_{t_{n+1}} \in{} A \mid{} X_{t_{0}},\dots{},X_{t_{n}} \big){} = \mathbb{P}{}\big( X_{t_{n+1}} \in{} A \mid{} X_{t_{n}} \big){}, \;\; \forall{} t_{0} < \dots{} < t_{n+1}, \;\; \forall{} A \in{} \mathcal{B}(S) $
