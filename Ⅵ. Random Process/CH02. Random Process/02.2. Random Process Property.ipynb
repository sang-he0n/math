# CH02.2. **Random Process Property**

> ## **이차 모멘트 유한성(Second-order Finiteness)**

#### **(1) 정의** : 확률 과정 $ \{ X_{t} \}_{t \in{} T} $ 가 모든 시점에서 이차 모멘트가 유한함을 만족하는 성질
#### $ = $ 각 시점에서 제곱값의 기댓값이 유한하여 평균과 자기상관(자기공분산)을 정의할 수 있는 성질
#### $ \Rrightarrow{} \mathbb{E}{}\big[{} |X_{t}|^{2} \big]{} < \infty{}, \;\; \forall{} t \in{} T,\;\; m_{X}(t) := \mathbb{E}{}\big[{} X_{t} \big]{}, \;\; R_{X}(t_{1},t_{2}) := \mathbb{E}{}\big[{} X_{t_{1}} X_{t_{2}} \big]{} $

<b></b>

> ## **엄격한 정상성(Strict Stationarity)**

#### **(1) 정의** : 확률 과정 $ \{ X_{t} \}_{t \in{} T} $ 가 모든 유한 차수의 유한차원 분포에 대해 시간 이동에 대한 불변성을 만족하는 성질
#### $ = $ 시간축을 평행 이동해도 모든 유한차원 분포가 동일하게 유지되는 강한 의미의 정상성 성질
#### $ \Rrightarrow{} \mathcal{L}{}\big( X_{t_{1}},\dots{},X_{t_{n}} \big){} = \mathcal{L}{}\big( X_{t_{1}+\tau{}},\dots{},X_{t_{n}+\tau{}} \big){}, \;\; \forall{} n \in{} \mathbb{N}{}, \;\; \forall{} t_{1},\dots{},t_{n} \in{} T, \;\; \forall{} \tau{} \text{ s.t. } t_{i}+\tau{} \in{} T $

##### **(`PLUS`) 정상성(Stationarity)** : **[~]**

<b></b>

> ## **약한 정상성(Wide-Sense Stationarity;WSS)**

#### **(1) 정의** : 이차 모멘트 유한 확률 과정에서 평균이 상수이고 자기공분산이 시차에만 의존함을 만족하는 성질
#### $ = $ 평균과 이차 통계량 관점에서만 보면 시간에 따라 통계 구조가 변하지 않는 약한 의미의 정상성 성질
#### $ \Rrightarrow{} \mathbb{E}{}\big[{} |X_{t}|^{2} \big]{} < \infty{}, \;\; \forall{} t \in{} T,\;\; m_{X}(t) := \mathbb{E}{}\big[{} X_{t} \big]{} = m_{X}, \;\; C_{X}(t_{1},t_{2}) := \text{cov}{}\big( X_{t_{1}},X_{t_{2}} \big){} = C_{X}(t_{2}-t_{1}) = C_{X}(\tau{}) $

<b></b>

> ## **정상 증분(Stationary Increments)**

#### **(1) 정의** : 확률 과정 $ \{ X_{t} \}_{t \in{} T} $ 의 증분 분포가 시작 시점이 아니라 구간 길이에만 의존함을 만족하는 성질
#### $ = $ 같은 길이의 시간 구간이면 언제 시작하더라도 동일한 증분 분포를 갖는 증분 정상성 성질
#### $ \Rrightarrow{} \mathcal{L}{}\big( X_{t+h}-X_{t} \big){} = \mathcal{L}{}\big( X_{s+h}-X_{s} \big){}, \;\; \forall{} s,t \in{} T, \;\; \forall{} h > 0 \text{ s.t. } s+h,t+h \in{} T $

<b></b>

> ## **독립 증분(Independent Increments)** 

#### **(1) 정의** : 확률 과정 $ \{ X_{t} \}_{t \in{} T} $ 의 서로 겹치지 않는 시간 구간들에서의 증분이 서로 독립임을 만족하는 성질
#### $ = $ 분리된 시간 구간마다의 변화량이 서로 영향을 주지 않는 독립 구조의 증분 성질
#### $ \Rrightarrow{} X_{t_{1}}-X_{t_{0}},\; X_{t_{2}}-X_{t_{1}},\;\dots{},\; X_{t_{n}}-X_{t_{n-1}} \text{ mutually independent}, \;\; \forall{} t_{0} < t_{1} < \dots{} < t_{n} $

<b></b>

> ## **연속 경로(Continuous Sample paths)** 

#### **(1) 정의** : 거의 모든 표본 경로가 시간에 대해 연속 함수가 됨을 만족하는 성질
#### $ = $ 개별 실현 경로가 시간축 위에서 끊김 없이 변화하는 연속 경로를 갖는 성질
#### $ \Rrightarrow{} \mathbb{P}{}\Big( \big\{ \omega{} \in{} \Omega{} : t \mapsto X_{t}(\omega{}) \text{ is continuous on } T \big\} \Big) = 1 $

<b></b>

> ## **우연속·좌극한 존재 경로(right-continuous with left limits;càdlàg)** 

#### **(1) 정의** : 거의 모든 표본 경로가 모든 시점에서 우연속이고 좌측 극한이 존재함을 만족하는 성질
#### $ = $ 경로가 오른쪽에서는 연속이고 왼쪽에서는 점프 직전 값이 항상 존재하는 형태의 경로를 갖는 성질
#### $ \Rrightarrow{} \mathbb{P}{}\Big( \big\{ \omega{} \in{} \Omega{} : \lim{}_{s \downarrow{} t} X_{s}(\omega{}) = X_{t}(\omega{}), \; \lim{}_{s \uparrow{} t} X_{s}(\omega{}) =: X_{t-}(\omega{}), \;\forall{} t \in{} T \big\} \Big) = 1 $

<b></b>

> ## **에르고딕 성질(Ergodic Property)**

#### **(1) 정의** : (적어도 WSS 인) 확률 과정에서 시간 평균이 ensemble 평균(기댓값)으로 거의 확실히 수렴함을 만족하는 성질
#### $ = $ 충분히 긴 한 경로의 시간 평균만으로도 전체 집단 평균을 회복할 수 있는 에르고딕 특성을 갖는 성질
#### $ \Rrightarrow{} \overline{X}_{T}(\omega{}) := \frac{1}{T} \int_{0}^{T} X_{t}(\omega{}) \, dt \xrightarrow[T \rightarrow{} \infty{}]{} m_{X} := \mathbb{E}{}\big[{} X_{t} \big]{} \;\; \text{a.s.} $

<b></b>

> ## **가우시안 성질(Gaussian Property)**

#### **(1) 정의** : 모든 유한차원 분포가 다변량 정규분포가 됨을 만족하는 성질
#### $ = $ 임의의 유한 개 시점을 선택해도 항상 가우시안 랜덤 벡터로 표현되는 가우시안 구조를 갖는 성질
#### $ \Rrightarrow{} (X_{t_{1}},\dots{},X_{t_{n}}) \sim \mathcal{N}{}\big( m(t_{1:n}), C(t_{1:n},t_{1:n}) \big){}, \;\; \forall{} n \in{} \mathbb{N}{}, \;\; \forall{} t_{1},\dots{},t_{n} \in{} T $

<b></b>

> ## **마팅게일 성질(Martingale Property)**

#### **(1) 정의** : 필터레이션 $ \{ \mathcal{F}_{t} \}_{t \in{} T} $ 에 대해 조건부 기댓값이 $ \mathbb{E}{}\big[{} X_{t} \mid \mathcal{F}_{s} \big]{} = X_{s} $ 를 만족하는 공정성 성질
#### $ = $ 과거 정보까지 고려했을 때 현재 값이 미래 값들의 조건부 기댓값과 일치하는, 기대 이득이 $ 0 $인 마팅게일 구조를 갖는 성질
#### $ \Rrightarrow{} \mathbb{E}{}\big[{} |X_{t}| \big]{} < \infty{}, \;\; \forall{} t \in{} T,\;\; \mathbb{E}{}\big[{} X_{t} \mid \mathcal{F}_{s} \big]{} = X_{s}, \;\; \forall{} s < t $

##### **(`PLUS`) 필터레이션(Filteration)** : 시간이 지날수록 늘어나는 정보의 집합을 $ \sigma{} $-대수들로 표현한 것

<b></b>

> ## **마르코프 성질(Markov Property)**

#### **(1) 정의** : 미래 분포가 과거 전체가 아니라 현재 상태에만 의존함을 만족하는 메모리리스 성질
#### $ = $ 현재 상태가 과거 정보를 모두 요약하여, 현재 상태만 알면 미래의 조건부 분포를 결정할 수 있는 마르코프 구조를 갖는 성질
#### $ \Rrightarrow{} \mathbb{P}{}\big( X_{t_{n+1}} \in{} A \mid X_{t_{0}},\dots{},X_{t_{n}} \big){} = \mathbb{P}{}\big( X_{t_{n+1}} \in{} A \mid X_{t_{n}} \big){}, \;\; \forall{} n \in{} \mathbb{N}{}, \;\; \forall{} t_{0} < \dots{} < t_{n+1}, \;\; \forall{} A \in{} \mathcal{B}(S) $
