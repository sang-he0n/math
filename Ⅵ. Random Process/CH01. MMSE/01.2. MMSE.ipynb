# CH01.2. **Minimum Mean Square Error**

> ## **MSE 최소화(Minimum Mean Square Error;MMSE)**

#### **(1) 정의** : 힐베르트 공간에서 MSE를 최소화하는 추정기(estimator) $ \hat{\textbf{Y}} $를 찾는 문제
#### $ = $ 힐베르트 공간에서 타겟 랜덤 벡터 $ \textbf{Y} $까지의 최소 거리를 정의하는 추정기 $ \hat{\textbf{Y}} $를 찾는 문제
#### $ \Rrightarrow{} \displaystyle{} \hat{\textbf{Y}}^{*} = \argmin_{\hat{\textbf{Y}}\in{}\mathcal{V}} \mathbb{E}\big[{}\|\textbf{Y}-\hat{\textbf{Y}}\|^{2}_{2}\big]{} $ 
#### $ \hspace{0.45cm} \text{where } \, \mathcal{V} \subset{} L^{2}(\Omega{};\mathbb{R}^{m}), \;\;  \langle{} \textbf{Y}, \hat{\textbf{Y}} \rangle{} := \mathbb{E}[\textbf{Y}^{T}\hat{\textbf{Y}}] $

<b></b>

> ## **$ L^{2} $ 공간에서 직교 투영(Orthogonal projection in $ \, L^{2} $)**

#### **(1) 정의** : 힐베르트 공간의 어떠한 랜덤 벡터를 닫힌 부분 공간에 직교하게 내리는 연산
#### $ \Rrightarrow{} \Pi{}_{\mathcal{V}}(\textbf{Y}) := \displaystyle{} \argmin_{\hat{\textbf{Y}}\in{}\mathcal{V}} \mathbb{E}\big[{}\|\textbf{Y}-\hat{\textbf{Y}}\|^{2}_{2}\big]{} $ 
#### $ \hspace{0.45cm} \text{where } \, \langle{} \textbf{Y}, \hat{\textbf{Y}} \rangle{} := \mathbb{E}[\textbf{Y}^{T}\hat{\textbf{Y}}], \;\; \, \mathcal{V} \, \text{ is closed subspace in } L^{2}(\Omega{};\mathbb{R}^{m}) $

#### **(2) 성질** : 
##### $ \hspace{0.15cm} $ ① $ \Pi{}_{\mathcal{V}}\big({}\Pi{}_{\mathcal{V}}(\textbf{Y})\big){} = \Pi{}_{\mathcal{V}}(\textbf{Y}) \;\; \because{} \Pi{}_{\mathcal{V}}(\textbf{Y}) \in{} \mathcal{V} $
##### $ \hspace{0.15cm} $ ② $ \Pi{}_{\mathcal{V}}(a_{1}\textbf{Y}_{1}+a_{2}\textbf{Y}_{2}) = a_{1} \, \Pi{}_{\mathcal{V}}(\textbf{Y}_{1}) + a_{2} \, \Pi{}_{\mathcal{V}}(\textbf{Y}_{2}) $

> ## (Theorem 1-5) **Orthogonality principle**

#### 힐베르트 공간에서 MSE 문제는 부분공간 $ \mathcal{V} $가 닫혀있을 때(closed)일 때 아래가 성립한다. 

#### **(0)** 부분공간 $ \mathcal{V} $에서 최적해 $ \hat{\textbf{Y}}^{*} $가 존재한다.
##### $ \Rrightarrow{} \displaystyle{} \exists{} \, \hat{\textbf{Y}}^{*} \in{} \mathcal{V} \;\; \text{ s.t. } \, \mathbb{E}\big[{}\|\textbf{Y}-\hat{\textbf{Y}}^{*}\|^{2}_{2}\big]{} = \inf_{\hat{\textbf{Y}} \in{} \mathcal{V}} \mathbb{E}\big[{}\|\textbf{Y}-\hat{\textbf{Y}}\|^{2}_{2}\big]{} \;\; (\text{infimum})$

##### **(`WHY?`)** 
##### $ \hspace{0.15cm} \text{let } \, d := \displaystyle{} \inf_{Z \in{} \mathcal{V}}\|\textbf{Y}-Z\|. $
##### $ \hspace{0.15cm} $ **[~]**

#### **(1)** 최적해 $ \hat{\textbf{Y}}^{*} $의 오차 $ \boldsymbol{\epsilon{}}^{*} $은 부분 공간 $ \mathcal{V} $와 직교한다. ($ = $ 부분 공간 $ \mathcal{V} $의 모든 벡터와 직교한다.)
#### $ \Rrightarrow{} \langle{} \boldsymbol{\epsilon{}}^{*},\hat{\textbf{Y}} \rangle{} = \mathbb{E}[(\boldsymbol{\epsilon{}}^{*})^{T}\hat{\textbf{Y}}] = 0 \; \Leftrightarrow{} \; \boldsymbol{\epsilon{}}^{*} \perp{} \mathcal{V} $
#### $ \hspace{0.45cm} \text{ where } \, \boldsymbol{\epsilon{}}^{*} = \textbf{Y} - \hat{\textbf{Y}}^{*}, \;\; \forall{} \, \hat{\textbf{Y}} \in{} \mathcal{V} $

##### **(`WHY?`)** 
##### $ \hspace{0.15cm} \text{let } \, \hat{\textbf{Y}}_{t} := \hat{\textbf{Y}}^{*} + t \textbf{V}, \;\; f(t) := \mathbb{E}[\| \textbf{Y} - \hat{\textbf{Y}}_{t} \|^{2}_{2}] \;\; \text{ where } \, t \in{} \mathbb{R}, \;\; \forall{} \, \textbf{V} \in{} \mathcal{V}, $ 
##### $ \hspace{0.15cm} f(t) = \mathbb{E}[\| \textbf{Y} - \hat{\textbf{Y}}_{t} \|^{2}_{2}] = \mathbb{E}[\|\textbf{Y}-(\hat{\textbf{Y}}^{*}+t\textbf{V}) \|^{2}_{2}] = \mathbb{E}[\|\textbf{Y}-\hat{\textbf{Y}}^{*}\|^{2}_{2}]-2t\mathbb{E}[(\textbf{Y}-\hat{\textbf{Y}}^{*})^{T}\textbf{V}] + t^{2} \mathbb{E}[\|\textbf{V}\|^{2}_{2}] $
##### $ \hspace{0.15cm} \frac{\mathrm{d}}{\mathrm{d}t}f(t) = \frac{\mathrm{d}}{\mathrm{d}t}\bigg({}\mathbb{E}[\|\textbf{Y}-\hat{\textbf{Y}}^{*}\|^{2}_{2}]-2t\mathbb{E}[(\textbf{Y}-\hat{\textbf{Y}}^{*})^{T}\textbf{V}] + t^{2} \mathbb{E}[\|\textbf{V}\|^{2}_{2}] \bigg){} $
##### $ \hspace{1.175cm} = -2\mathbb{E}[(\textbf{Y}-\hat{\textbf{Y}}^{*})^{T}\textbf{V}] + 2t \mathbb{E}[\|\textbf{V}\|^{2}_{2}] $
##### $ \hspace{0.15cm} \left.{} \frac{\mathrm{d}}{\mathrm{d}t} f(t) \right|{}_{t=0} = -2\mathbb{E}[(\textbf{Y}-\hat{\textbf{Y}}^{*})^{T}\textbf{V}] = 0 \;\; \because{} \hat{\textbf{Y}}_{t=0}=\hat{\textbf{Y}}^{*} \, \text{ is best-estimator over } \, \mathcal{V}, \;\; f(0) \leq{} f(t) \;\; \forall{} \, t $
##### $ \hspace{0.15cm} \therefore{} \mathbb{E}[(\textbf{Y}-\hat{\textbf{Y}}^{*})^{T}\textbf{V}] = 0 \;\;\;\; \forall{} \, \textbf{V} \in{} \mathcal{V} $ 

#### **(2)** 최적해 $ \hat{\textbf{Y}}^{*} $는 정사영으로 유일하게 정의된다
#### $ \Rrightarrow{} \exist{}! \, \hat{\textbf{Y}}^{*} = \Pi{}_{\mathcal{V}}(\textbf{Y}) \;\; \text{ s.t. } \, \mathbb{E}[\|\textbf{Y}-\hat{\textbf{Y}}^{*}\|^{2}_{2}] \leq{} \mathbb{E}[\|\textbf{Y}-\hat{\textbf{Y}}\|^{2}_{2}], \;\; \forall{} \, \hat{\textbf{Y}} \in{} \mathcal{V}  $

##### **(`WHY?`)** 
##### $ \hspace{0.15cm} \text{assume that there are } \, \hat{\textbf{Y}}^{*}, \; \tilde{\textbf{Y}} \, \text{ that minimize the MSE. and } \, \hat{\textbf{Y}}^{*} \neq{} \tilde{\textbf{Y}} $
##### $ \hspace{0.15cm} \text{and let } \, m := \mathbb{E}{}\big[{}\|\textbf{Y}-\hat{\textbf{Y}}^{*}\|^{2}_{2}\big]{} = \mathbb{E}{}\big[{}\|\textbf{Y}-\tilde{\textbf{Y}}\|^{2}_{2}\big]{} = \displaystyle{} \inf_{\hat{\textbf{Y}} \in{} \mathcal{V}} \mathbb{E}{}\big[{}\|\textbf{Y}-\hat{\textbf{Y}}\|^{2}_{2}\big]{} $
##### $ \hspace{0.15cm} \mathbb{E}[\|\textbf{Y}-\hat{\textbf{Y}}^{*}\|^{2}_{2}] = \mathbb{E}[\|\textbf{Y}-\hat{\textbf{Y}}^{*}\|^{2}_{2}] $
##### $ \hspace{2.4cm}  = \mathbb{E}[\|\textbf{Y}-\hat{\textbf{Y}}^{*}+\tilde{\textbf{Y}}-\tilde{\textbf{Y}}\|^{2}_{2}] $
##### $ \hspace{2.4cm} = \mathbb{E}[\|(\textbf{Y}-\tilde{\textbf{Y}})+(\tilde{\textbf{Y}}-\hat{\textbf{Y}}^{*})\|^{2}_{2}] $
##### $ \hspace{2.4cm} = \mathbb{E}[\|\textbf{Y}-\tilde{\textbf{Y}}\|^{2}_{2}]+2\mathbb{E}\big[{}(\textbf{Y}-\tilde{\textbf{Y}})^{T}(\tilde{\textbf{Y}}-\hat{\textbf{Y}}^{*})\big]{}+\mathbb{E}[\|\tilde{\textbf{Y}}-\hat{\textbf{Y}}^{*}\|^{2}_{2}] $
##### $ \hspace{2.4cm} = \mathbb{E}[\|\textbf{Y}-\tilde{\textbf{Y}}\|^{2}_{2}]+\mathbb{E}[\|\tilde{\textbf{Y}}-\hat{\textbf{Y}}^{*}\|^{2}_{2}] \;\; \because{} \tilde{\textbf{Y}}-\hat{\textbf{Y}}^{*} \in{} \mathcal{V}, \;\; \langle{} \textbf{Y}-\tilde{\textbf{Y}}, \tilde{\textbf{Y}}-\hat{\textbf{Y}}^{*} \rangle{} = 0 $
##### $ \hspace{0.15cm} m = m + \mathbb{E}[\|\tilde{\textbf{Y}}-\hat{\textbf{Y}}^{*}\|^{2}_{2}] $ 
##### $ \hspace{0.15cm} 0 = \mathbb{E}[\|\tilde{\textbf{Y}}-\hat{\textbf{Y}}^{*}\|^{2}_{2}] \; \Rightarrow{} \; \hat{\textbf{Y}}^{*} = \tilde{\textbf{Y}} \;\; (\text{contradiction}) \;\; \because{} \hat{\textbf{Y}}^{*} \neq{} \tilde{\textbf{Y}} $ 
##### $ \hspace{0.15cm} \therefore{} \hat{\textbf{Y}}^{*} = \tilde{\textbf{Y}} $

#### **(3)** MMSE를 아래와 같이 표현할 수 있다.
#### $ \Rrightarrow{} \mathbb{E}[\|\boldsymbol{\epsilon{}}^{*}\|^{2}_{2}] = \mathbb{E}[\|\textbf{Y}\|^{2}_{2}] - \mathbb{E}[\|\hat{\textbf{Y}}^{*}\|^{2}_{2}] \;\; \because{} \text{pythagorean theorem} $

> ## (Theorem 1-6) **Subspace calculus**

#### **(1)** (닫힌) 부분공간이 $ \mathcal{V_{2}} \subset{} \mathcal{V_{1}} $일 때, $ \, \mathcal{V}_{2}$으로 정사영한 결과와 $\mathcal{V}_{1}$ 이후 $ \mathcal{V}_{2}$로 정사영한 결과가 동일하다.
#### $ \Rrightarrow{} \Pi{}_{\mathcal{V_{2}}}(\textbf{Y}) = \Pi{}_{\mathcal{V_{2}}} \big({} \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) \big){} \;\; \text{ s.t. }\, \mathcal{V}_{2} \subset{} \mathcal{V}_{1} $

##### **(`WHY?`)** 
##### $ \hspace{0.15cm} \text{let } \, \forall{} \, \textbf{V} \in{} \mathcal{V}_{2}. $
##### $ \hspace{0.15cm} <\textbf{Y} - \Pi{}_{\mathcal{V_{2}}} \big({} \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) \big){}, \textbf{V}> = <\textbf{Y} - \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}), \textbf{V}> + <\Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) - \Pi{}_{\mathcal{V_{2}}} \big({} \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) \big){}, \textbf{V}> $
##### $ \hspace{4.075cm} = 0 + <\Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) - \Pi{}_{\mathcal{V_{2}}} \big({} \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) \big){}, \textbf{V}> \;\; \because{} \textbf{Y} - \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) \perp{} \textbf{V}, \;\; \textbf{V} \in{} \mathcal{V}_{2} \subset{} \mathcal{V}_{1} $
##### $ \hspace{4.075cm} = 0 \;\; \because{} \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) - \Pi{}_{\mathcal{V_{2}}} \big({} \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) \big){} \perp{} \textbf{V}, \;\; \textbf{V} \in{} \mathcal{V}_{2} $
##### $ \hspace{0.15cm} \therefore{} \Pi{}_{\mathcal{V_{2}}}(\textbf{Y}) = \Pi{}_{\mathcal{V_{2}}} \big({} \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) \big){} $ 

#### **(2)** 부분공간이 $ \mathcal{V_{2}} \subset{} \mathcal{V_{1}} $일 때, 오차 $ \boldsymbol{\epsilon{}}^{*} $을 두 부분으로 분해할 수 있다.
#### $ \Rrightarrow{} \boldsymbol{\epsilon{}}^{*} := \textbf{Y} - \Pi{}_{\mathcal{V}_{2}}(\textbf{Y}) = \boldsymbol{\epsilon{}}_{1}^{*} + \boldsymbol{\epsilon{}}_{2}^{*} $
#### $ \hspace{0.45cm} \text{s.t. } \, \mathcal{V}_{2} \subset{} \mathcal{V}_{1} $
#### $ \hspace{0.45cm} \text{where } \, \boldsymbol{\epsilon{}}_{1}^{*} := \textbf{Y} - \Pi{}_{\mathcal{V}_{1}}(\textbf{Y}), \;\; \boldsymbol{\epsilon{}}_{2}^{*} := \Pi{}_{\mathcal{V}_{1}}(\textbf{Y}) - \Pi{}_{\mathcal{V}_{2}}\big({}\Pi{}_{\mathcal{V}_{1}}(\textbf{Y})\big){} $

##### **(`WHY?`)** 
##### $ \hspace{0.15cm} \boldsymbol{\epsilon{}}^{*}_{1} + \boldsymbol{\epsilon{}}_{2}^{*} = \big({} \textbf{Y} - \Pi{}_{\mathcal{V}_{1}}(\textbf{Y}) \big){} + \bigg({} \Pi{}_{\mathcal{V}_{1}}(\textbf{Y}) - \Pi{}_{\mathcal{V}_{2}}\big({}\Pi{}_{\mathcal{V}_{1}}(\textbf{Y})\big){} \bigg){} $
##### $ \hspace{1.3cm} = \big({} \textbf{Y} - \Pi{}_{\mathcal{V}_{1}}(\textbf{Y}) \big){} + \big({} \Pi{}_{\mathcal{V}_{1}}(\textbf{Y}) - \Pi{}_{\mathcal{V}_{2}}(\textbf{Y}) \big){} \;\; \because{} \Pi{}_{\mathcal{V_{2}}}(\textbf{Y}) = \Pi{}_{\mathcal{V_{2}}} \big({} \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) \big){} $
##### $ \hspace{1.3cm} = \textbf{Y} - \Pi{}_{\mathcal{V}_{2}}(\textbf{Y}) $
##### $ \hspace{0.15cm} \therefore{} \boldsymbol{\epsilon{}}^{*} = \boldsymbol{\epsilon{}}^{*}_{1} + \boldsymbol{\epsilon{}}^{*}_{2} $

#### **(3)** 부분공간이 $ \mathcal{V_{2}} \subset{} \mathcal{V_{1}} $일 때, $ \mathcal{V}_{1} $의 MSE가 $ \mathcal{V}_{2} $의 MSE보다 작거나 같다.
#### $ \Rrightarrow{} \displaystyle{} \min_{\hat{\textbf{Y}}\in{}\mathcal{V}_{1}} \mathbb{E}[\|\textbf{Y}-\hat{\textbf{Y}}\|^{2}_{2}] \leq{} \min_{\hat{\textbf{Y}}\in{}\mathcal{V}_{2}}\mathbb{E}[\|\textbf{Y}-\hat{\textbf{Y}}\|^{2}_{2}] \;\; \text{ s.t. } \, \mathcal{V}_{2} \subset{} \mathcal{V}_{1} $

#### **(4)** 부분공간 $ \mathcal{V}_{1}, \mathcal{V}_{2} $가 직교하고 이를 직교합하는 $ \mathcal{V} $가 있을 때, $ \mathcal{V} $로 정사영한 결과는 두 부분공간으로 정사영한 결과와 같다.
#### $ \Rrightarrow{} \Pi{}_{\mathcal{V}}(\textbf{Y}) = \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) + \Pi{}_{\mathcal{V_{2}}}(\textbf{Y}) $ 
#### $ \hspace{0.45cm} \text{s.t. } \, \mathcal{V} = \mathcal{V}_{1} \oplus{} \mathcal{V}_{2}, \;\; \mathcal{V}_{1} \perp{} \mathcal{V}_{2} $
#### $ \hspace{0.45cm} \text{where } \oplus{} \text{ is orthogonal sum operator}. $

##### **(`WHY?`)** 
##### $ \hspace{0.15cm} \text{let } \, \textbf{Z} := \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) + \Pi{}_{\mathcal{V_{2}}}(\textbf{Y}) \in{} \mathcal{V}, \;\; \textbf{V} := \textbf{V}_{1} + \textbf{V}_{2} \in{} \mathcal{V} \;\; \text{ where } \, \textbf{V}_{i} \in{} \mathcal{V}_{i} \; (i=1,2) $
##### $ \hspace{0.15cm} <\textbf{Y} - \textbf{Z}, \textbf{V}> = <\textbf{Y} - \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) - \Pi{}_{\mathcal{V_{2}}}(\textbf{Y}), \textbf{V}_{1}+\textbf{V}_{2}> $
##### $ \hspace{2.4cm} = <\textbf{Y} - \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) - \Pi{}_{\mathcal{V_{2}}}(\textbf{Y}), \textbf{V}_{1}> + <\textbf{Y} - \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) - \Pi{}_{\mathcal{V_{2}}}(\textbf{Y}), \textbf{V}_{2}> $
##### $ \hspace{2.4cm} = <\textbf{Y} - \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}), \textbf{V}_{1}> - <\Pi{}_{\mathcal{V_{2}}}(\textbf{Y}), \textbf{V}_{1}> + <\textbf{Y} - \Pi{}_{\mathcal{V_{2}}}(\textbf{Y}), \textbf{V}_{2}> - <\Pi{}_{\mathcal{V_{1}}}(\textbf{Y}), \textbf{V}_{2}> $
##### $ \hspace{2.4cm} = 0 - <\Pi{}_{\mathcal{V_{2}}}(\textbf{Y}), \textbf{V}_{1}> + <\textbf{Y} - \Pi{}_{\mathcal{V_{2}}}(\textbf{Y}), \textbf{V}_{2}> - <\Pi{}_{\mathcal{V_{1}}}(\textbf{Y}), \textbf{V}_{2}> \;\; \because{} \textbf{Y} - \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) \perp{} \textbf{V}_{1} $
##### $ \hspace{2.4cm} = 0 + <\textbf{Y} - \Pi{}_{\mathcal{V_{2}}}(\textbf{Y}), \textbf{V}_{2}> - <\Pi{}_{\mathcal{V_{1}}}(\textbf{Y}), \textbf{V}_{2}> \;\; \because{} \Pi{}_{\mathcal{V_{2}}}(\textbf{Y}) \in{} \mathcal{V}_{2} \perp{} \mathcal{V}_{1} $
##### $ \hspace{2.4cm} = 0 - <\Pi{}_{\mathcal{V_{1}}}(\textbf{Y}), \textbf{V}_{2}> \;\; \because{} \textbf{Y} - \Pi{}_{\mathcal{V_{2}}}(\textbf{Y}) \perp{} \textbf{V}_{2} $
##### $ \hspace{2.4cm} = 0 \;\; \because{} \Pi{}_{\mathcal{V_{1}}}(\textbf{Y}) \in{} \mathcal{V}_{1} \perp{} \mathcal{V}_{2} $
##### $ \hspace{0.15cm} \therefore{} \textbf{Y} - \textbf{Z} \perp{} \textbf{V} \Rightarrow{} \Pi{}_{\mathcal{V}}(\textbf{Y}) $
