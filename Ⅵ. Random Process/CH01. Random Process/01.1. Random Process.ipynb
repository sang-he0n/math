# CH03.1. **Random Process**

> ## **시간 집합(Time set;Index set)**

#### **(1) 정의** : 어떤 시스템(모델링할 현실 대상)의 관찰 시점을 정의하는 순서가 존재하는 집합(ordered set)
#### $ \Rrightarrow{} T := \begin{cases} \{ t_{0} < t_{1} < t_{2} < \cdots{} \} & (\text{discrete}) \\ \text{interval} & (\text{continuous}) \end{cases} $

<b></b>

> ## **상태 공간(State Space)**

#### **(1) 정의** : 어떤 시스템의 한 시점에서 가능한 상태값들을 표현하기 위해 사용하는 집합
#### $ \Rrightarrow{} S := \begin{cases} \{ s_{0}, s_{1}, s_{2}, \cdots{} \} & (\text{discrete}) \\ \mathbb{R} \, \text{ or } \, \mathbb{R}^{m} & (\text{continuous})\end{cases} $
#### $ \hspace{0.45cm} \text{where } \, s_{i} \, \text{ is state} $

<b></b>

> ## **확률 과정(Random Process)**

#### **(1) 정의** : 시간(혹은 인덱스)에 따라 정의된, 상태공간 값을 갖는 확률변수들의 집합족(family;인덱스로 묶인 객체들의 묶음)
#### $ = $ 시간에 따라 확률적으로 변화하는 상태들의 집합족
#### $ \Rrightarrow{} X : \Omega{} \times{} T \rightarrow{} S \;\; \Leftrightarrow{} \;\; \{ X_{t} \}_{t\in{}T}, \;\; X_{t}(\omega) := X(\omega{},t) $
#### $ \hspace{0.45cm} \text{s.t. } \, X_{t} : (\Omega{}, \mathcal{F}) \rightarrow{} (S, \mathcal{B}(S)) $
#### $ \hspace{0.45cm} \text{where } \, T \, \text{ is time set}, \;\; S \, \text{ is state space}, \;\; \omega{} \in{} \Omega{}, \;\; t \in{} T, \;\; X(\omega{},t) \in{} S $

#### **(2) 종류** :
##### $ \hspace{0.15cm} $ ① 이산-시간, 이산-값(discrete-time, discrete-value;DTDV)
##### $ \hspace{0.15cm} $ ② 이산-시간, 연속-값(discrete-time, continuous-value;DTCV)
##### $ \hspace{0.15cm} $ ③ 연속-시간, 이산-값(continuous-time, discrete-value;CTDV)
##### $ \hspace{0.15cm} $ ④ 연속-시간, 연속-값(continuous-time, continuous-value;CTCV)

<b></b>

> ## (Theorem 1-1) **Two Representations of Random Process**

#### **(1)** 확률 과정에서 시간 인덱스를 $ \, t \in{} T $로 고정하면, 확률 변수 혹은 랜덤 벡터가 된다.
#### $ = $ 확률 과정에서 어떠한 시간을 하나 지정하면, 그 단면은 확률 변수 혹은 랜덤 벡터가 된다.
#### $ \Rrightarrow{} X(\cdot{},t) = X_{t} : \Omega{} \rightarrow{} S $

##### **(2)** 확률 과정에서 결과(outcome)를 $ \, \omega{} \in{} \Omega{} $로 고정하면, 표본 경로(sample path)가 된다.
##### $ = $ 확률과정에서 어떠한 결과 하나가 실제로 관측됐을 때, 시간을 변수로 갖는 함수는 표본 경로가 된다
#### $ \Rrightarrow{} X(\omega{}, \cdot{}) : T \rightarrow{} S $

> ## **평균 함수(Mean function)**

#### **(1) 정의** : 각 시간 인덱스 $ t $에서의 확률 과정 $ X $의 기대값을 정리한 함수
#### $ = $ 시간에 따른 확률 과정 $ X $의 평균 궤적
#### $ \Rrightarrow{} m_{X}(t) := \mathbb{E}[X_{t}] $

<b></b>

> ## **자기-상관 함수(Auto-Correlation function)**

#### **(1) 정의** : **[~]**
#### $ = $ 두 시점에서의 값의 곱에 대한 기댓값, second-order 특징을 요약
#### $ \Rrightarrow{} R_{X}(t_{1},t_{2}) := \mathbb{E}[X_{t_{1}} X_{t_{2}}] $

<b></b>

> ## **자기-공분산 함수(Auto-Covariance function)**

#### **(1) 정의** : **[~]**
#### $ \Rrightarrow{} C_{X}(t_{1},t_{2}) := \text{cov}(X_{t_{1}},X_{t_{2}}) $
#### $ \hspace{2.125cm} = \mathbb{E}[\big({}X_{t_{1}} - m_{X}(t_{1})\big){} \big({} X_{t_{2}} - m_{X}(t_{2})\big){}] $ 

<b></b>

> ## **증분(Increment;변화량)**

#### **(1) 정의** : 두 시점 $ s < t \, $ 사이에서 확률 과정 $ X $의 상태가 얼마나 변했는지를 나타내는 새로운 확률 변수
#### $ \Rrightarrow{} \Delta{}X(s,t] := X_{t} - X_{s} \;\; \text{ s.t. } s < t $
