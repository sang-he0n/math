# CH03.2. **Random Process Property**

> ## **이차 모멘트 유한성(Second-order Finiteness)**

#### **(1) 정의** : 확률 과정 $ \{ X_{t} \} $가 모든 시점에서 이차 모멘트가 유한함을 만족하는 성질
#### $ = $ 각 시점에서 제곱값의 기댓값이 유한하여 평균과 자기상관(자기공분산)을 정의할 수 있는 성질
#### $ \Rrightarrow{} \mathbb{E} \big[{} |X_{t}|^{2} \big]{} < \infty{} \;\;\;\; \forall{} t \in{} T $

<b></b>

> ## **엄격한 정상성(Strict Stationarity)**

#### **(1) 정의** : 확률 과정 $ \{ X_{t} \} $가 모든 유한 차수의 유한 차원 분포에 대해 시간 이동에 대한 불변성을 만족하는 성질
#### $ = $ 시간축을 평행 이동해도 모든 유한 차원 분포가 동일하게 유지되는 성질
#### $ \Rrightarrow{} \mathcal{L}\big( X_{t_{1}},\cdots{},X_{t_{n}} \big){} = \mathcal{L}\big( X_{t_{1}+\tau{}},\cdots{},X_{t_{n}+\tau{}} \big){} \;\;\;\; \forall{} \, t_{i} \in{} T $
#### $ \hspace{0.45cm} \text{s.t. } \, t_{i}+\tau{} \in{} T $

##### **(`PLUS`) 정상성(Stationarity)** : 확률 과정의 시간(인덱스) 축을 이동해도 통계적 특징이 변하지 않는 성질

<b></b>

> ## **약한 정상성(Wide-Sense Stationarity;WSS)**

#### **(1) 정의** : 이차 모멘트 유한 확률 과정 $ \{ X_{t} \} $에서 평균이 상수이고 자기공분산이 시차에만 의존함을 만족하는 성질
#### $ = $ 시간축을 평행 이동해도 평균은 일정하고, 공분 함수는 시간차(lag)에만 의존하는 성질
#### $ \Rrightarrow{} m_{X}(t_{i}) = m \;\; \text{ and } \;\; C_{X}(t_{1}+\tau{},t_{2}+\tau{}) = C_{X}(t_{1},t_{2}) , \;\;\;\; \forall{} t_{i} \in{} T $
#### $ \hspace{0.45cm} \text{s.t. } \, \mathbb{E} \big[{} |X_{t}|^{2} \big]{} < \infty{}, \;\; t_{i}+\tau{} \in{} T $

<b></b>

> ## **정상 증분(Stationary Increments)**

#### **(1) 정의** : 확률 과정의 증분 $ \Delta{} X(t_{0},t_{1}] $에 대한 유한 차원 분포가 구간 길이에만 의존함을 만족하는 성질
#### $ = $ 각 시간 간격이 동일하면 언제 시작하더라도 동일한 증분 유한 차원 분포를 갖는 성질
#### $ \Rrightarrow{} \mathcal{L}\big({} \Delta{} X(t_{0},t_{1}],\cdots{}, \Delta{} X(t_{n-1},t_{n}] \big){} = \mathcal{L}\big({}{} \Delta{} X(s_{0},s_{1}],\cdots{},\Delta{} X(s_{n-1},s_{n}] \big){} $
#### $ \hspace{0.45cm} \text{whenever } t_{i}-t_{i-1} = s_{i}-s_{i-1}, \;\; 1 \leq{} i \leq{} n $

<b></b>

> ## **독립 증분(Independent Increments)** 

#### **(1) 정의** : 확률 과정에서 서로 겹치지 않는 시간 구간들에서의 증분 $ \Delta{} X(t_{0},t_{1}] $이 서로 독립임을 만족하는 성질
#### $ = $ 분리된 시간 구간마다의 변화량이 서로 영향을 주지 않는 성질
#### $ \Rrightarrow{} \{ \Delta{} X(t_{0},t_{1}],\; \Delta{} X(t_{1},t_{2}],\;\dots,\; \Delta{} X(t_{n-1},t_{n}] \} \, \text{ are mutually independent}. \;\;\;\; t_{0} < t_{1} < \dots{} < t_{n} $

<b></b>

> ## **연속 경로(Continuous Sample paths)** 

#### **(1) 정의** : 개별 결과 $ \omega{} \in{} \Omega{} $ 에 대해 정의되는 표본 경로 $ X_{t}(\omega{}) $가 거의 모든(almost surely) $ \omega{} $ 에 대해 연속인 성질
#### $ = $ 거의 모든 표본 경로가 시간 축 위에서 끊김 없는(연속인) 경로를 갖는 성질
#### $ \Rrightarrow{} \mathbb{P} \Big({} \{ \omega{} \in \Omega{} \mid{} t \mapsto X_{t}(\omega{}) \, \text{ is continuous on } \, T \} \Big){} = 1. $

<b></b>

> ## **캐드랙 성질(càdlàg;right-continuous with left limits)** 

#### **(1) 정의** : 거의 모든 표본 경로 $ X_{t}(\omega{}) $가 모든 시점에서 우연속이고 좌극한이 존재하는 성질
#### $ = $ 표본 경로가 우측에선 연속이고 좌측에선 점프(jump) 직전 값이 항상 존재하는 형태의 경로를 갖는 성질
#### $ \Rrightarrow{} \displaystyle{} \mathbb{P}{}\Big({}\big\{{} \omega{} \in{} \Omega{} \mid{} \lim_{s \downarrow{} t} X_{s}(\omega{}) = X_{t}(\omega{}), \;\; \lim_{s \uparrow{} t} X_{s}(\omega{}) =: X_{t-}(\omega{}) \big\}{} \Big){}= 1 \;\;\;\; \forall{} t \in{} T $

<b></b>

> ## **가우시안 성질(Gaussian Property)**

#### **(1) 정의** : 모든 유한차원 분포가 다변량 정규분포가 됨을 만족하는 성질
#### $ = $ 임의의 유한 개 시점을 선택해도 항상 가우시안 랜덤 벡터로 정의되는 성질
#### $ \Rrightarrow{} \begin{bmatrix} X_{t_{1}} & \cdots{} & X_{t_{n}} \end{bmatrix}^{T} \sim \mathcal{N}{}\big( m_{X}(t_{1:n}), C_{X}(t_{1:n},t_{1:n}) \big){}, \;\;\;\; \forall{} t_{i} \in{} T $

<b></b>

> ## **마팅게일 성질(Martingale Property)**

#### **(1) 정의** : 필터레이션 $ \{ \mathcal{F}_{t} \} $에 대해, 과거 정보 $ \mathcal{F}_{s} $ 에 대한 미래 값의 조건부 기댓값이 현재 값과 일치함을 만족하는 성질
#### $ = $ 과거까지의 정보로 봤을 때 앞으로의 기대 이득이 $ 0 $인 마팅게일 구조를 갖는 성질
#### $ \Rrightarrow{} \mathbb{E}{}\big[{} X_{t} | \mathcal{F}_{s} \big]{} = X_{s} $
#### $ \hspace{0.45cm} \text{s.t. } \, \mathbb{E} \big[{} |X_{t}| \big]{} < \infty{}, \;\; \forall{} s < t $

<b></b>

> ## **마르코프 성질(Markov Property)**

#### **(1) 정의** : 미래 분포가 과거 전체가 아니라 현재 상태에만 의존하는 성질(memoryless)
#### $ = $ 현재 상태가 과거 정보를 모두 요약하여, 현재 상태만 알면 미래의 조건부 분포를 결정할 수 있는 마르코프 구조를 갖는 성질
#### $ \Rrightarrow{} \mathbb{P}{}\big( X_{t_{n+1}} \in{} A \mid X_{t_{0}},\dots{},X_{t_{n}} \big){} = \mathbb{P}{}\big( X_{t_{n+1}} \in{} A \mid X_{t_{n}} \big){} $
#### $ \hspace{0.45cm} \text{where } \, t_{0} < \dots{} < t_{n+1}, \;\; \forall{} A \in{} \mathcal{B}(S)  $
