# CH05.1. **$ L^{2} $ space**

> ## **$ L^{2} $ 공간($ L^{2} $ space)**

#### **(1) 정의** : $ 2 $차 모멘트가 유한한 랜덤 벡터들의 집합
#### $ \Rrightarrow{} L^{2}(\Omega{};\mathbb{R}^{m}) := \{ \textbf{X} : \Omega{} \to \mathbb{R}^{m} \;|\; \mathbb{E}[\|\textbf{X}\|^{2}_{2}] < \infty{} \} $

##### **(`PLUS`)** $ L^{2} $ 공간에는 고정(상수) 벡터도 포함함

<b></b>

> ## (Theorem 5-1) **Vector space property of $ \, L^{2} $**

#### $ L^{2} $ 공간은 벡터 공간(vector space)이다.

##### **(`WHY?`)** 
##### $ \hspace{0.15cm} \text{define } \, \alpha{} \in{} \mathbb{R}, \;\; \textbf{X}, \textbf{Y} \in{} L^{2}(\Omega{};\mathbb{R}^{m}). $
##### $ \hspace{0.15cm} \cdot{} \, \text{zero vector existence} $
##### $ \hspace{0.45cm} \text{if } \, \textbf{0} : (\Omega{}, \mathcal{F}) \rightarrow{} \big({} \mathbb{R}^{m}, \mathcal{B}(\mathbb{R}^{m}) \big){}, \;\; \omega{} \mapsto{} \begin{bmatrix} 0 & 0 & \cdots{} & 0 \end{bmatrix}^{T} $
##### $ \hspace{0.45cm} \mathbb{E}[\|\textbf{0}\|^{2}_{2}] = \displaystyle{} \sum^{m}_{i=1} |0|^{2} = 0 < \infty{} $
##### $ \hspace{0.45cm} \Rrightarrow{} \textbf{0} \in{} L^{2}(\Omega{};\mathbb{R}^{m}) $
##### $ \hspace{0.15cm} \cdot{} \, \text{closed under (vector) addition} $
##### $ \hspace{0.45cm} \mathbb{E}[\|\textbf{X}+\textbf{Y}\|^{2}_{2}] = \displaystyle{} \mathbb{E}\big[{} \sum^{m}_{i=1}|X_{i}+Y_{i}|^{2} \big]{} = \sum^{m}_{i=1} \mathbb{E}\big[{}|X_{i}+Y_{i}|^{2} \big]{} \;\; \cdots{} \; ⓐ $ 
##### $ \hspace{0.45cm} ⓐ \leq{} \displaystyle{} 2 \mathbb{E} \bigg[{} \sum^{m}_{i=1}(|X_{i}|^{2}+|Y_{i}|^{2}) \bigg]{} \;\; \because{} |a+b|^{2} \leq{} 2(|a|^{2}+|b|^{2})$
##### $ \hspace{0.83cm} = 2 \big({} \mathbb{E}[\|\textbf{X}\|^{2}_{2}] + \mathbb{E}[\|\textbf{Y}\|^{2}_{2}] \big){} < \infty{} $
##### $ \hspace{0.45cm} \Rrightarrow{} \textbf{X}+\textbf{Y} \in{} L^{2}(\Omega{};\mathbb{R}^{m}) $
##### $ \hspace{0.15cm} \cdot{} \, \text{closed under (scalar) multiplication} $
##### $ \hspace{0.45cm} \mathbb{E}[\|\alpha{}\textbf{X}\|^{2}_{2}] = \mathbb{E}\big[ \sum^{m}_{i=1}|\alpha{}X_{i}|^{2} \big] = |\alpha{}|^{2}\mathbb{E}\big[ \sum^{m}_{i=1}|X_{i}|^{2} \big] = | \alpha{}|^{2} \mathbb{E}[\|\textbf{X}\|^{2}_{2}] < \infty{} $
##### $ \hspace{0.45cm} \Rrightarrow{} \alpha{}\textbf{X} \in{} L^{2}(\Omega{};\mathbb{R}^{m}) $
##### $ \hspace{0.15cm} \text{(Other axioms omitted)} $
##### $ \hspace{0.45cm} \therefore{} L^{2}(\Omega{};\mathbb{R}^{m}) \, \text{ is vector space}.  $

<b></b>

> ## **$ L^{2} $ 노름($ L^{2} $ norm)**

#### **(1) 정의** : 어떠한 랜덤 벡터의 제곱근 기댓값 및 제곱 유클리드 노름 연산
#### $ \Rrightarrow{} \|\textbf{X}\|_{L^{2}} := \big({}\mathbb{E}[ \|\textbf{X}\|^{2}_{2} ] \big){}^{\frac{1}{2}} = \big({}\mathbb{E}[ \textbf{X}^{T}\textbf{X} ] \big){}^{\frac{1}{2}} = \big({}\mathbb{E}[ \sum_{i=1}^{m} |X_{i}|^{2}] \big){}^{\frac{1}{2}} $
#### $ \hspace{0.45cm} \text{where } \, \|\textbf{x}\|_{2} := \big({} \sum_{i=1}^{m} |x_{i}|^{2} \big){}^{\frac{1}{2}}  \;\; (\text{euclidean L2 norm})$

#### **(2) 성질** : **[~]**
##### $ \hspace{0.15cm} $ ① $ \|\textbf{X}\|_{L^{2}} = 0 \; \Leftrightarrow{} \; \textbf{X} = 0 \;\; \text{a.s.(almost surely)} $ 

##### **(`PLUS`) a.s.(almost surely)** : 확률이 $ 1 $인 집합에서만 성질이 성립한다는 의미로 정의됨

##### $ \hspace{0.15cm} $ ② $ \|\alpha{} \textbf{X}\|_{L^{2}} = |\alpha{}| \|\textbf{X}\|_{L^{2}} $
##### $ \hspace{0.15cm} $ ③ $ \|\textbf{X} + \textbf{Y} \|_{L^{2}} \leq{} \|\textbf{X}\|_{L^{2}} + \|\textbf{Y}\|_{L^{2}} $

<b></b>

> ## (Theorem 5-2) **Inner Product of $ L^{2} $ Space**

#### $ L^{2} $ 공간에서 내적을 아래와 같이 정의할 수 있다.
#### $ \Rrightarrow{} \langle{} \textbf{X}, \textbf{Y} \rangle{} := \mathbb{E}[\textbf{X}^{T}\textbf{Y}]  $

##### **(`WHY?`)** 
##### $ \hspace{0.15cm} \text{define } \, \alpha{},\beta{} \in{} \mathbb{R}, \;\; \textbf{X},\textbf{Y},\textbf{Z} \in{} L^{2}(\Omega{}; \mathbb{R}^{m}). $
##### $ \hspace{0.15cm} \cdot{} \, \text{finiteness} $
##### $ \hspace{0.45cm} \Rrightarrow{} \mathbb{E}[|\textbf{X}^{T}\textbf{Y}|] \le \mathbb{E}[\|\textbf{X}\|_{2}\|\textbf{Y}\|_{2}] \le{} \sqrt{\mathbb{E}[\|\textbf{X}\|^{2}_{2}] \, \mathbb{E}[\|\textbf{Y}\|^{2}_{2}]} < \infty{} \;\; \because{} \text{Cauchy–Schwarz inequality} $
##### $ \hspace{0.15cm} \cdot{} \, \text{linearity} $
##### $ \hspace{0.45cm} \Rrightarrow{} \langle{} \alpha{} \textbf{X} + \beta{} \textbf{Y}, \textbf{Z} \rangle{} = \mathbb{E}\big[(\alpha{} \textbf{X} + \beta{} \textbf{Y})^{T}\textbf{Z}\big] = \alpha{} \mathbb{E}[\textbf{X}^{T}\textbf{Z}] + \beta{} \mathbb{E}[\textbf{Y}^{T}\textbf{Z}] = \alpha{} \langle{} \textbf{X},\textbf{Z} \rangle{} + \beta{} \langle{} \textbf{Y},\textbf{Z} \rangle{} $
##### $ \hspace{0.15cm} \cdot{} \, \text{symmetry} $
##### $ \hspace{0.45cm} \Rrightarrow{} \langle{} \textbf{X},\textbf{Y} \rangle{} = \mathbb{E}[\textbf{X}^{T}\textbf{Y}] = \mathbb{E}\big[\sum_{i} X_{i}Y_{i}\big] = \mathbb{E}[\textbf{Y}^{T}\textbf{X}] = \langle{} \textbf{Y},\textbf{X} \rangle{} $
##### $ \hspace{0.15cm} \cdot{} \, \text{positive-definiteness} $
##### $ \hspace{0.45cm} \Rrightarrow{} \langle{} \textbf{X},\textbf{X} \rangle{} = \mathbb{E}[\|\textbf{X}\|^{2}_{2}] \ge 0, \;\; \langle{} \textbf{X},\textbf{X} \rangle{} = 0 \Rightarrow{} \textbf{X} = \textbf{0} \;\; \text{a.s.} $
##### $ \hspace{0.15cm} \therefore{} \langle{} \cdot{}, \cdot{} \rangle{} \, \text{ is the inner product on } \, L^{2}(\Omega{}; \mathbb{R}^{m}). $

##### **(`PLUS`)** 위와 같이 내적을 정의하면 $ L^{2}(\Omega{}; \mathbb{R}^{m}) $ 공간을 유클리드 공간처럼 직교, 투영 등의 개념을 동일한 방식으로 정의하고 사용할 수 있음

> ## **$ L^{2} $ 공간의 닫힌 부분공간(Closed subspace in $ L^{2} $ space)**

#### **(1) 정의** : $ L^{2} $ 노름에 대해 닫혀있는 (선형) 부분공간
#### $ \Rrightarrow{} \mathcal{V} := (\Gamma{}, +, \cdot{}, \|\cdot{}\|_{L^{2}}) $
#### $ \hspace{0.45cm} \text{ where } \, \Gamma{} \, \text{ is a subspace of } \, L^{2}(\Omega{};\mathbb{R}^{m}) $

#### **(2) 성질** : 스칼라 $ a_{1}, a_{2} $와 랜덤 벡터 $ \textbf{V}_{i} $가 존재할 때 아래가 성립함
##### $ \hspace{0.15cm} $ ① $ \textbf{V}_{1}, \textbf{V}_{2} \in{} \mathcal{V} \; \Rightarrow{} \; a_{1}\textbf{V}_{1} + a_{2}\textbf{V}_{2} \in{} \mathcal{V} \hspace{1.55cm} \cdots{} \, \text{linearity} $
##### $ \hspace{0.15cm} $ ② $ \displaystyle{} \forall{} \, \{ \textbf{V}_{n} \} \subset{} \mathcal{V}, \;\; \exists{} \textbf{V} \in{} L^{2}(\Omega{};\mathbb{R}^{m}), \;\;\|\textbf{V}_{n} - \textbf{V}\|_{L^{2}} \rightarrow{} 0 \; \Rightarrow{} \;\textbf{V} \in{} \mathcal{V} \;\; \cdots{} \, \text{closedness} $
