# CH01.2. **Function**

> ## **함수(Function)**

#### **(1) 정의** : 특정 집합에서 다른 집합으로의 대응 관계를 나타내는 규칙
#### $ \Rrightarrow{} f : A \rightarrow{} B, \;\; a \mapsto{} f(a) \;\; \text{ where } \, A, B \, \text{ are sets}, \;\; a \in{} A $

#### **(2) 형식** : 
##### $ \hspace{0.15cm} $ <img src="../img/03.1. Random Variables (1).png" width="40%" height="40%"></img>
##### $ \hspace{0.15cm} $ ① 정의역(domain) : 함수에 입력할 수 있는 모든 원소(input)들의 집합 $ A $
##### $ \hspace{0.15cm} $ ② 공역(codomain) : 함수값이 속할 수 있는 가능한 출력값들의 집합 $ B $ 
##### $ \hspace{0.15cm} $ ③ 치역(range) : 정의역 내 입력값이 함수를 통해 대응되는 출력값들의 집합 $ f(A) := \{ f(a) \mid{} a \in{} A \} \subseteq{} B $

> ## **함수의 종류(Classification of Function)**

#### **(1) 전사(Onto;Surjective)** : 치역(range)와 공역(codomain)이 동일한 관계
#### $ \Rrightarrow{} f(A) = B $   

#### **(2) 단사(One-to-one;Injective)** : 다른 입력(원소)은 항상 다른 출력을 갖는 관계
#### $ \Rrightarrow{} f(a_{1}) \neq{} f(a_{2}) \;\; \text{ where } \, \forall{} \, a_{1} \neq{} a_{2}, \;\; a_{1}, a_{2} \in{} A $

#### **(3) 전단사(Bijective)** : 전사이면서 단사인 관계
#### $ \Rrightarrow{} f(A) = B, \; \; f(a_{1}) \neq{} f(a_{2}) $
