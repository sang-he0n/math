# CH01.2. **Measure**

> ## **측도론(Measure Theory)**

#### **(1) 정의** : 집합에 길이, 넓이, 부피, 확률 등의 크기 개념을 적용 및 일반화하는 이론

##### **(`PLUS`)** 확률론은 사건을 집합으로 정의하고 측도론을 통해 집합 간의 연산을 통해 연산을 정의함

<b></b>

> ## **시그마-대수($\sigma{}$-Algebra)**

#### **(1) 정의** : 어떠한 (바탕) 집합 $ U $의 아래 공리를 만족하는 집합족
#### $ \Rrightarrow{} \mathcal{F} \subseteq{} 2^{U} $

#### **(2) 공리** : 어떠한 집합 $ U $에 대응되는 시그마-대수 $ \mathcal{F} $가 존재할 때 아래를 만족해야함
##### $ \hspace{0.15cm} $ ① $ \empty{} \in{} \mathcal{F} $ 
##### $ \hspace{0.15cm} $ ② $ A \in{} \mathcal{F} \; \Rightarrow{} \; A^{c} \in{} \mathcal{F} \;\; \text{ where } \, A \subset{} U $
##### $ \hspace{0.15cm} $ ③ $ A_{i} \in{} \mathcal{F} \; \Rightarrow{} \; \bigcup^{\infty{}}_{i=1} A_{i} \in{} \mathcal{F} \;\; \text{ where } \, A_{i} \subset{} U $

#### **(3) 성질** : 어떠한 집합 $ U $에 대응되는 시그마-대수 $ \mathcal{F} $가 존재할 때 아래를 만족함
##### $ \hspace{0.15cm} $ ① $ U \in{} \mathcal{F} $
##### $ \hspace{0.15cm} $ ② $ A_{i} \in{} \mathcal{F} \; \Rightarrow{} \; \displaystyle{} \cap^{\infty{}}_{i=1} A_{i} \in{} \mathcal{F} \;\; \text{ where } \, A_{i} \subset{} U $
##### $ \hspace{0.15cm} $ ③ $ A, B \in{} \mathcal{F} \; \Rightarrow{} \; \displaystyle{} (A \cup{} B) \in{} \mathcal{F} $
##### $ \hspace{0.15cm} $ ④ $ A, B \in{} \mathcal{F} \; \Rightarrow{} \; \displaystyle{} (A \cap{} B^{c}) \in{} \mathcal{F} $

#### **(4) 종류** :
##### $ \hspace{0.15cm} $ ① 자명한 시그마-대수(Trivial $ \sigma{} $-Algebra) :
##### $ \hspace{0.45cm} \Rrightarrow{} \{ \empty{}, U \} $
##### $ \hspace{0.15cm} $ ② 생성 시그마-대수(Generated $ \sigma{} $-Algebra) : 집합족 $ \mathcal{C} \subseteq{} 2^{U} $가 생성하는 가장 작은 시그마-대수

##### **(`PLUS`) 생성자(Generator)** : 주어진 집합족으로부터 가장 작은 시그마-대수를 만들어내는 연산
##### $ \hspace{0.15cm} \Rrightarrow{} \sigma{}(\mathcal{C}) := \bigcap{}\{\mathcal{G}\subseteq{} 2^{U} \; : \; \mathcal{G} \, \text{ is a }\sigma\text{-algebra on } U, \;\; \mathcal{C} \subseteq{} \mathcal{G} \} $

##### $ \hspace{0.15cm} $ ③ 보렐 시그마-대수(Borel $ \sigma{} $-Algebra) : 열린 집합을 생성하는 가장 작은 시그마-대수
##### $ \hspace{0.45cm} \Rrightarrow{} \mathcal{B}(\mathbb{R}) := \sigma{}(\{(a, b)  \; : \; a<b, \;\;  a, b \in{} \mathbb{Q} \}) \;\; \text{ where } \, \sigma{}(\cdot{}) : \text{ generation operator} $ 

<b></b>

> ## **가측 공간(Measurable Space)**

#### **(1) 정의** : 어떠한 집합과 그 집합 위의 시그마-대수으로 구성된 객체(튜플)
#### $ \Rrightarrow{} (U, \mathcal{F}) $ 

<b></b>

> ## **함수의 가측성(Measurability of Function)**

#### **(1) 정의** : 두 가측 공간을 매핑하는 함수 $ f $가 존재할 때, $ \, f $의 역상(preimage)이 정의역(domain)의 시그마-대수 안에 포함되는 성질
#### $ \Rrightarrow{} f : (U, \mathcal{F}) \rightarrow{} (W, \mathcal{G}), \;\; f \, \text{measurable} \;\; \Leftrightarrow{} \;\; f^{-1}(G) \in{} \mathcal{F} \;\;\;\; \forall{} G \in{} \mathcal{G} $
#### $ \hspace{0.45cm} \text{where } \, \mathcal{F}, \mathcal{G} \, \text{ are } \, \sigma{}\text{-algebra}, \;\; U, W \, \text{ are set} $

<b></b>

> ## **측도(Measure)**

#### **(1) 정의** : 어떠한 가측 공간에서 시그마-대수의 원소를 사용하여 비음수 실수 스칼라를 반환하는 집합 함수(set function)
#### $ = $ 집합의 크기를 산정하는 함수
#### $ \Rrightarrow{} \mu{} : \mathcal{F} \rightarrow{} [0, \infty{} ] $

#### **(2) 공리** : 어떠한 시그마-대수 $ \mathcal{F} $에 대응되는 측도 $ \mu{} $가 존재할 때 아래를 만족해야함
##### $ \hspace{0.15cm} $ ① $ \mu{}(A) \geq{} 0 \;\;\;\; \forall{} \, A \in{} \mathcal{F} $
##### $ \hspace{0.15cm} $ ② $ \mu{}(\empty{}) = 0 $ 
##### $ \hspace{0.15cm} $ ③ $ \displaystyle{} A_{i} \cap{} A_{j} = \empty{} \; \Rightarrow{} \; \mu{}(\cup^{\infty{}}_{i=1}A_{i}) = \sum^{\infty{}}_{i=1} \mu{}(A_{i}) \;\; \text{ where } \, \forall{} \, i \neq{} j, \;\; A_{i} \in{} \mathcal{F} $ 

#### **(3) 성질** : 어떠한 집합 $ U $의 부분집합 $ A, B $와 측도 $ \mu{} $가 존재할 때 아래를 만족함
##### $ \hspace{0.15cm} $ ① $ A \subseteq{} B \; \Rightarrow{} \; \mu{}(A)\le{} \mu{}(B) $
##### $ \hspace{0.15cm} $ ② $ \mu{}\left({} \bigcup_{i=1}^\infty{} A_{i}\right) \le{} \sum_{i=1}^\infty{} \mu{}(A_{i}) \;\; \text{ s.t. } \, A_{i} \cap{} A_{j} = \empty{} $
##### $ \hspace{0.15cm} $ ③ $ A_{n} \uparrow{} A \; \Rightarrow{} \; \mu{}(A_{n})\uparrow{} \mu{}(A) $
##### $ \hspace{0.15cm} $ ④ $ A_{n} \downarrow{} A \; \Rightarrow{} \; \mu{}(A_{n}) \downarrow{} \mu{}(A) \;\; \text{ s.t. } \, \mu{}(A_1)<\infty{} $

#### **(4) 종류** :
##### $ \hspace{0.15cm} $ ① 계수 측도(Counting measure)
##### $ \hspace{0.45cm} \Rrightarrow{} \mu{}(A) = \begin{cases} |A| & \text{if } \, A \, \text{ is finite set} \\ \infty{} & \text{if } \, A \, \text{ is infinite set} \end{cases} $
##### $ \hspace{0.15cm} $ ② 보렐 길이 측도(Borel Length measure) : 실수들의 집합의 길이를 일반화한 측도
##### $ \hspace{0.45cm} \Rrightarrow{} \lambda{}([a,b]) = b - a \;\; \text{ where } \, \lambda{} : \mathcal{B}(\mathbb{R}) \rightarrow{} [0, \infty{}] $
##### $ \hspace{0.15cm} $ ③ 확률 측도(Probability measure)
##### $ \hspace{0.45cm} \Rrightarrow{} \mathbb{P}(U) = 1 $

##### **(`PLUS`) 영집합(Empty set)** : 측도가 $ 0 $인 집합 
##### $ \hspace{0.15cm} \Rrightarrow{} \mu{}(A) = 0 $

<b></b>

> ## **측도 공간(Measure space)**

#### **(1) 정의** : 어떠한 집합과 그 집합과 대응되는 시그마-대수, 측도와 함께 구성된 객체
#### $ \Rrightarrow{} (U, \mathcal{F}, \mu{}) $ 

<b></b>

> ## **측도 공간의 완비성(Completeness of Measure space)**

#### **(1) 정의** : 측도 공간 $ (U,\mathcal{F},\mu{}) $에서 모든 영집합의 부분집합까지 시그마-대수 $ \mathcal{F} $가 포함하는 성질
#### $ \Rrightarrow (U,\mathcal{F},\mu{}) \, \text{ complete} $
#### $ \hspace{0.45cm} \text{s.t. } \, \forall{} N \in{} \mathcal{F}(\mu{}(N)=0), \;\; \forall{} A \subseteq{}  N\ \Rightarrow\ A\in\mathcal{F}.$
