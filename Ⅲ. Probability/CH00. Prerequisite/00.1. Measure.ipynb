> ## **시그마-대수($\sigma{}$-Algebra)**

#### **(1) 정의** : 집합 $ U $에서 여집합 및 (가산) 합집합 연산에 닫혀있고 공집합을 포함하는 집합족 $ \mathcal{F} $
#### $ \Rrightarrow{} \mathcal{F} \subseteq{} 2^{U} $

#### **(2) 공리** : 어떠한 집합 $ U $에 대응되는 시그마-대수 $ \mathcal{F} $가 존재할 때 아래를 만족해야함
##### $ \hspace{0.15cm} $ ① $ \empty{} \in{} \mathcal{F} $ 
##### $ \hspace{0.15cm} $ ② $ A \in{} \mathcal{F} \; \Rightarrow{} \; A^{c} \in{} \mathcal{F} \;\; \text{ where } \, A^{c} := U \setminus{} A \in{} \mathcal{F} $ 
##### $ \hspace{0.15cm} $ ③ $ \displaystyle{} A_{i} \in{} \mathcal{F} \; (\forall{}\,i) \; \Rightarrow{} \; \bigcup^{\infty{}}_{i=1} A_{i} \in{} \mathcal{F} $

#### **(3) 성질** : 어떠한 집합 $ U $에 대응되는 시그마-대수 $ \mathcal{F} $가 존재할 때 아래를 만족함
##### $ \hspace{0.15cm} $ ① $ U \in{} \mathcal{F} $
##### $ \hspace{0.15cm} $ ② $ \displaystyle{} A_{i} \in{} \mathcal{F} \; \Rightarrow{} \; \bigcap^{\infty{}}_{i=1} A_{i} \in{} \mathcal{F} \;\; \text{ where } \, A_{i} \subset{} U $
##### $ \hspace{0.15cm} $ ③ $ A, B \in{} \mathcal{F} \; \Rightarrow{} \; (A \cup{} B) \in{} \mathcal{F}, \;\; (A \cap{} B) \in{} \mathcal{F}, \;\;  (A \cap{} B^{c}) \in{} \mathcal{F} $ 

#### **(4) 종류** :
##### $ \hspace{0.15cm} $ ① 자명한 시그마-대수(Trivial $ \sigma{} $-Algebra)
##### $ \hspace{0.45cm} \Rrightarrow{} \{ \empty{}, U \} $
##### $ \hspace{0.15cm} $ ② 생성 시그마-대수(Generated $ \sigma{} $-Algebra) : 집합족 $ \mathcal{C} \subseteq{} 2^{U} $가 생성하는 가장 작은 시그마-대수
##### $ \hspace{0.45cm} \Rrightarrow{} \displaystyle{} \sigma{}(\mathcal{C}) := \bigcap_{i}\{\mathcal{G}_{i} \subseteq{} 2^{U} \; : \; \mathcal{G}_{i} \, \text{ is a }\sigma\text{-algebra on } \, U, \;\; \mathcal{C} \subseteq{} \mathcal{G}_{i} \} $
##### $ \hspace{0.15cm} $ ③ 보렐 시그마-대수(Borel $ \sigma{} $-Algebra) : 어떠한 열린 집합들을 포함하는 가장 작은 시그마-대수
##### $ \hspace{0.45cm} \Rrightarrow{} \mathcal{B}(\mathbb{R}) := \sigma{}(\{(a, b)  \; : \; a<b, \;\;  a, b \in{} \mathbb{Q} \}) \;\; \text{ where } \, \sigma{}(\cdot{}) : \text{ generation operator} $

> ## **가측 공간(Measurable Space)**

#### **(1) 정의** : 어떠한 집합과 그 집합 위의 시그마-대수으로 구성된 객체
#### $ \Rrightarrow{} (U, \mathcal{F}) \, \text{ is a measurable space } $

<b></b>

> ## **함수의 가측성(Measurability of Function)**

#### **(1) 정의** : 두 가측 공간을 매핑하는 함수 $ f $가 존재할 때, 그 역상(preimage) $ f^{-1}(G) $이 정의역(domain)의 시그마-대수 안에 포함되는 성질
#### $ \Rrightarrow{} f \, \text{ is measurable} \; \Leftrightarrow{} \; f^{-1}(G) \in{} \mathcal{F} \;\;\;\; \forall{} G \in{} \mathcal{G} $
#### $ \hspace{0.45cm} \text{where } \, f : (U, \mathcal{F}) \rightarrow{} (W, \mathcal{G}), \;\; \mathcal{F}, \mathcal{G} \, \text{ are } \, \sigma{}\text{-algebra}, \;\; U, W \, \text{ are set} $

<b></b>

> ## **측도(Measure)**

#### **(1) 정의** : 어떠한 가측 공간에서 시그마-대수의 원소를 사용하여 비음수 실수값을 반환하는 집합 함수(set function)
#### $ = $ 집합의 크기를 산정하는 함수
#### $ \Rrightarrow{} \mu{} : \mathcal{F} \rightarrow{} [0, \infty{} ] $

#### **(2) 공리** : 어떠한 시그마-대수 $ \mathcal{F} $에 대응되는 측도 $ \mu{} $가 존재할 때 아래를 만족해야함
##### $ \hspace{0.15cm} $ ① $ \mu{}(A) \geq{} 0 \;\;\;\; \forall{} \, A \in{} \mathcal{F} $
##### $ \hspace{0.15cm} $ ② $ \mu{}(\empty{}) = 0 $ 
##### $ \hspace{0.15cm} $ ③ $ \displaystyle{} A_{i} \cap{} A_{j} = \empty{} \;\; (\forall{} \, i \neq{} j) \; \Rightarrow{} \; \mu{}(\bigcup^{\infty{}}_{i=1}A_{i}) = \sum^{\infty{}}_{i=1} \mu{}(A_{i}) \;\; \text{ where } \, A_{i} \in{} \mathcal{F} $ 

#### **(3) 성질** : 어떠한 집합 $ U $와 측도 $ \mu{} $가 존재할 때 아래를 만족함
##### $ \hspace{0.15cm} $ ① $ A \subseteq{} B \; \Rightarrow{} \; \mu{}(A)\le{} \mu{}(B) \;\; \text{ where } \, A, B \in{} \mathcal{F} $
##### $ \hspace{0.15cm} $ ② $ \displaystyle{} \mu{}\left({} \bigcup_{i=1}^\infty{} A_{i}\right) \le{} \sum_{i=1}^\infty{} \mu{}(A_{i}) $

#### **(4) 종류** :
##### $ \hspace{0.15cm} $ ① 계수 측도(Counting measure)
##### $ \hspace{0.45cm} \Rrightarrow{} \mu{}(A) = \begin{cases} |A| & \text{if } \, A \, \text{ is finite set} \\ \infty{} & \text{if } \, A \, \text{ is infinite set} \end{cases} $
##### $ \hspace{0.15cm} $ ② 보렐 길이 측도(Borel Length measure) : 실수들의 집합의 길이를 일반화한 측도
##### $ \hspace{0.45cm} \Rrightarrow{} \lambda{}([a,b]) = b - a \;\; \text{ where } \, \lambda{} : \mathcal{B}(\mathbb{R}) \rightarrow{} [0, \infty{}] $
##### $ \hspace{0.15cm} $ ③ 확률 측도(Probability measure)
##### $ \hspace{0.45cm} \Rrightarrow{} \mathbb{P}(\Omega{}) = 1 $

##### **(`PLUS`) 영측도 집합(measure-zero set)** : 측도가 $ 0 $인 집합 
##### $ \hspace{0.15cm} \Rrightarrow{} \mu{}(A) = 0 $

> ## (Theorem 1-6) **Continuity of Measure**

#### **(1)** 어떠한 집합열 $ A_{1}\subseteq{}A_{2}\subseteq{}\cdots{} $이 존재할 때 그 합집합의 측도는 각 집합의 측도의 극한과 같다.
#### $ \Rrightarrow{} \displaystyle{} \mu{}\Big({}\bigcup_{n=1}^{\infty{}}A_{n}\Big) = \lim_{n\to{}\infty{}}\mu{}(A_{n}) \;\; \text{ s.t. } \, A_{1}\subseteq{}A_{2}\subseteq{}\cdots{}, \;\; A_{n}\in{}\mathcal{F} $

#### **(2)** 어떠한 집합열 $ A_{1}\supseteq{}A_{2}\supseteq{}\cdots{} $이 존재하고 $ \mu{}(A_{1})<\infty{} $일 때 그 교집합의 측도는 각 집합의 측도의 극한과 같다.
#### $ \Rrightarrow{} \displaystyle{} \mu{}\Big({}\bigcap_{n=1}^{\infty{}}A_{n}\Big) = \lim_{n\to{}\infty{}}\mu{}(A_{n}) \;\; \text{ s.t. } \, A_{1}\supseteq{}A_{2}\supseteq{}\cdots{}, \;\; A_{n}\in{}\mathcal{F}, \;\; \mu{}(A_{1})<\infty{} $
