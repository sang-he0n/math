# CH02.2. **Bayesian View of Machine Learning**

> ## **베이즈 법칙(Bayes Rule)**

#### **(1) 정의** : 우도와 사전 분포를 통해 관측 데이터 이후의 파라미터에 대한 사후분포를 구성하는 규칙
#### $ \Rrightarrow{} \displaystyle{} p(\theta{}|\mathcal{D}_{\mathrm{obs}}) = \frac{p(\mathcal{D}_{\mathrm{obs}}|\theta{})p(\theta{})}{p(\mathcal{D}_{\mathrm{obs}})} = \frac{p(\mathcal{D}_{\mathrm{obs}}|h_{\theta{}})p(\theta{})}{p(\mathcal{D}_{\mathrm{obs}})} $
##### $ \hspace{0.6cm} \text{where } \, \theta{} \in{} \Theta{}, \;\; h_{\theta{}} \in{} \mathcal{H} := \{ h_{\theta{}} \mid{} \theta{} \in{} \Theta{} \} $

##### **(`NOTE`)** 여기서 정의된 확률 $ p(\cdot{}) $는 문맥에 따라 확률분포의 밀도 표현(밀도,질량함수 등) 혹은 우도(likelihood) 값을 나타내는 계산용 표기로 정의함

#### **(2) 구성 요소** :
##### $ \hspace{0.15cm} $ ① 우도(Likelihood) : 파라미터 $ \theta{} $ 혹은 모델 $ h_{\theta{}} $가 주어졌을 때, 관측 데이터 $ \mathcal{D}_{\mathrm{obs}} $에 부여되는 확률 분포(의 밀도표현)
##### $ \hspace{0.45cm} \Rrightarrow{} p(\mathcal{D}_{\mathrm{obs}}|\theta{}) = p(\mathcal{D}_{\mathrm{obs}}|h_{\theta{}}) \approx{} \text{data-fit term} $
##### $ \hspace{0.15cm} $ ② 사전 분포(Prior) : 데이터 관측 이전에 파라미터 $ \theta{} $에 대해 갖는 불확실성 혹은 사전 선호를 표현하는 확률 분포
##### $ \hspace{0.45cm} \Rrightarrow{} p(\theta{}) \approx{} \text{distribution before observing data} $
##### $ \hspace{0.15cm} $ ③ 사후 분포(Posterior) : 데이터 $ \mathcal{D}_{\mathrm{obs}} $ 관측 이후 파라미터 $ \theta{} $에 대한 불확실성이 업데이트된 확률 분포
##### $ \hspace{0.45cm} \Rrightarrow{} p(\theta{}|\mathcal{D}_{\mathrm{obs}}) \approx{} \text{distribution after observing data} $
##### $ \hspace{0.15cm} $ ④ 증거(Evidence) : 모든 가능한 파라미터가 관측 데이터 $ \mathcal{D}_{\mathrm{obs}} $에 부여하는 우도를 사전 분포로 가중 평균한 값
##### $ \hspace{0.45cm} \Rrightarrow{} p(\mathcal{D}_{\mathrm{obs}}) = \int_{\Theta{}} p(\mathcal{D}_{\mathrm{obs}}|\theta{})p(\theta{})\mathrm{d}\theta{} = \int_{\Theta{}} p(\mathcal{D}_{\mathrm{obs}}|h_{\theta{}})p(\theta{})\mathrm{d}\theta{} \approx{} \text{normalization constant} $

> ## (Theorem 2-1) **Limitation of MLE from Bayesian View**

#### **(1)** 베이지안 관점에서 최대 우도 추정법(MLE)은 사전 분포 $ p(\theta{}) $를 고려하지 않고, 우도(likelihood)만을 최대화하는 파라미터 $ \theta{} $를 선택한다.
#### $ \Rrightarrow{} \displaystyle{} \theta{}_{\mathrm{MLE}} = \argmax_{\theta{} \in{} \Theta{}} p(\mathcal{D}_{\mathrm{obs}}|\theta{}) = \argmax_{\theta{} \in{} \Theta{}} p(\mathcal{D}_{\mathrm{obs}}|h_{\theta{}}) $
#### $ \hspace{1.175cm} \displaystyle{} = \argmax_{\theta{} \in{} \Theta{}} \bigg({} \text{likelihood}_{\mathrm{cond}}(\theta{}) \bigg){} = \argmax_{\theta{} \in{} \Theta{}} \bigg({} \prod_{i=1}^{N} p_{\theta{}}(y_i|x_i) \bigg){} $
#### $ \hspace{0.3cm} \text{where } \, \mathcal{D}_{\mathrm{obs}} = \{(x_i,y_i)\}_{i=1}^{N} $

#### **(2)** MLE의 한계는 동일한 최대 우도를 갖는 파라미터 후보가 여러 개 존재하더라도, 각 후보에 대한 사전 분포 $ p(\theta{}) $의 차이를 선택 기준에 반영하지 못한다는 점이다.
#### $ \Rrightarrow{} \displaystyle{} \text{assume } \, \theta{}_1 \neq{} \theta{}_2, \;\; p(\mathcal{D}_{\mathrm{obs}}|\theta{}_1) = p(\mathcal{D}_{\mathrm{obs}}|\theta{}_2) = \max_{\theta{} \in{} \Theta{}} p(\mathcal{D}_{\mathrm{obs}}|\theta{}), \;\; p(\theta{}_1) \neq{} p(\theta{}_2) $
#### $ \hspace{0.3cm} \therefore{} \displaystyle{} \theta{}_1 \text{ and } \theta{}_2 \text{ are indistinguishable under MLE criterion} $

> ## (Theorem 2-2) **Maximum A Posteriori Estimation**

#### **(1)** 베이지안 관점에서 최대 사후 추정(MAP)은 관측 데이터 $ \mathcal{D}_{\mathrm{obs}} $ 이후의 사후분포(posterior)를 최대화하는 파라미터 $ \theta{} $를 선택하는 과정이다.
#### $ \Rrightarrow{} \displaystyle{} \theta{}_{\mathrm{MAP}} = \argmax_{\theta{} \in{} \Theta{}} p(\theta{}|\mathcal{D}_{\mathrm{obs}}) = \argmax_{\theta{} \in{} \Theta{}} \frac{p(\mathcal{D}_{\mathrm{obs}}|\theta{})p(\theta{})}{p(\mathcal{D}_{\mathrm{obs}})} $
#### $ \hspace{0.3cm} \therefore{} \displaystyle{} h_{\theta{}_{\mathrm{MAP}}} \in{} \mathcal{H} \text{ is the selected model candidate} $

#### **(2)** MAP에서 evidence $ p(\mathcal{D}_{\mathrm{obs}}) $는 $ \theta{} $에 대해 상수이므로, MAP는 우도(likelihood)와 사전 분포(prior)를 함께 최대화하는 문제로 정리된다.
#### $ \displaystyle{} \text{since } \, p(\theta{}|\mathcal{D}_{\mathrm{obs}}) = \frac{p(\mathcal{D}_{\mathrm{obs}}|\theta{})p(\theta{})}{p(\mathcal{D}_{\mathrm{obs}})} \propto{} p(\mathcal{D}_{\mathrm{obs}}|\theta{})p(\theta{}) $
#### $ \Rrightarrow{} \displaystyle{} \theta{}_{\mathrm{MAP}} := \argmax_{\theta{} \in{} \Theta{}} p(\mathcal{D}_{\mathrm{obs}}|\theta{})p(\theta{}) = \argmax_{\theta{} \in{} \Theta{}} \left[ \log p(\mathcal{D}_{\mathrm{obs}}|\theta{}) + \log p(\theta{}) \right]{} $
#### $ \hspace{1.325cm} \displaystyle{} = \argmax_{\theta{} \in{} \Theta{}} \left[ \log p(\mathcal{D}_{\mathrm{obs}}|h_{\theta{}}) + \log p(\theta{}) \right]{} $
#### $ \hspace{1.325cm} \displaystyle{} = \argmax_{\theta{} \in{} \Theta{}} \left[ \sum_{i=1}^{N} \log p_{\theta{}}(y_i|x_i) + \log p(\theta{}) \right]{} $
#### $ \hspace{0.3cm} \text{where } \, \mathcal{D}_{\mathrm{obs}} = \{(x_i,y_i)\}_{i=1}^{N}, \;\; \displaystyle{} p(\mathcal{D}_{\mathrm{obs}}|h_{\theta{}}) = \prod_{i=1}^{N} p_{\theta{}}(y_i|x_i) $

#### **(3)** MAP는 NLL에 negative log-prior 항을 추가하여 목적 함수를 최소화하는 문제로 정의된다.
#### $ \Rrightarrow{} \displaystyle{} \theta{}_{\mathrm{MAP}} = \argmin_{\theta{} \in{} \Theta{}} \left[ \sum_{i=1}^{N} -\log p_{\theta{}}(y_i|x_i) - \log p(\theta{}) \right]{} = \argmin_{\theta{} \in{} \Theta{}} \left[{} \mathcal{L}_{\mathrm{NLL}}(\theta{}) + \Omega{}_{\mathrm{prior}}(\theta{}) \right]{} $
#### $ \hspace{0.3cm} \text{where } \, \displaystyle{} \Omega{}_{\mathrm{prior}}(\theta{}) := -\log p(\theta{}) $
