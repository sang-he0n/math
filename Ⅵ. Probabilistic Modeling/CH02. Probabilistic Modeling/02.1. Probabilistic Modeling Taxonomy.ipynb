# CH02.1. **Probabilistic Modeling Taxonomy**

> ## **판별 모델(Discriminative Model)**

#### **(1) 정의** : 입력 $ X $가 주어졌을 때 출력 $ Y $에 대한 참 조건부 분포(의 밀도 표현) $ \, p^{*}_{Y|X}(y|x) $를 모델이 직접 근사하는 확률론적 모델
#### $ = $ 관측된 입력 $ x $에 대해 어떤 출력 $ y $가 적절한 지를 학습하는 모델
#### $ \Rrightarrow{} p_{\theta{}}(y|x) \approx{} p^{*}_{Y|X}(y|x) $

#### **(2) 목적(비용) 함수** : 실제 학습 과정에서 관측 데이터 $ \mathcal{D}_{\mathrm{obs}}=\{(x_i,y_i)\}_{i=1}^{N}$에 대해 최적화하는 함수
#### $ \Rrightarrow{} \displaystyle{} J(\theta{}) := \mathcal{L}_{\text{NLL}}(\theta{}) + \Lambda{}(\theta{}) = \frac{1}{N}\sum^{N}_{i=1}-\log{}\big({}p_{\theta{}}(y_{i}|x_{i})\big){} + \Lambda{}(\theta{}) $
#### $ \hspace{0.15cm} \text{where } \, \Lambda{}(\theta{}) : \text{additional term} $

#### **(3) 추론 관점** : 모델의 학습 (조건부) 분포 $ p_{\theta{}}(y|x) $를 이용하여 주어진 입력 $ x $에 대한 최종 출력 $ \hat{y} $를 결정(decision)
#### $ = $ 모델의 학습 분포로부터 정의되는 결정 규칙(decision rule)을 이용해 출력(label;class;target)에 대해 예측
#### $ \Rrightarrow{} p_{\theta{}}(y|x) : \text{distribution} \to{} \text{decision}, \;\; \displaystyle{} \hat{y}= \delta{}_{\bar{\theta{}}}(x) = \argmax_{y}p_{\bar{\theta{}}}(y|x) $
#### $ \hspace{0.3cm} \text{where } \, \delta{}_{\bar{\theta{}}}(x) : \text{decision rule}, \;\; \bar{\theta{}} : \text{fixed parameter} $

##### **(`PLUS`) 결정 규칙(Decision Rule)** : 학습된 확률분포를 최종 예측값으로 변환하는 규칙

#### **(4) 예시** :
#### $ \hspace{0.15cm} $ ① Logistic Regression
#### $ \hspace{0.15cm} $ ② Softmax Classifier
#### $ \hspace{0.15cm} $ ③ BERT-based Classifier
#### $ \hspace{0.15cm} $ ④ Probabilistic Regression Model

> ## **무조건부 생성 모델(Unconditional Generative Model)**

#### **(1) 정의** : 관측된 데이터 $ x $에 대하여 참 확률 분포 $ p^{*}_{X}(x) $를 모델이 근사하는 모델
#### $ = $ 관측 데이터가 어떤 확률적 모델에 의해 생성되는 지를 학습하는 모델
#### $ \Rrightarrow{} p_{\theta{}}(x) \approx{} p^{*}_{X}(x) $

#### **(2) 목적(비용) 함수** :
#### $ \Rrightarrow{} \displaystyle{} J(\theta{}) := \hat{\mathcal{L}}_{\text{NLL}}(\theta{}) + \Lambda{}(\theta{}) = \frac{1}{N}\sum^{N}_{i=1}-\log{}\big({}p_{\theta{}}(x_{i})\big){} + \Lambda{}(\theta{}) $
#### $ \hspace{0.15cm} \text{where } \, \Lambda{}(\theta{}) : \text{additional term} $

#### **(3) 추론 관점** : 모델의 학습 분포 $p_{\theta{}}(x)$로부터 새로운 데이터 $ x $를 생성
#### $ = $ 별도의 조건 없이 원 데이터 분포 $ p^{*}_{X} $에서 관측될 법한 표본(sample)을 생성
#### $ \Rrightarrow{} p_{\theta{}}(x) : \text{distribution} \to{} \text{sample}, \;\; x_\text{gen} \sim{} p_{\theta{}}(X) $

#### **(3) 예시** :
#### $ \hspace{0.15cm} $ ① Gaussian Mixture Model
#### $ \hspace{0.15cm} $ ② Variational Autoencoder
#### $ \hspace{0.15cm} $ ③ Generative Adversarial Network
#### $ \hspace{0.15cm} $ ④ Diffusion Model
#### $ \hspace{0.15cm} $ ⑤ Autoregressive Language Model

<b></b>

> ## **조건부 생성 모델(Conditional Generative Model)**

#### **(1) 정의** : 조건 변수 $ C $가 주어졌을 때 데이터 $ X $의 조건부분포 $ p^{*}_{X|C}(x|c) $를 모델이 학습하는 생성 모델
#### $ = $ 주어진 조건 $ c $에 맞추어 새로운 데이터 $ x $를 생성하는 모델
#### $ \Rrightarrow{} p_{\theta{}}(x|c) \approx{} p^{*}_{X|C}(x|c) $

##### **(`PLUS`) 조건 변수(Condition Variable)** : 조건 $ C $는 데이터 생성을 제어하는 객체를 의미함

#### **(2) 목적(비용) 함수** :
#### $ \Rrightarrow{} \displaystyle{} J(\theta{}) := \hat{\mathcal{L}}_{\text{NLL}}(\theta{})+\Lambda{}(\theta{}) = \frac{1}{N}\sum_{i=1}^{N}-\log{}\big({}p_{\theta{}}(x_i|c_i)\big){}+\Lambda{}(\theta{}) $
#### $ \hspace{0.15cm} \text{where } \, \Lambda{}(\theta{}) : \text{additional term} $

#### **(3) 추론 관점** : 모델이 학습한 조건부 분포 $ p_{\theta{}}(x|c) $로부터 조건 $ c $에 부합하는 새로운 데이터 $ x $를 생성
#### $ \Rrightarrow{} p_{\theta{}}(x|c) : \text{distribution} \to{} \text{sample}, \;\; x_\text{gen} \sim{} p_{\theta{}}(X|C=c) $

#### **(3) 예시**
#### $ \hspace{0.3cm} $ ① Class-conditional Generation : $c=\text{class label}, \;\; x=\text{image}$
#### $ \hspace{0.3cm} $ ② Text-to-Image Diffusion with CFG : $c=\text{text prompt}, \;\; x=\text{image}$
#### $ \hspace{0.3cm} $ ③ Machine Translation : $c=\text{source sentence}, \;\; x=\text{target sentence}$
#### $ \hspace{0.3cm} $ ④ Instruction-tuned Language Model : $c=\text{instruction}, \;\; x=\text{response}$

##### **(`PLUS`) CFG(Classifier-Free Guidance)** : 조건부 생성에서 condition $c$의 반영 정도를 조절하는 sampling 기법
##### $ \hspace{0.875cm} $ CFG는 별도의 classifier를 사용하지 않고, conditional prediction과 unconditional prediction의 차이를 이용해 condition 방향으로 generation을 유도함
##### $ \hspace{0.3cm} \Rrightarrow{} \epsilon{}_{\mathrm{CFG}} = \epsilon{}_{\theta{}}(x_t,\varnothing{},t) + w\big({}\epsilon{}_{\theta{}}(x_t,c,t)-\epsilon{}_{\theta{}}(x_t,\varnothing{},t)\big){} $
##### $ \hspace{0.3cm} \text{where } \, c : \text{condition}, \;\; \varnothing{} : \text{null condition}, \;\; w : \text{guidance scale} $
