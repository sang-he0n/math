# CH01.1. **Machine Learning Overview**

> ## **머신 러닝(Machine Learning)**

#### **(1) 정의** : 
##### $ \hspace{0.15cm} $ ① 명시적 규칙을 사람이 직접 코딩하지 않고, 데이터를 통해 어떠한 작업을 성능 지표를 통해 개선하는 분야 (Tom Mitchell)
##### $ \hspace{0.15cm} $ ② 컴퓨터가 명시적으로 프로그래밍되지 않고도 학습할 수 있는 능력을 갖게 하는 연구 분야 (Arthur Samuel)

#### **(2) 구성요소** :
##### $ \hspace{0.15cm} $ ① 경험(Experience) : 모델이 학습에 사용하는 경험
##### $ \hspace{0.45cm} = $ 모델이 파라미터를 업데이트하기 위해 관측하는 데이터
##### $ \hspace{0.45cm} $ [`ex`] tabular data, time series, text, image, video, speech, graph, multimodal, $ \cdots{} $
##### $ \hspace{0.15cm} $ ② 임무(Task) : 모델이 수행해야 하는 문제
##### $ \hspace{0.45cm} $ [`ex`] classification, regression, detection, segmentation, translation, policy learning, $ \cdots{} $
##### $ \hspace{0.15cm} $ ③ 측정 지표(Performance measure) : 업무 수행 능력을 측정하는 지표
##### $ \hspace{0.45cm} $ [`ex`] accuracy, MSE, cross-entropy, NLL, FID, $ \cdots{} $

#### **(3) 학습 설정에 대한 분류** :
##### $ \hspace{0.15cm} $ ① 지도 학습(Supervised Learning)
##### $ \hspace{0.15cm} $ ② 비지도 학습(Unsupervised Learning)
##### $ \hspace{0.15cm} $ ③ 강화 학습(Reinforce Learning)

#### **(4) 확률 모델링 기준 분류** :
##### $ \hspace{0.15cm} $ ① 판별 모델(Discriminative model) 
##### $ \hspace{0.15cm} $ ② 생성 모델(Generative model)
