# CH02.2. **Learning Problem Formulation**

> ## **지도 학습(Supervised Learning)**

#### **(1) 경험(Experience)** : 
#### $ \text{assume } \, (X, Y) \sim{} P^{*}_{X,Y}, \;\; (X_{i}, Y_{i}) \overset{\text{i.i.d.}}{\sim} P^{*}_{X,Y}, \;\; i = 1, 2, \cdots{}, N $
#### $ \Rrightarrow{} \mathcal{D} = \{ (X_{i}, Y_{i}) \}_{i=1}^{N} \xrightarrow[\text{observation at } \, \omega{}_{0}]{} \mathcal{D}_{\text{obs}} = \{ (x_{i}, y_{i} )\}_{i=1}^{N} $
#### $ \hspace{0.3cm} \text{where } \, X_{i} : \Omega{} \to{} \mathcal{X}, \;\; Y_{i} : \Omega{} \to{} \mathcal{Y}, \;\; x_{i} = X_{i}(\omega{}_{0}) \in{} \mathcal{X}, \;\; y_{i} = Y_{i}(\omega{}_{0}) \in{} \mathcal{Y} $
#### $ \hspace{0.3cm} \text{and } \, \Omega{} : \text{sample space}, \;\; \omega{}_{0} \in{} \Omega{} : \text{outcome} $

#### **(2) 임무(Task)** :
##### $ \hspace{0.15cm} $ ① 결정론적 모델링(Deterministic Modeling) : 관측된 출력값 $ y_{i} $에 대한 점추정(point prediction)
##### $ \hspace{0.45cm} \Rrightarrow{} \hat{y}_{i} = h_{\theta{}}(x_{i}) \approx{} y_{i}, \;\;  h_{\theta{}} : \mathcal{X} \to{} \mathcal{Y} $
##### $ \hspace{0.45cm} \hspace{0.3cm} \text{where } \, h_{\theta{}} \in{} \mathcal{H} := \{ h_{\theta{}} \mid{} \theta{} \in{} \Theta{} \} $
##### $ \hspace{0.15cm} $ ② 확률론적 모델링(Probabilistic Modeling) : 모집단(참) 조건부분포(conditional distribution)의 밀도표현 $ p^{*}_{Y|X}(y|x) $ 근사
#### $ \hspace{0.45cm} \Rrightarrow{} p_{\theta{}}(y|x) \approx{} p^{*}_{Y|X}(y|x) $
#### $ \hspace{0.45cm} \hspace{0.3cm} \text{where } \, p^{*}_{Y|X}(y|x) : \text{density representation of true conditional distribution (we don't know)} $ 

#### **(3) 측정 지표(Performance measure)** : 원 확률 측도 $ P^{*}_{X,Y} $에서 새롭게 추출될 샘플에 대한 기대 손실, 모집단 위험(Population Risk)을 추정 
##### $ \hspace{0.15cm} $ ① 결정론적 모델링(Deterministic Modeling)
##### $ \hspace{0.45cm} \Rrightarrow{} \displaystyle{} \mathcal{R}(\theta{}) = \mathbb{E}_{(X,Y)\sim{}P^{*}_{X,Y}} \big[{}\ell{}(h_{\theta{}}(X), Y)\big]{} = \int_{\mathcal{X}\times{}\mathcal{Y}} {}\ell{}(h_{\theta{}}(x), y) P^{*}_{X,Y}( \text{d} (x,y)) $
##### $ \hspace{0.45cm} \hspace{0.3cm} \text{where } \, \mathcal{R}(\theta{}) : \text{population risk}, \;\; \ell{}(h_{\theta{}}(X), Y) : \text{loss function} $
##### $ \hspace{0.15cm} $ ② 확률론적 모델링(Probabilistic Modeling) : 손실 함수(loss function;$ \ell{}(\cdot{}) $)을 Negative Log Likelihood로 설정
##### $ \hspace{0.45cm} \Rrightarrow{} \displaystyle{} \mathcal{R}_{\text{NLL}}(\theta{}) = \mathbb{E}_{(X,Y)\sim{}P^{*}_{X,Y}} \big[{}-\log\big({}p_{\theta{}}(Y|X)\big){}\big]{} $

<b></b>

> ## (Theorem 1-1) **Empirical Approximation of Population Risk**

#### 모집단 위험(Population Risk)은 실제 학습 단계에서는 알 수 없으므로, 관측된 유한 표본 $ \mathcal{D}_{\text{obs}} $의 평균인 경험적 위험(Empirical Risk)으로 근사한다.
#### 해당 근사는 독립 항등 분포(i.i.d) 하에서 큰수의 법칙에 의해 정당화된다.

#### **(1)** 결정론적 모델링
#### $ \Rrightarrow{} \displaystyle{} \hat{\mathcal{R}}_{N}(\bar{\theta{}}) = \frac{1}{N} \sum^{N}_{i=1} \ell{}(h_{\bar{\theta{}}}(x_{i}), y_{i}) \xrightarrow[N\to{}\infty{}]{} \mathcal{R}(\bar{\theta{}}) = \mathbb{E}_{(X,Y)\sim{}P^{*}_{X,Y}} \big[{}\ell{}(h_{\bar{\theta{}}}(X), Y)\big]{} $ 
#### $ \hspace{0.3cm} \text{where } \bar{\theta{}} : \text{fixed parameter} $

#### **(2)** 확률론적 모델링
#### $ \Rrightarrow{} \displaystyle{} \hat{\mathcal{R}}_{\text{NLL}}(\bar{\theta{}}) = \frac{1}{N}\sum^{N}_{i=1}-\log{}\big({}p_{\bar{\theta{}}}(y_{i}|x_{i})\big){} \xrightarrow[N\to{}\infty{}]{} \mathcal{R}_{\text{NLL}}(\bar{\theta{}}) = \mathbb{E}_{(X,Y)\sim{}P^{*}_{X,Y}} \big[{}-\log\big({}p_{\bar{\theta{}}}(Y|X)\big){}\big]{} $

<b></b>

> ## (Theorem 1-2) **Probabilistic Learning via Maximum Likelihood Estimation(MLE)**

#### **(1)** 확률적 모델의 학습은 관측 데이터 $ \mathcal{D}_{\text{obs}} = \{(x_i,y_i)\}_{i=1}^{N} $를 가장 높은 우도(likelihood)로 설명하는 모델 $ h_{\theta{}} $ (또는 파라미터 $ \theta{} $)를 선택하는 과정이다. 
#### $ \Rrightarrow{} \theta{}_{\text{MLE}} = \displaystyle{} \argmax_{\theta{} \in{} \Theta{}} \bigg({} \text{likelihood}_{\text{cond}}(\theta{}) \bigg){} = \argmax_{\theta{} \in{} \Theta{}} \bigg({} \prod^{N}_{i=1} p_{\theta{}}(y_{i}|x_{i}) \bigg){} $
#### $ \hspace{0.15cm} \because{} \, \displaystyle{} \text{likelihood}_{\text{cond}}(\theta{}) = p_{\theta{}}(y_{1}, y_{2}, \cdots{}, y_{N} | x_{1}, x_{2}, \cdots{}, x_{N}) = \displaystyle{} \prod^{N}_{i=1} p_{\theta{}}(y_{i}|x_{i}) \;\; (\text{conditional i.i.d. factorization}) $

#### **(2)** MLE는 (조건부) likelihood maximization과 NLL minimization의 동치 관계이다
#### $ \Rrightarrow{} \theta{}_{\text{MLE}} = \displaystyle{} \argmax_{\theta{}\in{}\Theta{}} \bigg({} \prod^{N}_{i=1} p_{\theta{}}(y_{i}|x_{i}) \bigg){} = \argmax_{\theta{}\in{}\Theta{}} \bigg({} \sum^{N}_{i=1} \log{}(p_{\theta{}}(y_{i}|x_{i})) \bigg){} $
#### $ \hspace{1.35cm} = \displaystyle{} \argmin_{\theta{}\in{}\Theta{}} \bigg({} \sum^{N}_{i=1} - \log{}(p_{\theta{}}(y_{i}|x_{i})) \bigg){} = \theta{}_{\text{NLL}} $

<b></b>

> ## (Theorem 1-3) **Population Risk via KL-divergence Perspective**

#### **(1)** 모집단 NLL 위험은 참 조건부 분포의 평균 정보량(entropy)의 기댓값과 참 조건부 분포-모델 조건부 분포 간의 KL 발산(KL divergence)의 기댓값으로 분해된다. 
#### $ \text{assume } \, (X,Y)\sim{}P^{*}_{X,Y}, \;\; X\sim{}P^{*}_{X} = P^{*}_{X}(\cdot{}) $
#### $ \text{let } \, P^{*}_{Y|X}(\cdot{}|X) : \text{conditional distribution of } \, Y \text{ given } \, X $
#### $ \mathcal{L}_{\mathrm{NLL}}(\theta{}) := \mathbb{E}_{(X,Y)\sim{}P^{*}_{X,Y}}\big[{}-\log{}p_{\theta{}}(Y|X)\big]{} = \mathbb{E}_{X\sim{}P^{*}_{X}}\bigg[{}\mathbb{E}_{Y\sim{}P^{*}_{Y|X}(\cdot{}|X)}\big[{}-\log{}p_{\theta{}}(Y|X)\big]{}\bigg]{} \;\; (\because{} \text{tower property}) $
##### $ \hspace{1.5cm} = \mathbb{E}_{X\sim{}P^{*}_{X}}\bigg[{}\mathbb{E}_{Y\sim{}P^{*}_{Y|X}(\cdot{}|X)}\bigg[{}-\log{}p^{*}_{Y|X}(Y|X)+\log{}p^{*}_{Y|X}(Y|X)-\log{}p_{\theta{}}(Y|X)\bigg]{}\bigg]{} $
##### $ \hspace{1.5cm} = \mathbb{E}_{X\sim{}P^{*}_{X}}\bigg[{}\mathbb{E}_{Y\sim{}P^{*}_{Y|X}(\cdot{}|X)}\bigg[{}-\log{}p^{*}_{Y|X}(Y|X)+\log{}\frac{p^{*}_{Y|X}(Y|X)}{p_{\theta{}}(Y|X)}\bigg]{}\bigg]{} $
##### $ \hspace{1.5cm} = \mathbb{E}_{X\sim{}P^{*}_{X}}\bigg[{}\mathbb{E}_{Y\sim{}P^{*}_{Y|X}(\cdot{}|X)}\big[{}-\log{}p^{*}_{Y|X}(Y|X)\big]{}\bigg]{}+\mathbb{E}_{X\sim{}P^{*}_{X}}\bigg[{}\mathbb{E}_{Y\sim{}P^{*}_{Y|X}(\cdot{}|X)}\bigg[{}\log{}\frac{p^{*}_{Y|X}(Y|X)}{p_{\theta{}}(Y|X)}\bigg]{}\bigg]{} $
##### $ \hspace{1.5cm} = \mathbb{E}_{X\sim{}P^{*}_{X}}\bigg[{}\underbrace{\mathbb{E}_{Y\sim{}P^{*}_{Y|X}(\cdot{}|X)}\big[{}-\log{}p^{*}_{Y|X}(Y|X)\big]{}}_{\text{entropy term}}\bigg]{}+\mathbb{E}_{X\sim{}P^{*}_{X}}\bigg[{}\underbrace{D_{\mathrm{KL}}\big({}P^{*}_{Y|X}(\cdot{}|X)\mid\mid{}P_{\theta{}}(\cdot{}|X)\big){}}_{\text{KL divergence term}}\bigg]{} $

#### **(2)** 모집단 NLL 최소화(Population NLL minimization)는 KL-발산항의 기댓값을 최소화하는 것과 동치이다.
#### $ \Rrightarrow{} \displaystyle{} \theta{}^{*}_{\text{NLL}} = \argmin_{\theta{}\in{}\Theta{}} \mathcal{L}_{\mathrm{NLL}}(\theta{}) = \argmin_{\theta{}\in{}\Theta{}} \mathbb{E}_{X\sim{}P^{*}_{X}}\bigg[{}D_{\mathrm{KL}}\big({}P^{*}_{Y|X}(\cdot{}|X)\mid\mid{}P_{\theta{}}(\cdot{}|X)\big){}\bigg]{} $

> ## **자기-지도 학습(Self-supervised Learning)**

#### **(1) 경험(Experience)** : 
#### $ \text{assume } \, X \sim{} P^{*}_{X}, \;\; X_{i} \overset{\text{i.i.d.}}{\sim} P^{*}_{X}, \;\; i = 1, 2, \cdots{}, N $
#### $ \hspace{0.3cm} \mathcal{D} = \{ X_{i} \}_{i=1}^{N} \xrightarrow[\text{observation at } \, \omega{}_{0}]{} \mathcal{D}_{\text{obs}} = \{ x_{i} \}_{i=1}^{N} $
#### $ \text{let } \, \mathcal{T} : \mathcal{X} \to{} \widetilde{\mathcal{X}} \times{} \mathcal{Y}_{\text{self}}, \;\; (\widetilde{X}_{i}, Y^{\text{self}}_{i}) := \mathcal{T}(X_{i}) $ 
#### $ \Rrightarrow{} \mathcal{D}_{\text{self, obs}} = \{(\widetilde{x}_{i}, y^{\text{self}}_{i})\}_{i=1}^{N} $
#### $ \hspace{0.3cm} \text{where } \, X_{i} : \Omega{} \to{} \mathcal{X}, \;\; \widetilde{X}_{i} : \Omega{} \to{} \widetilde{\mathcal{X}}, \;\; Y^{\text{self}}_{i} : \Omega{} \to{} \mathcal{Y}_{\text{self}} $
#### $ \hspace{0.3cm} \text{and } \, \Omega{} : \text{sample space}, \;\; \omega{}_{0} \in{} \Omega{}, \;\; \mathcal{T} : \text{self-supervised task generator}, \;\; y^{\text{self}}_{i} : \text{pseudo-label constructed from } x_{i} $

#### **(`PLUS`)** 임무(task), 측정 지표(performance measure)는 기존 지도학습과 동일

<b></b>

> ## **비지도 학습(Unsupervised Learning)**

#### **(1) 경험(Experience)** :
#### $ \text{assume } \, X \sim{} P^{*}_{X}, \;\; X_{i} \overset{\text{i.i.d.}}{\sim} P^{*}_{X}, \;\; i = 1, 2, \cdots{}, N $
#### $ \Rrightarrow{} \mathcal{D} = \{ X_{i} \}_{i=1}^{N} \xrightarrow[\text{observation at } \, \omega{}_{0}]{} \mathcal{D}_{\text{obs}} = \{ x_{i} \}_{i=1}^{N} $
#### $ \hspace{0.3cm} \text{where } \, X_{i} : \Omega{} \to{} \mathcal{X}, \;\; x_{i} = X_{i}(\omega{}_{0}) \in{} \mathcal{X} $
#### $ \hspace{0.3cm} \text{and } \, \Omega{} : \text{sample space}, \;\; \omega{}_{0} \in{} \Omega{} : \text{outcome} $

#### **(2) 임무(Task)** :
##### $ \hspace{0.15cm} $ ① 결정론적 함수 모델링(Deterministic Function Modeling) : label 없이 입력 데이터 $ x_i $의 구조적 산출물(structural output)을 학습
##### $ \hspace{0.45cm} \Rrightarrow{} s_i = \phi_{\theta{}}(x_i), \;\; \phi_{\theta{}} : \mathcal{X} \to{} \mathcal{S} $
##### $ \hspace{0.45cm} \hspace{0.3cm} \text{where } \, s_i \in{} \mathcal{S} : \text{task-induced structural output} $
##### $ \hspace{0.45cm} \hspace{0.3cm} \text{and } \, \mathcal{S} : \text{structure space induced by the unsupervised objective} $
##### $ \hspace{0.15cm} $ ② 확률론적 모델링(Probabilistic Modeling) : 모집단(참) 주변분포(marginal distribution) \, $ p^{*}_{X}(x) $ 근사
##### $ \hspace{0.45cm} \Rrightarrow{} p_{\theta{}}(x) \approx{} p^{*}_{X}(x) $
##### $ \hspace{0.45cm} \hspace{0.3cm} \text{where } \, p^{*}_{X}(x) : \text{true marginal distribution(we don't know)} $

#### **(3) 측정 지표(Performance measure)** : 모집단 위험(Population Risk)로 정의
##### $ \hspace{0.15cm} $ ① 결정론적 함수 모델링(Deterministic Function Modeling)
##### $ \hspace{0.45cm} \Rrightarrow{} \displaystyle{} \mathcal{L}(\theta{}) = \mathbb{E}_{X\sim{}P^{*}_{X}} \big[{}\ell{}(X, \phi_{\theta{}}(X))\big]{} $
##### $ \hspace{0.45cm} \hspace{0.3cm} \text{where } \, \mathcal{L}(\theta{}) : \text{population risk}, \;\; \ell{}(X, \phi_{\theta{}}(X)) : \text{task-induced loss function} $
##### $ \hspace{0.15cm} $ ② 확률론적 모델링(Probabilistic Modeling) : loss function $ \ell{}(\cdot{}) $을 Negative Log Likelihood로 설정
##### $ \hspace{0.45cm} \Rrightarrow{} \displaystyle{} \mathcal{L}_{\text{NLL}}(\theta{}) = \mathbb{E}_{X\sim{}P^{*}_{X}} \big[{}-\log{}\big({}p_{\theta{}}(X)\big){}\big]{} $
