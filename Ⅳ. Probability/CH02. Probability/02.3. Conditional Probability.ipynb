# CH01.4. **Conditional Probability**

> ## **조건부 확률(Conditional Probability)**

#### **(1) 정의** : 어떠한 사건이 이미 발생했다고 가정할 때 다른 사건이 일어날 확률 
#### $ \Rrightarrow{} P(A|B) = \displaystyle{} \frac{P(A\cap{}B)}{P(B)} $

##### **(`PLUS`)** 조건부 확률은 표본 공간이 $ B $로 축소됐다고 이해할 수 있음

#### **(2) 예시** :
##### $ \hspace{0.15cm} $ ① 공평한 주사위를 $ 1 $회 던졌을 때 $ 3 $ 이하 눈을 뽑았고 그 중 홀수를 뽑을 확률을 구하면
##### $ \hspace{0.45cm} \Omega{} = \{ 1, 2, 3, 4, 5, 6 \} $
##### $ \hspace{0.45cm} A = \{ 1, 3, 5 \} $ : 홀수를 뽑을 사건
##### $ \hspace{0.45cm} B = \{ 1, 2, 3 \} $ : $ 3 $ 이하를 뽑을 사건
##### $ \hspace{0.45cm} A \cap{} B = \{ 1, 3, 5 \} \cap{} \{ 1, 2, 3 \} = \{ 1 \} $ : 홀수 및 $ \, 3 $ 이하를 뽑을 사건
##### $ \hspace{0.45cm} \therefore{} P(A|B) = \frac{P(A\cap{}B)}{P(B)} = \frac{|A\cap{}B|}{|B|} = \frac{|\{ 1 \}|}{|\{ 1, 2, 3 \}|} = \frac{1}{3} \;\; \because{} \text{equally likely outcome} $
##### $ \hspace{0.15cm} $ ② **[CONTENTS]**

> ## (Theorem 2-1) **Chain rule of Conditional Probability**

#### 서로소가 아닌 사건 $ A_{1}, A_{2}, \cdots{}, A_{n} $의 교집합 확률은 아래와 같다.
#### $ \Rrightarrow{} P(\cap{}^{n}_{i=1}A_{i}) = P(A_{1})P(A_{2}|A_{1})P(A_{3}|A_{1}\cap{}A_{2})\cdots{}P(A_{n}|A_{1}\cap{}A_{2}\cap{}A_{3}\cap{}\cdots{}\cap{}A_{n-1}) $

##### **(`WHY?`)** 
##### $ \hspace{0.15cm} \text{if } \,B_{1} = ( A_{2}\cap{}A_{3}\cap{}\cdots{}\cap{}A_{n} ) $
##### $ \hspace{0.3cm} P(A_{1}\cap{}A_{2}\cap{}\cdots{}\cap{}A_{n}) = P(A_{1}\cap{}B_{1}) = P(A_{1}|B_{1})P(B_{1}) $
##### $ \hspace{0.15cm} \text{if } \,B_{2} = ( A_{3}\cap{}A_{4}\cap{}\cdots{}\cap{}A_{n} ), \;\; B_{1} = A_{2}\cap{}B_{2} $
##### $ \hspace{0.3cm} P(B_{1}) = P(A_{2}\cap{}B_{2}) = P(A_{2}|B_{2})P(B_{2}) $
##### $ \hspace{0.2cm} \vdots{} $ 
##### $ \hspace{0.15cm} \text{if } \,B_{n-1} = A_{n}, \;\; B_{n-1} = A_{n-1}\cap{}B_{n-1} $
##### $ \hspace{0.3cm} P(B_{n-1}) = P(A_{n}\cap{}B_{n}) = P(A_{n}|B_{n})P(B_{n}) $
##### $ \hspace{0.15cm} P(A_{1}\cap{}A_{2}\cap{}\cdots{}\cap{}A_{n}) = P(A_{1}|B_{1})P(A_{2}|B_{2})\cdots{}P(A_{n}|B_{n-1}) $
##### $ \hspace{0.15cm} \therefore{} \; P(A_{1}\cap{}A_{2}\cap{}\cdots{}\cap{}A_{n}) = P(A_{1})P(A_{2}|A_{1})P(A_{3}|A_{1}\cap{}A_{2})\cdots{}P(A_{n}|A_{1}\cap{}A_{2}\cap{}A_{3}\cap{}\cdots{}\cap{}A_{n-1}) $

> ## **사건의 독립(Independence of Event)**

#### **(1) 정의** : 어떠한 사건의 확률이 다른 사건의 확률에 영향을 미치지 않는 관계
#### $ \Rrightarrow{} P(A\cap{}B) = P(A)P(B) \;\; \text{ where } \, A, B \in{} \mathcal{F} $ 

#### **(2) 예시** :
##### $ \hspace{0.15cm} $ ① 공평한 주사위 $ 1 $회 던졌을 때 눈이 짝수인 사건 $ A $와 $ \, 3 $ 이하인 사건 $ B $가 있을 때 두 사건의 관계를 파악하려 하는 경우   
##### $ \hspace{0.45cm} \Omega{} = \{ 1, 2, 3, 4, 5, 6 \}, \;\; A = \{ 2, 4 \}, \;\; B = \{ 1, 2, 3 \}, \;\; A \cap{} B = \{ 2 \} $
##### $ \hspace{0.45cm} P(A\cap{}B) = \frac{|A\cap{}B|}{|\Omega{}|} = \frac{1}{6} $
##### $ \hspace{0.45cm} P(A)P(B) = \frac{|A|}{|\Omega{}|} \times{} \frac{|B|}{|\Omega{}|} = \frac{2}{6}\times{} \frac{3}{6} = \frac{1}{6} = P(A\cap{}B) $
##### $ \hspace{0.45cm} \therefore{}  A, \, B \, \text{ is independent (event)}. $
##### $ \hspace{0.15cm} $ ② **[CONTENTS]**

<b></b>

> ## (Theorem 2-2) **Mutually independence**

#### 어떠한 사건 $ A, B, C $가 존재할 때 해당 사건들의 아래 네 조건을 만족해야지 상호(mutually) 독립 관계임을 정의한다. 
#### $ \Rrightarrow{} \left.{} \begin{array}{l} P(A\cap{}B) = P(A)P(B) \\ \\ P(A\cap{}C) = P(A)P(C) \\ \\ P(C\cap{}B) = P(C)P(B) \end{array} \right\}{} \;\; \cdots{} \; \text{pairwise independence} $
#### $ \hspace{0.675cm} P(A\cap{}B\cap{}C) = P(A)P(B)P(C) $

<b></b>

> ## (Theorem 2-3) **Union of Independent Event**

#### 어떠한 사건 $ A_{1}, A_{2}, \cdots{}, A_{n} $이 서로 독립 관계일 때 그 사건들의 합집합 확률은 아래와 같다. 
#### $ \Rrightarrow{} P(\cup^{n}_{i=1}A_{i}) = P\big({}(A_{1}^{c}\cap{}A_{2}^{c}\cap{}\cdots{}\cap{}A_{n}^{c})^{c}\big){} \;\; \because{} $ Theorem 1-1
#### $ \hspace{2.125cm} = 1 - P(A_{1}^{c}\cap{}A_{2}^{c}\cap{}\cdots{}\cap{}A_{n}^{c}) $ 
#### $ \hspace{2.125cm} = 1 - P(A_{1}^{c})P(A_{2}^{c})\cdots{}P(A_{n}^{c}) $
#### $ \hspace{0.45cm} \text{ s.t. } \, \{ A_{1}, \cdots{}, A_{n} \} \, \text{ is mutually independent}. $

<b></b>

> ## **사건의 배반(Event of Mutual Exclusivity Relationship)**

#### **(1) 정의** : 두 사건이 동시에 일어날 수 없는 관계
#### $ \Rrightarrow{} P(A \cap{} B) = 0 \;\; \text{ s.t. } \, P(A) \neq{} 0, \;\; P(B) \neq{} 0 $

<b></b>

> ## (Theorem 2-4) **Law of Total Probability**

#### 표본 공간을 분할하는 사건 $ B_{1}, B_{2}, \cdots{}, B_{n} $가 존재할 때 사건 $ A $의 확률은 아래와 같다.
#### $ \Rrightarrow{} P(A) = \displaystyle{} \sum^{n}_{i=1}P(A\cap{}B_{i}) $
#### $ \hspace{1.375cm} = \displaystyle{} \sum^{n}_{i=1} P(A|B_{i})P(B_{i}) $
#### $ \hspace{0.45cm} \text{s.t. } \, B_{i}\cap{}B_{k} = \emptyset{}, \;\; \cup{}_{i=1}B_{i} = \Omega{}, \;\; i \neq{} k $

> ## **베이즈 정리(Bayes Theorem)**

#### **(1) 정의** : 어떤 사건 $ 𝐵 $가 발생한다고 가정했을 때, 그 원인으로 사건 $ A $가 일어났을 확률을 계산하는 방법
#### $ \Rrightarrow{} P(A|B) = \displaystyle{} \frac{P(A\cap{}B)}{P(B)} = \frac{P(B|A)P(A)}{P(B)} $

#### **(2) 구성** :
##### $ \hspace{0.15cm} $ ① 사후확률(posterior) : 사건 $ B $가 관측된 후에 사건 $ A $가 참일 확률 $ \rightarrow{} P(A|B) $
##### $ \hspace{0.15cm} $ ② 우도(likelihood) : 사건 $ A $가 주어졌을 때 $ B $가 관측될 가능성 $ \rightarrow{} P(B|A) $
##### $ \hspace{0.15cm} $ ③ 사전확률(prior) : 관측 이전에 사건 $ A $가 일어날 것이라고 믿는 신념(확률) $ \rightarrow{} P(A) $
##### $ \hspace{0.15cm} $ ④ 증거(evidence) : 전체적으로 $ B $가 관측될 주변 확률 $ \rightarrow{} P(B) $

#### **(3) 예시** :
##### $ \hspace{0.15cm} $ ① 어떠한 제품이 실제 불량품인 비율은 $ 5\% $이고 QC에서 오검사할 확률이 $ 10 \% $일 때, 양성으로 판별했지만 실제 불량품일 확률을 구하면
##### $ \hspace{0.45cm} \Omega{} = \{ (B, P), (B, N), (G, P), (G, N) \} $
##### $ \hspace{0.45cm} B $ : 생산 제품이 불량품일 사건 
##### $ \hspace{0.45cm} G $ : 생산 제품이 양품(정상 제품)일 사건
##### $ \hspace{0.45cm} N $ : QC에서 양품이라고 판별할 사건
##### $ \hspace{0.45cm} P $ : QC에서 불량품이라고 판별할 사건 
##### $ \hspace{0.45cm} P(B) = 0.05 $ : 생산 제품이 불량품일 확률 
##### $ \hspace{0.45cm} P(P|B) = P(N|G) = 0.1 $ : 오검사(QC에서 실제 불량품을 양품 혹은 실제 양품을 불량품이라고 판별)할 확률
##### $ \hspace{0.45cm} P(B, P) = P(B) \times{} P(P|B) = 0.05 \times{} 0.1 = 0.005 $  
##### $ \hspace{0.45cm} P(B, N) = P(B) \times{} P(N|B) = 0.05 \times{} (1 - 0.1) = 0.05 \times{} 0.9 = 0.045 $  
##### $ \hspace{0.45cm} P(G, P) = P(G) \times{} P(P|G) = (1 - 0.05) \times{} 0.1 = 0.95 \times{} 0.1 = 0.095 $  
##### $ \hspace{0.45cm} P(G, N) = P(G) \times{} P(N|G) = (1 - 0.05) \times{} (1 - 0.1) = 0.855 $
##### $ \hspace{0.45cm} \therefore{} P(B|P) = \frac{P(B,P)}{P(P)} = \frac{P(B,P)}{P(B,P)+P(G,P)} = \frac{0.005}{0.005+0.095} =  0.05 $
##### $ \hspace{0.15cm} $ ② **[CONTENTS]**

<b></b>

> ## (Theorem 2-5) **Expaned Form of Bayes's Rule**

#### 표본 공간을 분할(partitioned)하는 사건 $ A_{i} $가 존재할 때 $ B $에 대한 $ A_{i} $의 베이즈 정리는 다음과 같이 정의된다.
#### $ \Rrightarrow{} P(A_{i}|B) = \frac{P(B|A_{i})P(A_{i})}{\sum_{j}P(B|A_{j})P(A_{j})} \;\; \text{ s.t. } \, \Omega{} = \cup_{j}A_{j}, \;\; A_{i} \cap{} A_{j} = \empty{} $

<b></b>

> ## **조건부 독립(Conditional Independence)**

#### **(1) 정의** : 어떠한 사건 $ C $가 주어졌을 때 $ A, B $가 독립인 관계
#### $ \Rrightarrow{} P(A\cap{}B|C) = P(A|C)P(B|C) \;\; \text{ s.t. } \, P(C) \neq{} 0 $

##### **(`PLUS`)** 조건부 독립은 사건 $ C $이 발생한 상황에서만 성립하는 관계이며, $ A, B $가 항상 독립이라는 의미는 아님
##### $ \hspace{0.15cm} \Rrightarrow{} P(A\cap{}B|C) = P(A|C)P(B|C) \not\leftrightarrow{} P(A\cap{}B) = P(A)P(B) $
