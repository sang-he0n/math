#### **(1) 정의** : 집합에 길이, 넓이, 부피, 확률 등의 크기 개념을 일반화하는 이론

> ## **시그마-대수($\sigma{}$-Algebra)**

#### **(1) 정의** : 어떠한 집합 $ U $를 측정 가능한 집합들로 정의하기 위한 그 집합족(collection)
#### $ \Rrightarrow{} \mathcal{F} $

##### **(`PLUS`)** 확률론에서 시그마-대수는 실험에서 연구자가 알고 있는 정보의 수준으로 해석함

#### **(2) 공리** : 어떠한 집합 $ U $에 대응되는 시그마-대수 $ \mathcal{F} $가 존재할 때 아래를 만족해야함
##### $ \hspace{0.15cm} $ ① $ \mathcal{B} \ni{} \empty{} $ 
##### $ \hspace{0.15cm} $ ② $ A \in{} B \; \Rightarrow{} A^{c} \in{} \mathcal{F} $ 
##### $ \hspace{0.15cm} $ ③ $ A_{i} \in{} B \; \Rightarrow{} \cup^{\infty{}}_{i=1} A_{i} \in{} \mathcal{F} $

#### **(3) 성질** : 어떠한 집합 $ U $에 대응되는 시그마-대수 $ \mathcal{F} $가 존재할 때 아래를 만족함
##### $ \hspace{0.15cm} $ ① $ U \in{} \mathcal{F} $
##### $ \hspace{0.15cm} $ ② $ A, B \in{} \mathcal{F} \; \Rightarrow{} \; \displaystyle{} (A \cup{} B) \in{} \mathcal{F} $
##### $ \hspace{0.15cm} $ ③ $ A_{i} \in{} \mathcal{F} \; \Rightarrow{} \; \displaystyle{} \cap^{\infty{}}_{i=1} A_{i} \in{} \mathcal{F} $
##### $ \hspace{0.15cm} $ ④ **[LATEX]**
##### $ \hspace{0.15cm} $ ⑤ **[LATEX]**

#### **(4) 종류** :
##### $ \hspace{0.15cm} $ ① 자명한 시그마-대수(Trivial $ \sigma{} $-Algebra) :
##### $ \hspace{0.45cm} \Rrightarrow{} \mathcal{F} := \{ \empty{}, U \} $
##### $ \hspace{0.15cm} $ ② 보렐 시그마-대수(Borel $ \sigma{} $-Algebra) : 열린 구간 $ (a, b) $를 포함하는 가장 작은 시그마-대수
##### $ \hspace{0.45cm} \Rrightarrow{} \mathcal{B}(\mathbb{R}) := \sigma{}({(a, b) \subseteq{} \mathbb{R}  \; : \; a<b, \;\;  a, b \in{} \mathbb{R}}) \;\; \text{ where } \, \sigma{}(\cdot{}) : \text{ generate operator} $ 
##### $ \hspace{0.15cm} $ ③ 멱집합 시그마 대수(Powerset $ \sigma{} $-Algebra) :
##### $ \hspace{0.45cm} \Rrightarrow{} \mathcal{F} := 2^{U} $
##### $ \hspace{0.15cm} $ ④ 생성 시그마-대수(Generated $ \sigma{} $-Algebra) : 어떠한 집합족 $ \mathcal{C} $를 포함하는 가장 작은 시그마-대수

> ## **가측 공간(Measurable Space)**

#### **(1) 정의** : 어떠한 집합과 그 집합과 대응되는 시그마-대수으로 구성된 객체(튜플)
#### $ \Rrightarrow{} (A, \mathcal{F}) $ 

<b></b>

> ## **함수의 가측성(Measurability of Function)**

#### **(1) 정의** : 두 가측 공간을 매핑하는 함수 $ f $가 존재할 때, $ \, f $의 역상(preimage)가 정의역(domain)의 시그마 대수 안에 포함되는 성질
#### $ \Rrightarrow{} f^{-1}(G) \in{} \mathcal{F} \;\; \text{ where } \, f : (A, \mathcal{F}) \rightarrow{} (B, \mathcal{G}), \;\; \, \mathcal{F}, \mathcal{G} \, \text{ are } \, \sigma{}-\text{alegbras} $

<b></b>

> ## **측도(Measure)**

#### **(1) 정의** : 어떠한 가측 공간에서 시그마-대수의 원소를 사용하여 비음수 실수 스칼라를 반환하는 집합 함수(set function)
#### $ = $ 집합의 크기를 산정하는 함수
#### $ \Rrightarrow{} \mu{} : \mathcal{B} \rightarrow{} [0, \infty{} ] $

#### **(2) 공리** : 어떠한 시그마-대수 $ \mathcal{B} $에 대응되는 측도 $ \mu{} $가 존재할 때 아래를 만족해야함
##### $ \hspace{0.15cm} $ ① $ \mu{}(A) \geq{} 0 \;\; \forall{} \, A \in{} \mathcal{B} $
##### $ \hspace{0.15cm} $ ② $ \mu{}(\empty{}) = 0 $ 
##### $ \hspace{0.15cm} $ ③ $ \displaystyle{} A_{i} \cap{} A_{j} = \empty{} \; \Rightarrow{} \; \mu{}(\cup^{\infty{}}_{i=1}A_{i}) = \sum^{\infty{}}_{i=1} \mu{}(A_{i}) $ 

#### **(3) 종류** :
##### $ \hspace{0.15cm} $ ① 가산 측도(Counting measure)
##### $ \hspace{0.15cm} $ ② 르베그 측도(Lebesgue measure) : 보렐 시그마-대수 위의 집합들의 길이를 정의하는 측도
##### $ \hspace{0.45cm} \Rrightarrow{} \lambda{} : \mathcal{B}(\mathbb{R}) \rightarrow{} [0, \infty{}] $
##### $ \hspace{0.15cm} $ ③ 확률 측도(Probability measure)
