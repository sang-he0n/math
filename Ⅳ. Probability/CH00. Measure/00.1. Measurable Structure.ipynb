> ## **시그마-대수($\sigma{}$-Algebra)**

#### **(1) 정의** : 집합 $ U $에서 여집합 및 (가산) 합집합 연산에 닫혀있고 공집합을 포함하는 집합족 $ \mathcal{F} $
#### $ \Rrightarrow{} \mathcal{F} \subseteq{} 2^{U} $

#### **(2) 공리** : 어떠한 집합 $ U $에 대응되는 시그마-대수 $ \mathcal{F} $가 존재할 때 아래를 만족해야함
##### $ \hspace{0.15cm} $ ① $ \emptyset{} \in{} \mathcal{F} $ 
##### $ \hspace{0.15cm} $ ② $ A \in{} \mathcal{F} \; \Rightarrow{} \; A^{c} := U \setminus{} A \in{} \mathcal{F} $ 
##### $ \hspace{0.15cm} $ ③ $ \displaystyle{} A_{i} \in{} \mathcal{F} \; (\forall{}\,i) \; \Rightarrow{} \; \bigcup^{\infty{}}_{i=1} A_{i} \in{} \mathcal{F} $

#### **(3) 성질** : 어떠한 집합 $ U $에 대응되는 시그마-대수 $ \mathcal{F} $가 존재할 때 아래를 만족함
##### $ \hspace{0.15cm} $ ① $ U \in{} \mathcal{F} $
##### $ \hspace{0.15cm} $ ② $ \displaystyle{} A_{i} \in{} \mathcal{F} \; \Rightarrow{} \; \bigcap^{\infty{}}_{i=1} A_{i} \in{} \mathcal{F} \;\; \text{ where } \, A_{i} \subset{} U $
##### $ \hspace{0.15cm} $ ③ $ A, B \in{} \mathcal{F} \; \Rightarrow{} \; (A \cup{} B) \in{} \mathcal{F}, \;\; (A \cap{} B) \in{} \mathcal{F}, \;\;  (A \cap{} B^{c}) \in{} \mathcal{F} $

> ## **생성 시그마-대수(Generated $ \sigma{} $-Algebra)**

#### **(1) 정의** : 집합족 $ \mathcal{C} $가 생성하는 가장 작은 시그마-대수
