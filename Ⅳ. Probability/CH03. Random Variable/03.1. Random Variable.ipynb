> ## **함수(Function)**

#### **(1) 정의** : 특정 집합에서 다른 집합으로의 대응 관계를 나타내는 규칙
#### $ \Rrightarrow{} f : A \rightarrow{} B \;\; \text{ where } \, A, B \, \text{ are sets} $

##### $ \hspace{0.15cm} $ ① 정의역(domain) : 함수에 입력할 수 있는 모든 원소(input)들의 집합 $ A $
##### $ \hspace{0.15cm} $ ② 공역(codomain) : 함수값이 속할 수 있는 가능한 출력값들의 집합 $ B $ 
##### $ \hspace{0.15cm} $ ③ 치역(range) : 정의역 내 입력값이 함수를 통해 대응되는 출력값들의 집합 $ f(A) $

> ## **확률 변수(Random Variable;RV)**

#### **(1) 정의** : 확률 공간 $ (\Omega{}, \mathcal{F}, P) $과 가측 공간 $ \big({}\mathbb{R}, \mathcal{B}(\mathbb{R})\big){} $을 매핑하는 가측(measurable) 함수
#### $ = $ 표본 공간의 원소(근원 사건)을 어떠한 규칙에 의해 실수로 반환하는 함수
#### $ \Rrightarrow{} X(\omega{}) : (\Omega{},\mathcal{F}, P) \rightarrow{} \big({}\mathbb{R}, \mathcal{B}(\mathbb{R})\big){} \;\; \text{ s.t. } \, X^{-1}(B) = \{ \omega{} \in{} \Omega{} : X(\omega{}) \in{} B \} \in{} \mathcal{F}, \;\; \forall{} B \in{} \mathcal{B}(\mathbb{R}) $ 

##### **(`PLUS`) 확률 변수의 치역(range)** : 어떠한 확률 변수가 가질 수 있는 모든 값으로 이루어진 집합
##### $ \hspace{0.15cm} \Rrightarrow{} R_{X} := X(\Omega{}) = \{ X(\omega{}) : \omega{} \in{} \Omega{} \} $ 

#### **(2) 종류** : 
##### $ \hspace{0.15cm} $ ① 이산 확률 변수(Discrete Random Variable)
##### $ \hspace{0.15cm} $ ② 연속 확률 변수(Continuous Random Variable)

#### **(3) 예시** : 
##### $ \hspace{0.15cm} $ ① 주사위를 $ 2 $번 던지고 확률 변수 $ X $를 두 주사위 눈의 합으로 정의할 때 그 치역을 구하고자 하는 경우
##### $ \hspace{0.45cm} \Omega{} = \{ (\omega{}_{i}, \omega{}_{k}) \mid{} 1 \leq{} i, k \leq{} 6 \} = \{ (1,1), (1,2), (1,3), \cdots{}, (6,6) \} $
##### $ \hspace{0.45cm} X(\omega{}_{i},\omega{}_{k}) = \omega{}_{1} + \omega{}_{2} $
##### $ \hspace{0.45cm} \therefore{} R_{X} = \{ 2, 3, \cdots{}, 12 \} $
##### $ \hspace{0.15cm} $ ② 동전을 앞면이 나올때까지 던지고 확률 변수 $ Y $를 $ \, i $번째에서 앞면이 나올 경우라고 할 때 그 치역을 구하고자 하는 경우
##### $ \hspace{0.45cm} \Omega{} = \{ H, (T,H), (T,T,H), \cdots{} \} $
##### $ \hspace{0.45cm} Y(\omega{}) = \text{length}(\omega{}) \;\; \text{ where } \, \text{length}(\cdot{}) \, \text{ is tuple-length operator}. $
##### $ \hspace{0.45cm} \therefore{} R_{Y} = \{ 1, 2, 3, 4, 5, \cdots{} \} $
##### $ \hspace{0.15cm} $ ③ 현재 시점부터 운석이 지구에 충돌까지 걸리는 시간(초)을 확률 변수 $ T $라고 할 때 그 치역을 구하고자 하는 경우
##### $ \hspace{0.45cm} \Omega{} = [ 0 , \infty{}) \approx{} \{ 0, 0.1, 0.001, \cdots{} \} $
##### $ \hspace{0.45cm} T(\omega{}) = \omega{} \;\; \text{ where } \, \omega{} \in{} \Omega{} $
##### $ \hspace{0.45cm} \therefore{} R_{T} = [ 0, \infty{} ) $
