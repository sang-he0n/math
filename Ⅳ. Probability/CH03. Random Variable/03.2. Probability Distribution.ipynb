# CH03.2. **Probability Distribution**

> ## **확률 분포(Probability Distribution;Pushforward Measure)**

#### **(1) 정의** : 원 확률 측도 $ P $를 확률 변수 $ X $를 통해 가측 공간 $ (\mathbb{R}, \mathcal{B}(\mathbb{R})\big){} $ 안에서 정의된 확률 측도(measure)
#### $ = $ 확률 변수 $ X $의 값 또는 구간에 대응되는 확률을 정의하는 함수
#### $ \Rrightarrow{} P_{X}(B) := P(X \in{} B) = P \circ{} X^{-1} = P(X^{-1}(B)\big){} \;\; \text{ where } \, B \in{} \mathcal{B}(\mathbb{R}) $ 

##### **(`PLUS`) 모수(Parameter)** : 확률 분포의 위치, 폭, 기울기 등의 형태를 결정하는 값

#### **(2) 종류** : 
##### $ \hspace{0.15cm} $ ① 누적 분포 함수(Cumulative Distribution Function)
##### $ \hspace{0.15cm} $ ② 확률 질량 함수(Probability Mass Function)
##### $ \hspace{0.15cm} $ ③ 확률 밀도 함수(Probability Density Function)
##### $ \hspace{0.15cm} $ ④ 혼합 분포(Mixture distribution)

<b></b>

> ## **누적 분포 함수(Cumulative Distribution Function;CDF)**

#### **(1) 정의** : **[~]**
#### $ \Rrightarrow{} F_{X}(x) := P_{X}\big({}(−∞,x]\big){} = P(X<x) $

#### **(2) 성질** : 어떠한 누적 분포 함수 $ F_{X}(x) $가 존재할 때 아래가 성립함
##### $ \hspace{0.15cm} $ ① $ x_{1} < x_{2} \; \Rightarrow{} \; F_{X}(x_{1}) < F_{X}(x_{2}) \;\; \cdots{} \, \text{monotone non-decreasing} $
##### $ \hspace{0.15cm} $ ② $ \displaystyle{} \lim_{x_{1}\downarrow{}x}F_{X}(x_{1}) = F_{X}(x) \hspace{1.425cm} \;\; \cdots{} \; \text{right-continuous} $ 
##### $ \hspace{0.15cm} $ ③ $ \displaystyle{} \lim_{x\rightarrow{}-\infty{}}F_{X}(x) = 0 $
##### $ \hspace{0.15cm} $ ③ $ \displaystyle{} \lim_{x\rightarrow{}+\infty{}}F_{X}(x) = 1 $

<b></b>

> ## **확률 질량 함수(Probability Mass Function)**

#### **(1) 정의** : 확률 변수의 분포 $ P_{X} $가 가산 측도 $ \mu{}_{c} $에 절대 연속일 때 존재하는 라돈-니코딤 도함수
#### $ = $ 확률 변수의 지지 집합(support set)이 가산 집합인 확률 분포
#### $ \Rrightarrow{} p_{X}(x) := \frac{\text{d}P_{X}}{\text{d}\mu{}_{c}}(x) $
#### $ \hspace{0.45cm} \text{s.t. } \, P_{X} \ll{} \mu{}_{c} \;\; \text{ where } \, \mu{}_{c} \, \text{ is counting measure}. $ 
#### <img src="../img/03.2. Discrete Random Variables (2).png" width="50%" height="50%"></img>

#### **(2) 성질** :
##### $ \hspace{0.15cm} $ ① $ 0 \leq{} p_{X}(x) \leq{} 1 \;\; \forall{}(\text{for all}) \; x $
##### $ \hspace{0.15cm} $ ② $ \displaystyle{} \sum_{x\in{}R_{X}} p_{X}(x) = 1 $
##### $ \hspace{0.15cm} $ ③ $ \displaystyle{} P(X\in{}A) = \sum_{x_{i}\in{}A}p_{X}(x_{i}) \;\; \text{ where } \, A \subset{} R_{X} $
##### $ \hspace{0.15cm} $ ④ $ \displaystyle{} P(X\in{}A) = \sum_{x_{i}\in{}(A\cap{}R_{X})}p_{X}(x_{i}) \;\; \text{ where } \, A \subset{} \mathbb{R} $

<b></b>

> ## **확률 밀도 함수(Probability Density Function;PDF)**

#### **(1) 정의** : 확률 변수의 분포 $ P_{X} $가 르베그 측도에 절대 연속일 때 존재하는 라돈-니코딤 도함수
#### $ = $ 확률 변수의 지지 집합이 불가산 집합인 확률 분포
#### $ \Rrightarrow{} f_{X}(x) := \frac{\text{d}P_{X}}{\text{d}\lambda{}}(x) $
#### $ \hspace{0.45cm} \text{s.t. } \, P_{X} \ll{} \lambda{} \;\; \text{ where } \, \lambda{} \, \text{ is lebesgue measure}. $ 
#### <img src="../img/03.4. Continuous Random Variables (2).png" width="50%" height="50%"></img>

#### **(2) 성질** :
##### $ \hspace{0.15cm} $ ① $ 0 \leq{} f_{X}(x) $
##### $ \hspace{0.15cm} $ ② $ P(X\in{}A) = \int_{x\in{}A}f_{X}(x)\text{d}x \;\; \text{ where } \, A \, \text{ is event}. $
##### $ \hspace{0.15cm} $ ③ $ P(a \leq{} X \leq{} b) = P(a < X < b) = F(X=b) - F(X=a) = \int^{b}_{a}f_{X}(x)\text{d}x $
##### $ \hspace{0.15cm} $ ④ $ \int^{\infty{}}_{-\infty{}} f_{X}(x) \text{d}x = \int_{R_{X}} f_{X}(x) \text{d}x = 1 $

<b></b>

> ## **기대값(Expectation)**

#### **(1) 정의** : 확률 변수 $ X $의 분포 $ P_{X} $를 기준으로 적분한 값
#### $ = $ 확률 변수에 대응되는 확률의 가중 평균(weighted average)
#### $ = $ 실험이나 과정이 무한히 반복할 때 확률 변수의 예상되는 척도
#### $ \Rrightarrow{} \mathbb{E}[X] = \displaystyle{} \int_{\Omega{}} X(\omega{}) \, \text{d} P(\omega{}) = \int_{\mathbb{R}} x \, \text{d} P_{X}(x) $ 
#### $ \hspace{1.275cm} = \begin{cases} \sum_{x_{i}\in{}R_{X}} x_{i} \times{} p_{X}(x_{i}) & \text{if } \, X \, \text{ has pmf } \, p_{X}(x). \\ \int^{\infty{}}_{-\infty{}} \big({} x \times{} f_{X}(x) \big){} \text{d}x & \text{if } \, X \, \text{ has pdf } \, f_{X}(x). \end{cases} $

#### **(2) 성질** (Theorem 3-1) : 확률 변수 $ X, Y $와 상수 $ a, b $가 존재할 때 아래가 성립함
##### $ \hspace{0.15cm} $ ① $ \mathbb{E}[aX+bY] = a\mathbb{E}[X] + b \mathbb{E}[Y] $
##### $ \hspace{0.15cm} $ ② $ X \perp{} Y \; (\text{independent}) \; \Rightarrow{} \; \mathbb{E}[XY] = \mathbb{E}[X]\mathbb{E}[Y] $

<b></b>

> ## (Theorem 3-n) **Law Of The Unconscious Statistician(LOTUS)**

#### 확률 변수 $ X $의 합성 함수 $ g(X) $의 기댓값은 아래와 같이 계산할 수 있다.
#### $ \Rrightarrow{} \mathbb{E}[g(X)] = \displaystyle{} \int_{\mathbb{R}} g(x) \, \text{d} P_{X}(\omega{}) = \begin{cases} \displaystyle{} \sum_{x_{i}\in{}R_{X}} g(x_{i})p_{X}(x_{i}) & \text{if } \, X \, \text{ has pmf } \, p_{X}(x). \\ \displaystyle{} \int^{\infty{}}_{-\infty{}} \big({} g(x) \times{} f_{X}(x) \big){} \text{d}x & \text{if } \, X \, \text{ has pdf } \, f_{X}(x). \end{cases} $

<b></b>

> ## **분산(Variance)**

#### **(1) 정의** : 확률 변수 $ X $의 분포 $ P_{X} $를 기준으로 기댓값으로부터의 제곱 편차를 적분한 값
#### $ = $ 확률 변수 값이 기댓값으로부터 얼마나 퍼져 있는지를 나타내는 척도
#### $ \Rrightarrow{} \text{var}[X] = \mathbb{E}[(X-E[X])^{2}] = \mathbb{E}[X^{2}] - (E[X])^{2} $

#### **(2) 성질** : 확률 변수 $ X, Y $와 상수 $ a, b $가 존재할 때 아래가 성립함
##### $ \hspace{0.15cm} $ ① $ \text{var}[aX+b] = a^{2} \text{var}[X] $
##### $ \hspace{0.15cm} $ ② $ X \perp{} Y \; \Rightarrow{} \; \text{var}[X+Y] = \text{var}[X] + \text{var}[Y] $
