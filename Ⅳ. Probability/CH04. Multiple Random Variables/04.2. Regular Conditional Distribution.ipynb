# CH04.3. **Regular Conditional Distribution**

> ## **정규 조건부 확률 분포(Regular Conditional Probability Distribution)**

#### **(1) 정의** : 확률 변수 $ X, Y $가 존재할 때 $ Y $의 값과 $ X $의 보렐 집합 $ A $를 매핑하는 커널(kernel) 함수
#### $ = $ 확률 변수 $ X $의 분포를 조건 확률 변수 $ Y $의 각 실현되지 않은 값 $ y $에 대해 정의된 함수
#### $ \Rrightarrow{} P_{X|Y} : R_{Y} \times{} \mathcal{B}(R_{X}) \rightarrow{} [0, 1] \; \Leftrightarrow{} \; P_{X|Y}(A | y) := P(X\in{}A|Y = y) $
#### $ \hspace{0.45cm} \text{where } \, X : (\Omega{}, \mathcal{F}, P) \rightarrow{} \big({}\mathbb{R}, \mathcal{B}(R_{X})\big){}, \;\; Y : (\Omega{}, \mathcal{F}, P) \rightarrow{} \big({}\mathbb{R}, \mathcal{B}(R_{Y})\big){} $
#### $ \hspace{0.45cm} \text{and } \, A \in{} \mathcal{B}(R_{X}), \;\; y \in{} R_{Y} $

> ## **(정규) 조건부 기댓값 (Regular Conditional Expectation)**

#### **(1) 정의** : 확률 변수 $ X, Y $가 존재할 때, 정규 조건부 확률분포 $ P_{X|Y}(\cdot | y) $ 위에서 $ X $의 $ 1 $차 모멘트를 정의한 함수
#### $ = $ 확률 변수 $ X $의 평균값을, 아직 실현되지 않은 조건 확률 변수 $ Y $의 각 값 $ y $에 대해 계산하는 함수
#### $ \Rrightarrow{} \mathbb{E}[X|Y] := \displaystyle{} \int_{R_{X}} x P_{X|Y}(\mathrm{d}x | y) $
#### $ \hspace{1.775cm} = \begin{cases} \displaystyle{} \sum_{x\in{}R_{X}} x p_{X|Y}(x | y) & \text{if } \, X, Y \, \text{ have pmf}. \\ \displaystyle{} \int_{{}R_{X}} x f_{X|Y}(x | y) \mathrm{d}x & \text{if } \, X, Y \, \text{ have pdf}. \end{cases} $

#### **(`PLUS`)** 실제 연구(논문)에서 정규 조건부 기댓값, 조건부 기댓값을 거의 구분하지 않고 사용함

#### **(2) 성질** : 확률 변수 $ X, Y, Z $와 상수 $ a, b $가 존재할 때 아래가 성립함
##### $ \hspace{0.15cm} $ ① $ \mathbb{E}[aX+bY|Z] = a\mathbb{E}[X|Z] + b \mathbb{E}[Y|Z] $
##### $ \hspace{0.15cm} $ ② $ X \perp{} Y \; (\text{mutually independent}) \; \Rightarrow{} \; \mathbb{E}[X|Y] = \mathbb{E}[X] $
##### $ \hspace{0.15cm} $ ③ $ \mathbb{E}[g(Y)|Y] = g(Y) $
##### $ \hspace{0.15cm} $ ④ $ \mathbb{E}\big[{}\mathbb{E}[X|Y] \big]{}{} = \mathbb{E}[X] \;\; \because{} $ Theorem 4-n. 

#### **(3) 예시** :
##### $ \hspace{0.15cm} $ ① 공평한 주사위를 $ 1 $회 던진 결과를 확률 변수 $ X $라고 정의할 때 주사위눈이 $ 3 $ 이하일 때 확률 변수 $ X $의 기댓값을 구하면
##### $ \hspace{0.45cm} \Omega{} = \{ 1, 2, 3, 4, 5, 6 \} $
##### $ \hspace{0.45cm} R_{X} = \Omega{} $
##### $ \hspace{0.45cm} R_{Y} = \begin{cases} 1 & \text{if } \, R_{X} \leq{} 3 \\ 0 & \text{otherwise} \end{cases} $
##### $ \hspace{0.45cm} P_{X|Y}(x | y) = \frac{P_{XY}(x, y)}{P_{Y}(y)} = \begin{cases} \frac{P_{XY}(1, y)}{P_{Y}(y)} = \frac{\frac{1}{6}}{\frac{1}{2}} = \frac{1}{3} \\ \frac{P_{XY}(2, y)}{P_{Y}(y)} = \frac{\frac{1}{6}}{\frac{1}{2}} = \frac{1}{3} \\ \frac{P_{XY}(3, y)}{P_{Y}(y)} = \frac{\frac{1}{6}}{\frac{1}{2}} = \frac{1}{3} \end{cases} $
##### $ \hspace{0.45cm} \therefore{} \mathbb{E}[X | Y] = \displaystyle{} \sum^{3}_{x=1} xP_{X|Y}(x | y) = 1 \times{} \frac{1}{3} + 2 \times{} \frac{1}{3} + 3 \times{} \frac{1}{3} = 2 $
##### $ \hspace{0.15cm} $ ② 앞면이 나올 확률이 $ p $이며 동전을 던질 때 확률 변수 $ X $는 앞면이 나올 때까지 던진 횟수이고, 첫 시행 결과에 대한 조건부 확률을 구하면
##### $ \hspace{0.45cm} \Omega{} = \{ (\text{H}), (\text{T,H}), (\text{T,T,H}), \cdots{} \} $
##### $ \hspace{0.45cm} R_{X} = \text{length}(\Omega{}), \;\; X \sim{} \text{Geom}(p) $
##### $ \hspace{0.45cm} R_{Y} = \begin{cases} 1 & \text{if } \, \omega{}_{1} = \text{H} \\ 0 & \text{if } \, \omega{}_{1} = \text{T} \end{cases} \;\; \text{ where } \, (\omega{}_{1}, \omega{}_{2}, \cdots{}) \in{} \Omega{} $ 
##### $ \hspace{0.45cm} P_{X|Y}(x | y) = \begin{cases} 1 & \text{if } \, x = 1 , \; \; y = 1 \\ 0 & \text{if } \, x \neq{} 1 , \; \; y = 1 \\ 0 & \text{if } \, x < 2 , \; \; y = 0 \\ (1-p)^{x-2}p & \text{if } \, x \geq{} 2 , \; \; y = 0 \end{cases} $
##### $ \hspace{0.45cm} \mathbb{E}[X | Y = 1 ] = \displaystyle{} \sum^{\infty{}}_{x=1} xP_{X|Y}(x | 1) = 1 \times{} 1 = 1 $
##### $ \hspace{0.45cm} \mathbb{E}[X | Y = 0 ] = \displaystyle{} \sum^{\infty{}}_{x=2} xP_{X|Y}(x | 0) = \sum^{\infty{}}_{x=2} x (1-p)^{x-2} p = \cdots{} = \frac{1+p}{p} \;\; \because{} (X|Y=0) \sim{} 1+\text{Geom}(p) $
##### $ \hspace{0.45cm} \therefore{} \mathbb{E}[X | Y ] = \begin{cases} 1 & \text{if } \, y = 1 \\ \frac{1+p}{p} & \text{if } \, y = 0 \end{cases} $

> ## **(정규) 조건부 분산(Regular Conditional Variance)**

#### **(1) 정의** : 확률 변수 $ X, Y $가 존재할 때, 정규 조건부 기댓값을 이용해 정의되는 $ Y $-측정가능(measurable)한 확률변수
#### $ = $ 확률 변수 $ Y $의 각 값에 대해, $ X $의 분산을 조건부 평균 기준으로 측정한 함수
#### $ \Rrightarrow{} \text{var}( X | Y ) = \mathbb{E}[X^{2} | Y ] - (\mathbb{E}[X | Y ])^{2} $

<b></b>
