> ## **결합 확률 변수의 독립(Independnece of Joint RV)**

#### **(1) 정의** : 어떠한 확률 변수가 일어날 확률이 다른 확률 변수가 일어날 확률에 영향을 미치지 않는 관계
#### $ \Rrightarrow{} P_{XY}(x, y) = P_{X}(X)P_{Y}(Y) \;\; \Leftrightarrow{} X, Y \text{ are independent}. $

#### **(2) 예시** :
##### $ \hspace{0.15cm} $ ① 결합 확률 질량 함수 $ P_{XY}(x, y) $가 아래의 표로 정의될 때
##### $ \hspace{0.45cm} \begin{array}{|c|c|c|c|} \hline{} P_{XY}(x,y) & y=1 & y=2 & y=5 \\ \hline{} x=0 & 0.1 & 0 & 0.2 \\ \hline{} x=1 & 0.5 & 0.2 & 0 \\ \hline \end{array} $
##### $ \hspace{0.3cm} \cdot{} $ 확률 변수 $ X, Y $가 독립 관계인지 확인하면?
##### $ \hspace{0.45cm} P_{Y}(y) = \begin{cases} 0.6 & \text{if } \,y=1 \\ 0.2 & \text{if } \,y=2 \\ 0.2 & \text{if } \,y=5 \end{cases} $
##### $ \hspace{0.45cm} P_{Y|X}(y \; | \; 0) = \begin{cases} \frac{1}{3} & \text{if } \,y=1 \\ 0 & \text{if } \,y=2 \\ \frac{2}{3} & \text{if } \,y=5 \end{cases} $
##### $ \hspace{0.45cm} P_{Y|X}(1 \; | \; 0) = \frac{1}{3} \neq{} P_{Y}(1) = 0.6 $
##### $ \hspace{0.45cm} \therefore{} X, Y \, \text{ are not independent}. $

> ## **독립 항등 분포(Independent and Identical Distribution;I.I.D)**

#### **(1) 정의** : 확률 변수 $ X_{1}, X_{2}, \cdots{}, X_{n} $가 독립 관계이고 모두 같은 분포를 갖는 관계
#### $ \Rrightarrow{} X_{i} \sim{} \text{i.i.d.} $ 
#### $ \Leftrightarrow{} \text{ s.t.} \, P_{X_{1}}(x_{1}) = P_{X_{2}}(x_{2}) = \cdots{} = P_{X_{n}}(x_{n}) \;\;\;\; \forall{} x \in{} \mathbb{R} $
#### $ \hspace{0.45cm} \text{and } \; P_{X_{1}X_{2}\cdots{}X_{n}}(x_{1}, x_{2}, \cdots{}, x_{n}) = P_{X_{1}}(x_{1}) P_{X_{2}}(x_{2}) \cdots{} P_{X_{n}}(x_{n}) $

<b></b>

> ## **확률 변수들의 합(Sum of Random Variables)**

#### **(1) 정의** : 여러 확률 변수의 선형 결합 중 계수가 $ 1 $로 구성된 확률 변수의 선형 합성 함수
#### $ \Rrightarrow{} Y = g(X_{1}, X_{2}, \cdots{}, X_{n}) = X_{1} + X_{2} + \cdots{} + X_{n} $

#### **(2) 기대값** :
#### $ \Rrightarrow{} E[Y] = E[X_{1}+X_{2}+\cdots{}+X_{n}] = E[X_{1}] + E[X_{2}] + \cdots{} + E[X_{n}] $

#### **(3) 분산** : 
#### $ \Rrightarrow{} \text{var}[Y] = \text{var}[X_{1}+X_{2}+\cdots{}+X_{n}] = \displaystyle{} \sum^{n}_{i=1}\text{var}[X_{i}] + 2 \sum_{1\leq{}i<k\leq{}n}\text{cov}[X_{i}, X_{k}] $

##### **(`WHY?`)**
##### $ \hspace{0.15cm} \text{var}[Y] = \text{var}[X_{1}] + \text{var}[X_{2} + \cdots{} + X_{n}] + 2\text{cov}[X_{1}, X_{2}+\cdots{}+X_{n}] $
##### $ \hspace{1.25cm} = \text{var}[X_{1}] + \text{var}[X_{2}] + \text{var}[X_{3} + \cdots{} + X_{n}] + 2\text{cov}[X_{1}, X_{2}+\cdots{}+x_{n}] + 2\text{cov}[X_{2}, X_{3}+\cdots{}+X_{n}] $
##### $ \hspace{1.35cm} \vdots{} $
##### $ \hspace{1.25cm} = \text{var}[X_{1}] + \cdots{} + \text{var}[X_{n}] + 2\text{cov}[X_{1}, X_{2}+\cdots{}+x_{n}] + \cdots{} + 2\text{cov}[X_{n-1}, X_{n}] $
##### $ \hspace{1.25cm} = \sum^{n}_{i=1}\text{var}[X_{i}] + 2 \sum_{1\leq{}i<k\leq{}n}\text{cov}[X_{i}, X_{k}] $

##### **(`PLUS`) $ X_{i} $가 독립 관계일 때의 분산** : 
##### $ \hspace{0.15cm} \text{var}[Y] = \sum^{n}_{i=1}\text{var}[X_{i}] $

> ## **확률 변수들의 곱(Product of Random Variables)**

#### **(1) 정의** : 여러 확률 변수의 곱 형태로 정의된 비선형 합성 확률 변수
#### $ \Rrightarrow{} Y = g(X_{1}, X_{2}, \cdots{}, X_{n}) = X_{1}X_{2}\cdots{}X_{n} $

#### **(2) 기대값** :
#### $ \Rrightarrow{} E[Y] = E[X_{1}X_{2}\cdots{}X_{n}] $

#### **(`PLUS`)** $ X_{i} $가 독립 관계일 때의 기댓값 : 
##### $ \hspace{0.15cm} \Rrightarrow{} E[X_{1}X_{2}\cdots{}X_{n}] = E[X_{1}]E[X_{2}]\cdots{}E[X_{n}] $ 

##### **(`WHY?`)**
##### $ \hspace{0.15cm} \text{if } \,X, Y \, \text{ is independent}. $
##### $ \hspace{0.3cm} E[XY] = \sum_{(x_{i}, y_{k})\in{}(R_{X}\times{}R_{Y})} x_{i}y_{k}P_{XY}(x_{i}, y_{k}) $
##### $ \hspace{1.5cm} = \sum_{X_{i}\in{}R_{X}} \sum_{y_{k}\in{}R_{Y}} x_{i}y_{k}P_{X}(x_{i})P_{Y}(y_{k}) $
##### $ \hspace{1.5cm} = \sum_{X_{i}\in{}R_{X}} x_{i} P_{X}(x_{i}) \sum_{y_{k}\in{}R_{Y}} y_{k} P_{Y}(y_{k}) $
##### $ \hspace{1.5cm} = E[X]E[Y] $
##### $ \hspace{0.15cm} \therefore{} E[XY] = E[X]E[Y] $

#### **(3) 분산** : 
#### $ \Rrightarrow{} \text{var}[Y] = \text{var}[X_{1}X_{2}\cdots{}X_{n}] = E[X_{1}^{2}X_{2}^{2}\cdots{}X_{n}^{2}] - (E[X_{1}X_{2}\cdots{}X_{n}])^{2} $
