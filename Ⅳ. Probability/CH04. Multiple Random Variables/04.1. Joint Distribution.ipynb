# CH00.3. **Probability Ⅲ**

> ## **결합 확률 분포(Joint probability distribution)**

#### **(1) 정의** : **[~]** 
#### $ \Rrightarrow{} P_{X_{1},\cdots{},X_{n}}(x_{1},\cdots{},x_{n}) = \begin{cases} F_{X_{1},\cdots{},X_{n}}(x_{1},\cdots{},x_{n}) = P(X_{1}\leq{}x_{1},\cdots{},X_{n}\leq{}x_{n}) & \text{if } \, X \, \text{ has joint cdf }. \\ p_{X_{1},\cdots{},X_{n}}(x_{1},\cdots{},x_{n}) =  P(X_{1}=x_{1},\cdots{}, X_{n}=x_{n}) & \text{if } \, X \, \text{ has joint pmf }. \\ f_{X_{1},\cdots{},X_{n}}(x_{1},\cdots{},x_{n}) = \frac{\partial{}^{n}}{\partial{}x_{1} \partial{}x_{2} \cdots{} \partial{}x_{n}} F_{X_{1},\cdots{},X_{n}}(x_{1},\cdots{},x_{n}) & \text{if } \, X \, \text{ has joint pdf }. \end{cases} $

> ## **주변 확률 분포(Marginal probability distribution)**

#### **(1) 정의** : **[~]**
#### $ \Rrightarrow{} P_{X_{1}}(x_{1}) = \begin{cases} F_{X_{1}}(x_{1}) = \displaystyle{} \sum_{x_{2}}\cdots{}\sum_{x_{n}} P(X_{1}\leq{}x_{1},\cdots{},X_{n}\leq{}x_{n}) & \text{if } \, X \, \text{ has joint cdf }. \\ p_{X_{1}}(x_{1}) =  \displaystyle{} \sum_{x_{2}}\cdots{}\sum_{x_{n}} p_{X_{1},\cdots{},X_{n}}(x_{1},  \cdots{}, x_{n}) & \text{if } \, X \, \text{ has joint pmf }. \\ f_{X_{1}}(x_{1}) = \displaystyle{} \int^{\infty{}}_{-\infty{}}\cdots{}\int^{\infty{}}_{-\infty{}} f_{X_{1},\cdots{},X_{n}}(x_{1},  \cdots{}, x_{n}) \text{d}x_{2:n} & \text{if } \, X \, \text{ has joint pdf }. \end{cases} $
