# CH04.5. **Multivariate Random Vectors**

> ## **랜덤 벡터(Random vector)**

#### **(1) 정의** : $ m $개의 확률 변수로 구성된 벡터
#### $ \Rrightarrow{} \textbf{X} := \begin{bmatrix} X_{1} \\ X_{2} \\ \vdots{} \\ X_{m} \end{bmatrix} $ 
#### $ \hspace{0.45cm} \text{where } \, \textbf{X} : (\Omega{}, \mathcal{F}, P) \rightarrow{} (\mathbb{R}^{m}, \mathcal{B}(\mathbb{R}^{m})) $

<b></b>

> ## **랜덤 벡터의 기댓값(Expectation of random vector)**

#### **(1) 정의** : 각 확률 변수의 기대값으로 정의된 벡터
#### $ \Rrightarrow{} \mathbb{E}[\textbf{X}] = \begin{bmatrix} \mathbb{E}[X_{1}] \\ \mathbb{E}[X_{2}] \\ \vdots{} \\ \mathbb{E}[X_{m}] \end{bmatrix} $

#### **(2) 성질** : 고정 행렬 $ A, C $ 및 고정 벡터 $ \textbf{b} $와 랜덤 벡터 $ \textbf{X}, \textbf{Y} $이 존재할 때 다음이 성립함
##### $ \hspace{0.15cm} $ ① $ \mathbb{E}[A\textbf{X}+\textbf{b}] = A \, \mathbb{E}[\textbf{X}] + \textbf{b} $
##### $ \hspace{0.15cm} $ ② $ \mathbb{E}[(A\textbf{X})(C\textbf{Y})^{T}] = A \, \mathbb{E}[\textbf{X}\textbf{Y}^{T}] C^{T} $

<b></b>

> ## **랜덤 벡터의 공분산(Covariance of random vector)**

#### **(1) 정의** : 두 랜덤 벡터 $ \textbf{X}, \textbf{Y} $의 각 성분 쌍(pair)에 대한 공분산으로 정의된 행렬
#### $ \Rrightarrow{} \text{cov}(\textbf{X}, \textbf{Y}) = \mathbb{E}\big[{} (\textbf{X}-\mathbb{E}[\textbf{X}])(\textbf{Y}-\mathbb{E}[\textbf{Y}])^{T} \big]{} $
#### $ \hspace{2.1cm} = \begin{bmatrix} \text{cov}(X_{1}, Y_{1}) & \text{cov}(X_{1}, Y_{2}) & \cdots{} & \text{cov}(X_{1}, Y_{m}) \\ \text{cov}(X_{2}, Y_{1}) & \text{cov}(X_{2}, Y_{2}) & \cdots{} & \text{cov}(X_{2}, Y_{m}) \\ \vdots{} & \vdots{} & \ddots{} & \vdots{} \\ \text{cov}(X_{m}, Y_{1}) & \text{cov}(X_{m}, Y_{2}) & \cdots{} & \text{cov}(X_{m}, Y_{m}) \end{bmatrix} $

#### **(2) 성질** : 고정 행렬 $ A, C $ 및 고정 벡터 $ \textbf{b}, \textbf{d} $와 랜덤 벡터 $ \textbf{X}, \textbf{Y}, \textbf{W}, \textbf{Z} $이 존재할 때 다음이 성립함
##### $ \hspace{0.15cm} $ ① $ \text{cov}(\textbf{X}, \textbf{X}) = \text{var}(\textbf{X}) $
##### $ \hspace{0.15cm} $ ② $ \text{cov}(\textbf{X}, \textbf{Y}) = \mathbb{E}[\textbf{X}\textbf{Y}^{T}] - \mathbb{E}[\textbf{X}](\mathbb{E}[\textbf{Y}])^{T} $
##### $ \hspace{0.15cm} $ ③ $ \text{cov}(\textbf{X}, \textbf{Y}) = \big({} \text{cov}(\textbf{Y}, \textbf{X}) \big){}^{T} $
##### $ \hspace{0.15cm} $ ④ $ \text{cov}(A\textbf{X} + \textbf{b}, C\textbf{Y} + \textbf{d}) = A \, \text{cov}(\textbf{X},\textbf{Y})C^{T} $
##### $ \hspace{0.15cm} $ ⑤ $ \text{cov}(\textbf{X} + \textbf{W}, \textbf{Y} + \textbf{Z}) = \text{cov}(\textbf{X}, \textbf{Y}) + \text{cov}(\textbf{X}, \textbf{Z}) + \text{cov}(\textbf{W}, \textbf{Y}) + \text{cov}(\textbf{W}, \textbf{Z}) $ 

<b></b>

> ## (Theorem 1-1) **Symmetric Positive Semidefinite of Covariance**

#### **(1)** 공분산 행렬은 대칭 행렬(symmetric matrix)이다.
#### $ \Rrightarrow{} \text{cov}(\textbf{X}, \textbf{X}) = \big({} \text{cov}(\textbf{X}, \textbf{X}) \big){}^{T} $

#### **(2)** 공분산 행렬은 양의 준정부호 행렬(positive semi-definite;PSD)이다.
#### $ \Rrightarrow{} \textbf{x}^{T} \cdot{} \text{cov}(\textbf{X}, \textbf{X}) \cdot{} \textbf{x} \geq{} 0 $

##### **(`WHY?`)**
##### $ \hspace{0.15cm}{} \textbf{x}^{T} \cdot{} \text{cov}(\textbf{X}, \textbf{X}) \cdot{} \textbf{x} = \textbf{x}^{T} \cdot{} \mathbb{E}\big[{} (\textbf{X}-\mathbb{E}[\textbf{X}])(\textbf{X}-\mathbb{E}[\textbf{X}])^{T} \big]{} \cdot{} \textbf{x} $
##### $ \hspace{2.935cm} = \mathbb{E}\big[{} \textbf{x}^{T} \cdot{} (\textbf{X}-\mathbb{E}[\textbf{X}])(\textbf{X}-\mathbb{E}[\textbf{X}])^{T} \cdot{} \textbf{x} \big]{} $
##### $ \hspace{2.935cm} = \mathbb{E}\big[{} \big({} \textbf{x}^{T} \cdot{} (\textbf{X}-\mathbb{E}[\textbf{X}]) \big){} \big({} \textbf{x}^{T} \cdot{} (\textbf{X}-\mathbb{E}[\textbf{X}]) \big){}^{T} \big]{} $
##### $ \hspace{2.935cm} = \mathbb{E}\big[{} \big({} \textbf{x}^{T} \cdot{} (\textbf{X}-\mathbb{E}[\textbf{X}]) \big){}^{2} \big]{} $ 
##### $ \hspace{2.935cm} \geq{} 0 $
