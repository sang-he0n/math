# CH05.1. **Moment Generating Function**

> ## **$ n $차 적률($ n $-th Moment)**

#### **(1) 정의** : 확률 변수 $ X $의 $ n $차 기대값
#### $ \Rrightarrow{} \mathbb{E}[X^{n}] $

<b></b>

> ## **$ n $차 중심 적률($ n $-th Central Moment)**

#### **(1) 정의** : 확률 변수 $ X $의 평균(mean)부터 멀어진 거리의 $ n $차 기대값
#### $ \Rrightarrow{} \mathbb{E}[(X-\mathbb{E}[X])^{n}] $

#### **(2) 의미** : 
##### $ \hspace{0.15cm} $ ① $ 2 $차 중심 적률 : 분산(variance)
##### $ \hspace{0.45cm} \Rrightarrow{} \mathbb{E}[(X-\mathbb{E}[X])^{2}] $
##### $ \hspace{0.15cm} $ ② 정규화 $ 3 $차 중심 적률 : 데이터의 비대칭성으로 정의되는 왜도(skewness)
##### $ \hspace{0.45cm} \Rrightarrow{} \frac{\mathbb{E}[(X-\mathbb{E}[X])^{3}]}{\sigma{}^{3}} \;\; \text{ where } \, \sigma{} \, \text{ is standard deviation}. $
##### $ \hspace{0.15cm} $ ③ 정규화 $ 4 $차 중심 적률 : 데이터의 꼬리 두꺼움 및 뾰족함으로 정의되는 첨도(kurtosis)
##### $ \hspace{0.45cm} \Rrightarrow{} \frac{\mathbb{E}[(X-\mathbb{E}[X])^{4}]}{\sigma{}^{4}} $

<b></b>

> ## (Theorem 5-1) **Existence of Moments**

#### $ k $차 적률이 존재하고 유한할 때, $ \, 0 $보다 크고 $ k $보다 낮은 $ l $차 절댓값 적률 역시 존재한다. 
#### $ \Rrightarrow{} \mathbb{E}[|X|^{l}] < \infty{} \;\; \text{ s.t. } \, \mathbb{E}[|X|^{k}] < \infty{}, \;\; 0 \leq{} l < k $

<b></b>

> ## **적률 생성 함수(Moment Generating Function; MGF)**

#### **(1) 정의** : 확률 변수의 적률이 존재하는 범위 내에서, 그 모든 적률을 생성하는 함수
#### $ = $ 어떠한 실수 구간에서 확률 변수의 분포를 특징짓는 함수
#### $ \Rrightarrow{} M_{X}(s) = \mathbb{E}[e^{sX}] \;\; \text{ s.t. } \, \forall{} \, s \in{} (-s_{0}, s_{0}), \;\; s_{0} > 0 $

##### **(`WHY?`)**
##### $ \hspace{0.15cm} e^{sX} = \sum^{\infty{}}_{k=0} \frac{(sX)^{k}}{k!} = \sum^{\infty{}}_{k=0} \frac{s^{k}X^{k}}{k!} \;\; \because{} e^{x} = 1 + x + \frac{x^{2}}{2!} + \cdots{} = \sum^{\infty{}}_{k=0} \frac{x^{k}}{k!} $
##### $ \hspace{0.15cm} M_{X}(s) = \mathbb{E}[e^{sX}] = \mathbb{E}[\sum^{\infty{}}_{k=0} \frac{s^{k}X^{k}}{k!}] = \sum^{\infty{}}_{k=0} \frac{s^{k}}{k!} \mathbb{E}[X^{k}] $
##### $ \hspace{1.3cm} = \mathbb{E}[1] + s\mathbb{E}[X] + \frac{s^{2}}{2!} \mathbb{E}[X^{2}] + \frac{s^{3}}{3!} \mathbb{E}[X^{3}] +\cdots{} $
##### $ \hspace{0.15cm} \frac{\mathrm{d}}{\mathrm{d}s} M_{X}(s) = M'_{X}(s) = \mathbb{E}[X] + \frac{2s}{2!}\mathbb{E}[X^{2}] + \frac{3s^{2}}{3!}\mathbb{E}[X^{3}] + \cdots{} $
##### $ \hspace{0.15cm} M'_{X}(0) = \mathbb{E}[X] $
##### $ \hspace{0.15cm} M''_{X}(s) = \frac{2}{2!}\mathbb{E}[X^{2}] + \frac{3\cdot{}2\cdot{}s}{3!}\mathbb{E}[X^{3}] + \cdots{} $
##### $ \hspace{0.15cm} M''_{X}(0) = \mathbb{E}[X^{2}] $
##### $ \hspace{0.45cm} \vdots{} $
##### $ \hspace{0.15cm} \therefore{} M^{(k)}_{X}(0) = \mathbb{E}[X^{k}] $

#### **(2) 확률 변수 별 계산** : 
##### $ \hspace{0.15cm} $ ① 이산 확률 변수 : 
##### $ \hspace{0.45cm} \Rrightarrow{} M_{X}(s) = \sum_{x_{i}\in{}X(s)} e^{sx}P_{X}(x_{i}) $
##### $ \hspace{0.15cm} $ ② 연속 확률 변수 : 
##### $ \hspace{0.45cm} \Rrightarrow{} M_{X}(s) = \int^{\infty{}}_{-\infty{}} e^{sx}f_{X}(x) \mathrm{d}x $

<b></b>

> ## (Theorem 5-2) **Uniqueness of MGF**

#### 두 확률 변수의 적률 생성 함수가 반드시 존재하며 같을 경우, 두 확률 변수는 같은 분포를 갖는다. 
#### $ \Rrightarrow{} X \stackrel{\mathrm{d}}{=} Y \;\; \text{ s.t. } \, M_{X}(s), M_{Y}(s) \, \text{ exist}, \;\; M_{X}(s) = M_{Y}(s) \;\; \forall{} \, s \in{} (-s_{0}, s_{0}) $
